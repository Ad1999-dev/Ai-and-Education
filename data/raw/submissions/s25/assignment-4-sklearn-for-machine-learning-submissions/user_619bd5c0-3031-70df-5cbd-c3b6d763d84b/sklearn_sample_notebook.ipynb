```python
# Imports section
import pandas as pd              
import numpy as np                
import sklearn as sk
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
df = pd.read_csv('science_data_large.csv')  

# Output the first 15 rows of the data
print(df.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
print(df.describe())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.drop('Size nm^3', axis=1)
y = df['Size nm^3']


# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

model = sk.linear_model.LinearRegression()
model.fit(X_train, y_train)

sample = pd.DataFrame([[0.5, 150]], columns=X_train.columns)    # Replace with appropriate sample values
prediction = model.predict(sample)
print("Sample Prediction:", prediction)

score = model.score(X_test, y_test)
print("Model Score:", score)

coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
scores = sk.model_selection.cross_val_score(model, X, y, cv=sk.model_selection.KFold(n_splits=5, random_state=42, shuffle=True))
print("Cross-validation scores:", scores)
print("Average score:", np.mean(scores))
print("Standard deviation:", np.std(scores))

# Report on their finding and their significance
```

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.
poly = sk.preprocessing.PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

X_train_poly, X_test_poly, y_train_poly, y_test_poly = sk.model_selection.train_test_split(
    X_poly, y, test_size=0.1, random_state=42
)

model_poly = sk.linear_model.LinearRegression()
model_poly.fit(X_train_poly, y_train_poly)

scores_poly = sk.model_selection.cross_val_score(
    model_poly, X_poly, y, cv=sk.model_selection.KFold(n_splits=5, random_state=42, shuffle=True)
)

print("Cross-validation scores:", scores_poly)
print("Average score:", np.mean(scores_poly))
print("Standard deviation:", np.std(scores_poly))
print("Intercept:", model_poly.intercept_)
print("Coefficients:", model_poly.coef_)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

df_ckd = pd.read_csv('ckd_feature_subset.csv')
print(df_ckd.head())


X_ckd = df_ckd.drop('Target_ckd', axis=1)
y_ckd = df_ckd['Target_ckd']

# Define the classification models using the sk alias
models = {
    "Logistic Regression": sk.linear_model.LogisticRegression(random_state=42, max_iter=1000),
    "Support Vector Machine": sk.svm.SVC(random_state=42),
    "k-Nearest Neighbors": sk.neighbors.KNeighborsClassifier(),
    "Neural Network": sk.neural_network.MLPClassifier(random_state=42, max_iter=1000)
}

# Set up 5-fold cross-validation with a fixed random seed
kf = sk.model_selection.KFold(n_splits=5, random_state=42, shuffle=True)

# Evaluate each model and collect cross-validation metrics
results = {}
for name, model in models.items():
    scores = sk.model_selection.cross_val_score(model, X_ckd, y_ckd, cv=kf)
    results[name] = (np.mean(scores), np.std(scores))
    print(f"{name}: Mean Accuracy = {np.mean(scores):.4f}, Std Deviation = {np.std(scores):.4f}")
```

## Results and Conclusion for Classification Experiments

### Classification Model Performance on the CKD Dataset

The following table summarizes the cross-validation results for the four classification models evaluated on the CKD dataset:

| Model                   | Mean Accuracy | Standard Deviation |
|-------------------------|---------------|--------------------|
| Logistic Regression     | 0.8566        | 0.0663             |
| Support Vector Machine  | 0.9282        | 0.0476             |
| k-Nearest Neighbors     | 0.9280        | 0.0524             |
| Neural Network          | 0.9351        | 0.0405             |

**Interpretation:**

- The **Neural Network** achieved the highest mean accuracy (0.9351) with the lowest standard deviation (0.0405), indicating not only superior predictive performance but also very consistent results across the folds.
- Both the **Support Vector Machine** and **k-Nearest Neighbors** models performed well, with mean accuracies of around 0.928 and moderate consistency.
- **Logistic Regression** showed the lowest performance among the four, with a mean accuracy of 0.8566 and a slightly higher standard deviation.

Overall, the neural network appears to be the best performing model on this CKD dataset, suggesting that its ability to capture non-linear patterns may be advantageous in predicting CKD status.


```python
# Experiments with Neural Network.
nn_configs = {
    "NN Config 1 (50 neurons)": sk.neural_network.MLPClassifier(hidden_layer_sizes=(50,), random_state=42, max_iter=1000),
    "NN Config 2 (100 neurons)": sk.neural_network.MLPClassifier(hidden_layer_sizes=(100,), random_state=42, max_iter=1000),
    "NN Config 3 (50, 50 neurons)": sk.neural_network.MLPClassifier(hidden_layer_sizes=(50,50), random_state=42, max_iter=2000)
}

nn_results = {}
for name, model in nn_configs.items():
    scores = sk.model_selection.cross_val_score(model, X_ckd, y_ckd, cv=kf)
    nn_results[name] = (np.mean(scores), np.std(scores))
    print(f"{name}: Mean Accuracy = {np.mean(scores):.4f}, Std Deviation = {np.std(scores):.4f}")
```

## Results and Conclusion for Neural Network Experiments

### Neural Network Configuration Performance

| Configuration               | Mean Accuracy | Standard Deviation |
|-----------------------------|---------------|--------------------|
| NN Config 1 (50 neurons)    | 0.9351        | 0.0405             |
| NN Config 2 (100 neurons)   | 0.9351        | 0.0405             |

**Interpretation:**

- **NN Config 1 (50 neurons)** and **NN Config 2 (100 neurons)** achieved identical performance, indicating that simply increasing the number of neurons in a single hidden layer from 50 to 100 did not improve classification accuracy for this dataset.
- **NN Config 3 (50, 50 neurons)** outperformed the single-layer configurations, reaching a higher mean accuracy (0.9546). This suggests that a deeper architecture with two hidden layers may better capture the underlying patterns in the CKD data.
