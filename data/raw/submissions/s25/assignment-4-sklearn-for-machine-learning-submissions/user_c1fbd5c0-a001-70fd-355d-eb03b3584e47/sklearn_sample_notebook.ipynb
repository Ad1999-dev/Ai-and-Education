## 1. Loading the dataset

```python
# Using pandas load the dataset

data = pd.read_csv('science_data_large.csv')

# Output the first 15 rows of the data

print(data.head(15))

# Display a summary of the table information (number of datapoints, etc.)

data.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = data[['Temperature °C', 'Mols KCL']]
y = data['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1 , random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

model = sk.linear_model.LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model

sample_datapoint = pd.DataFrame([[413, 612]], columns=['Temperature °C', 'Mols KCL'])
predicted_output = model.predict(sample_datapoint)
print(predicted_output)

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

model_score = model.score(X_test, y_test)
print(model_score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

coefficients = model.coef_
intercept = model.intercept_

print(coefficients)
print(intercept)
```

Write the linear equation of a slime: (example equation: $E = mc^2$)

$h(x) = -409391.47958 + 866.14641 * x_1 + 1032.69506 * x_2$

Report on score and explain meaning:

By default, the score method for LinearRegression calculates the R^2 value to score the model. The R^2 value divides the residual sum of squares (the variance of the target variable not accounted for by the model) b the total sum of squares (the total variance of the target variable) and subtracts it from 1. This means that the closer the R^2 value is to 1, the less the fraction is, which translates to less variance of the target variable not account for by the model and thus a more accurate prediction. The model above has an R^2 value of 0.85524, which is overall pretty good. 

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores = sk.model_selection.cross_val_score(model, X, y, cv=kf)  

print(scores)

# Report on their finding and their significance
```

Write findings here:

The R^2 value across shuffles of data remains between 0.82 to 0.89, so within +/- 0.04 range of the score calculated in 1.5. This range is relatively narrow, which suggests that the high score in 1.3 was not a lucky outlier but rather than the model is stable and consistently performs well in predicting the target value. 

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

poly = sk.preprocessing.PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model_poly = sk.linear_model.LinearRegression()
model_poly.fit(X_poly,y)

kf_poly = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_poly = sk.model_selection.cross_val_score(model_poly, X_poly, y, cv=kf_poly)  

print(scores_poly)

coefficients_poly = model_poly.coef_
intercept_poly = model_poly.intercept_

print(coefficients_poly)
print(intercept_poly)

# Report on the metrics and output the resultant equation as you did in Part 3.
```

## Classification Model Evaluation

```python
# Load the dataset. Then train and evaluate the classification models.

data = pd.read_csv('ckd_feature_subset.csv')

X = data[['age', 'bp', 'wbcc', 'appet_poor', 'appet_good', 'rbcc']]
y = data['Target_ckd']

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1 , random_state=42)
```

### Logistic Regression

```python
model_logistic = sk.linear_model.LogisticRegression()
model_logistic.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_logistic = sk.model_selection.cross_val_score(model_logistic, X, y, cv=kf, scoring='accuracy')

mean_logistic = np.mean(scores_logistic)
std_logistic = np.std(scores_logistic)

print(mean_logistic)
print(std_logistic)
```

### Support Vector Machines

```python
model_svc = sk.svm.SVC()
model_svc.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_svc = sk.model_selection.cross_val_score(model_svc, X, y, cv=kf, scoring='accuracy')

mean_svc = np.mean(scores_svc)
std_svc = np.std(scores_svc)

print(mean_svc)
print(std_svc)
```

### K-Nearest Neighbors

```python
model_neighbors = sk.neighbors.KNeighborsClassifier()
model_neighbors.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_neighbors = sk.model_selection.cross_val_score(model_neighbors, X, y, cv=kf, scoring='accuracy')

mean_neighbors = np.mean(scores_neighbors)
std_neighbors = np.std(scores_neighbors)

print(mean_neighbors)
print(std_neighbors)
```

### Neural Networks

```python
model_nn = sk.neural_network.MLPClassifier()
model_nn.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_nn = sk.model_selection.cross_val_score(model_nn, X, y, cv=kf, scoring='accuracy')

mean_nn = np.mean(scores_nn)
std_nn = np.std(scores_nn)

print(mean_nn)
print(std_nn)
```

Both had very similar mean accuracies and standard deviations, with the SVC model being slightly more accurate with a lower standard deviation. This is likely because SVC models are more effective then others when working with a dataset with a relatively small amount of samples and a large amount of features, as well as a dataset that has non-linear relationships between features. These characteristics are present in the ckd_feature_subset dataset. 

## Experiments with Neural Networks

### Hidden Layer Sizes

```python
model_nn_layer = sk.neural_network.MLPClassifier(max_iter=1000)
model_nn_layer.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_nn_layer = sk.model_selection.cross_val_score(model_nn_layer, X, y, cv=kf, scoring='accuracy')

mean_nn_layer = np.mean(scores_nn_layer)
std_nn_layer = np.std(scores_nn_layer)

print(mean_nn_layer)
print(std_nn_layer)
```

### Solver

```python
model_nn_solver = sk.neural_network.MLPClassifier(solver='lbfgs')
model_nn_solver.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_nn_solver = sk.model_selection.cross_val_score(model_nn_solver, X, y, cv=kf, scoring='accuracy')

mean_nn_solver = np.mean(scores_nn_solver)
std_nn_solver = np.std(scores_nn_solver)

print(mean_nn_solver)
print(std_nn_solver)
```

### Activation

```python
model_nn_activation = sk.neural_network.MLPClassifier(activation='tanh')
model_nn_activation.fit(X_train, y_train)

kf = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
scores_nn_activation = sk.model_selection.cross_val_score(model_nn_activation, X, y, cv=kf, scoring='accuracy')

mean_nn_activation = np.mean(scores_nn_activation)
std_nn_activation = np.std(scores_nn_activation)

print(mean_nn_activation)
print(std_nn_activation)
```
