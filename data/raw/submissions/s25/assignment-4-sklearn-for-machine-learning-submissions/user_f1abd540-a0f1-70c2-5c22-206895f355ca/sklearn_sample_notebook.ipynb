```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import PolynomialFeatures
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
sdl = pd.read_csv("science_data_large.csv")
print(sdl.head(15))
sdl.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X = sdl.drop(columns=['Size nm^3']) 
y = sdl['Size nm^3']                 
# Split the dataset into training and testing sets (90% train, 10% test)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Display the shapes of the resulting datasets
print("Training set size:", x_train,y_train)
print("Test set size:", x_test,y_test)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
model = LinearRegression()
model.fit(x_train, y_train)

# Score the model
score = model.score(x_test, y_test)

# Create a sample datapoint for predictions
sample_data = pd.DataFrame([[25, 1]], columns=['Temperature °C', 'Mols KCL'])  # Example: Temperature = 25°C, Mols KCL = 1 
predicted_output = model.predict(sample_data)

# Display the score and the predicted output
print("Model Score:", score)
print("Predicted Output for Sample Data Point:", predicted_output)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

# Report the cross-validation scores
print("Cross-validation R^2 scores:", cv_scores)
print("Mean R^2 score:", cv_scores.mean())
print("Standard deviation of R^2 scores:", cv_scores.std())
```

Write findings here: the outcome of cross validation suggsts that the linear model is effeective in predicting size, has high mean R^2 score and a low deviation.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
model_poly = LinearRegression()

# Define the KFold cross-validation strategy
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores_poly = cross_val_score(model_poly, X_poly, y, cv=kf, scoring='r2')

# Report the cross-validation scores
print("Cross-validation R^2 scores (Polynomial Features):", cv_scores_poly)
print("Mean R^2 score (Polynomial Features):", cv_scores_poly.mean())
print("Standard deviation of R^2 scores (Polynomial Features):", cv_scores_poly.std())
model_poly.fit(X_poly, y)

# Extract coefficients and intercept
coefficients_poly = model_poly.coef_
intercept_poly = model_poly.intercept_

# Print coefficients and intercept
print("Coefficients (Polynomial Features):", coefficients_poly)
print("Intercept (Polynomial Features):", intercept_poly)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier  # Import MLPClassifier
from sklearn.metrics import accuracy_score

# Load the cleaned CKD dataset
ckd_data = pd.read_csv("ckd_feature_subset.csv")

# Check for null values
print(ckd_data.isnull().sum())

# Define features and target variable
X = ckd_data.drop(columns=['Target_ckd'])  # Assuming 'Target_ckd' is the correct column name
y = ckd_data['Target_ckd']  # Convert to binary

# Standardizing the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Function to evaluate models using cross-validation
def evaluate_model(model, X, y):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X, y, cv=kf)
    return scores.mean(), scores.std()

# Step 1: Logistic Regression
logistic_model = LogisticRegression(max_iter=1000)
logistic_accuracy, logistic_std = evaluate_model(logistic_model, X, y)

# Step 2: Support Vector Machines (SVC)
svc_model = SVC()
svc_accuracy, svc_std = evaluate_model(svc_model, X, y)

# Step 3: k-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier()
knn_accuracy, knn_std = evaluate_model(knn_model, X, y)

# Step 4: Neural Network using Multi-layer Perceptron (MLP)
mlp_model = MLPClassifier(hidden_layer_sizes=(64,), max_iter=1000, random_state=42)  # One hidden layer, 64 neurons
mlp_accuracy, mlp_std = evaluate_model(mlp_model, X, y)

# You can also evaluate different configurations of the MLP
mlp_config2 = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=42)  # Two hidden layers
mlp_config2_accuracy, mlp_config2_std = evaluate_model(mlp_config2, X, y)

mlp_config3 = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=1000, random_state=42)  # Two hidden layers, more neurons
mlp_config3_accuracy, mlp_config3_std = evaluate_model(mlp_config3, X, y)

# Compile results into a DataFrame
results = {
    'Model': ['Logistic Regression', 'SVC', 'KNN', 'MLP Config 1', 'MLP Config 2', 'MLP Config 3'],
    'Average Accuracy': [logistic_accuracy, svc_accuracy, knn_accuracy, mlp_accuracy, mlp_config2_accuracy, mlp_config3_accuracy],
    'Standard Deviation': [logistic_std, svc_std, knn_std, mlp_std, mlp_config2_std, mlp_config3_std]
}

results_df = pd.DataFrame(results)
print(results_df)
```

## Results and Conclusion for Classification Experiments

SVC had the highet average of 96.7% in predicting CKD and a standard deviation of 0.0293.This method is highly effective in classifying if patient has CKD or not. Logistic regression had a 94.79% accuracy in predicting and KNN had a 92.15% accuracy in predicting CKD. SVC uses different kernel functions to capture more complex relations, leads to better performance on datasets with non-linear relationships.

## Results and Conclusion for Neural Network Experiments

While the MLP configurations performed well, they did not exceed the SVC’s accuracy. The neural network's performance could potentially improve with additional tuning of hyperparameters, increased depth, more neurons, or more training epochs.
- neural networks can take longer to train and may require careful adjustment of learning rates and architecture to achieve optimal performance compared to the other models evaluated. The more number of layers and greater number of neurons yeilded better results.
