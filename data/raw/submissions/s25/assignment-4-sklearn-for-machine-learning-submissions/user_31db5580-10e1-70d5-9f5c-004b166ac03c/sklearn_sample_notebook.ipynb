```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression  
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
df = pd.read_csv('science_data_large.csv')  
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.drop(columns=['Size nm^3'])  
y = df['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression().fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_point = pd.DataFrame([[469,647]], columns=X.columns)

prediction = model.predict(sample_point)
print("Predicted Output:", prediction)

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print("Model Score:", score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
equation = f"h(x) = {intercept:.4f} + " + " + ".join([f"{coef:.4f}x_{i}" for i, coef in enumerate(coefficients, start=1)])
print(equation)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cross_score = cross_val_score(model, X, y, cv=kf)

# Report on their finding and their significance
print("Cross-val R² scores:", cross_score)
print(f"Mean R² Score: {cross_score.mean():.4f}")
print(f"Standard Deviation: {cross_score.std():.4f}")
```

Write findings here:


- The Mean $R^2$ score represents that the model explains 85.97% of the variance in the target variable on average across the 5 cross-validation folds. Therefore, it is a good fit to data, but can be better.
- The standard deviation of 0.0184 shows good variability in the model’s performance. The relatively low standard deviation suggests that the model’s performance is constant over several subsets of data 

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures

# Create polynomial regression model (degree 2)
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X_train, y_train)

# Cross-validation for polynomial regression
kf = KFold(n_splits=5, shuffle=True, random_state=42)
poly_cv_scores = cross_val_score(poly_model, X, y, cv=kf)
print("Polynomial Regression CV Scores:", poly_cv_scores)
print("Mean Polynomial CV Score:", poly_cv_scores.mean())
print("Standard Deviation:", poly_cv_scores.std())

score_poly = poly_model.score(X_test, y_test)
print("Model Score:", score_poly)

coeff = poly_model.named_steps['linearregression'].coef_
inter = poly_model.named_steps['linearregression'].intercept_

print("Intercept:", inter)
print("Coefficients:", coeff)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
df = pd.read_csv('ckd_feature_subset.csv')
X = df.drop(columns=['Target_ckd'])  
y = df['Target_ckd']

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Support Vector Machines": SVC(),
    "k-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Networks": MLPClassifier(max_iter=1000)  
}

results = {}

for model_name, model in models.items():
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cross_score = cross_val_score(model, X, y, cv=kf)
    results[model_name] = {
        "Mean Accuracy": np.mean(cross_score),
        "Standard Deviation": np.std(cross_score)
    }

results_df = pd.DataFrame(results).T
print(results_df)
```

## Results and Conclusion for Classification Experiments
The table above summarizes the mean accuracy and standard deviation of the four classification models evaluated using 5-fold cross-validation on the CKD dataset.
- The Neural Network model achieved the highest mean accuracy of 0.9351 with a standard deviation of 0.0405. This implies that the model performed vbetter than other models and also exhibited high consistency across different folds of cross-validation. Here's some reasons why:
- Neural networks have the ability to learn complex patterns and interactions in the data through multiple layers of interconnected nodes. This allows identification of more intricate patterns.  
- Unlike Logistic Regression, which assumes a linear relationship, neural networks can learn feature interactions without explicit manual engineering.



```python
# Experiments with Neural Network.
nn_configs = {
    "Config 1 (1 layer, 32 neurons, relu, adam)": MLPClassifier(hidden_layer_sizes=(32,), activation='relu', solver='adam', max_iter=1000, random_state=42),
    "Config 2 (2 layers, 64-32 neurons, tanh, sgd)": MLPClassifier(hidden_layer_sizes=(64, 32), activation='tanh', solver='sgd', learning_rate_init=0.01, max_iter=1000, random_state=42),
    "Config 3 (3 layers, 128-64-32 neurons, relu, adam)": MLPClassifier(hidden_layer_sizes=(128, 64, 32), activation='relu', solver='adam', max_iter=1000, random_state=42),
}

nn_results = {}
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for config_name, model in nn_configs.items():
    cross_score = cross_val_score(model, X, y, cv=kf)
    nn_results[config_name] = {
        "Mean Accuracy": np.mean(cross_score),
        "Standard Deviation": np.std(cross_score)
    }

nn_results_df = pd.DataFrame(nn_results).T
print(nn_results_df)
```

## Results and Conclusion for Neural Network Experiments
- Config 3 outperformed the other configurations with a mean accuracy of 0.9675 and a relatively low standard deviation of 0.0353. This indicates high performance and consistency across different folds of cross-validation.
- The substantial improvement from Config 2 to Config 3 (from 0.9415 to 0.9675) indicates that increasing the number of layers and neurons significantly enhances the model’s ability to learn complex relationships. The three-layer configuration with 128, 64, and 32 neurons allowed for deeper representation learning with a further increase in mean accuracy. 
- The choice of the Adam optimizer in Config 1 and Config 3 helped those models converge faster and achieve better performance than traditional methods like stochastic gradient descent (SGD) used in Config 2.


The experimentation illustrates that:
Increasing Model Capacity: Adding more layers and neurons allows the model to capture richer representations of the data, significantly improving classification performance.
Tuning Activation Functions and Optimizers: The selection of appropriate activation functions and optimizers can substantially affect training efficiency and performance. In this case, using ReLU and Adam optimized configurations yielded the best results.
