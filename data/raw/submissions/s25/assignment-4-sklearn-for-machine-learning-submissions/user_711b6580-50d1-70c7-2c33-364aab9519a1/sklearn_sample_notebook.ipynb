```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
science_data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(science_data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(science_data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = science_data[['Temperature °C', 'Mols KCL']]
y = science_data['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample_datapoint = pd.DataFrame([[500, 700]], columns=['Temperature °C', 'Mols KCL'])
predicted_size = model.predict(sample_datapoint)
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
model_score = model.score(X_test, y_test)
print(f"R² Score of the model: {model_score:.4f}")
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficients (Temperature, Mols KCL): {coefficients}")
print(f"Intercept: {intercept}")
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
scores = cross_val_score(model, X, y, cv=5)
# Report on their finding and their significance
print("Cross-Validation Scores:", scores)
print(f"Average R² Score from Cross-Validation: {scores.mean():.4f}")
print(f"Standard Deviation of R² Scores: {scores.std():.4f}")
```

Write findings here:

Answer: The individual R² scores for the 5 cross-validation scores show how good the model is at predicting the target variable, slime size, on the 5 subsets.
The average R² score of 85.68% shows that the model is pretty good at predicting across the 5 subsets. 
The standard deviation of R² scores, 0.0135 shows that the model is nearly perfectly consistent across the subsets. There is not much deviation between the models predictions. 

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
model_poly = make_pipeline(poly, LinearRegression())
# Perform k-fold cross validation (as above)
scores_poly = cross_val_score(model_poly, X, y, cv=5)
model_poly.fit(X, y)

model_poly_trained = model_poly.named_steps['linearregression']
coefficients_poly = model_poly_trained.coef_
intercept_poly = model_poly_trained.intercept_
# Report on the metrics and output the resultant equation as you did in Part 3.
print("Cross-Validation Scores for Polynomial Regression:", scores_poly)
print(f"Average R² Score from Cross-Validation: {scores_poly.mean():.4f}")
print(f"Standard Deviation of R² Scores: {scores_poly.std():.4f}")
print(f"Intercept: {intercept_poly}")
print(f"Coefficients: {coefficients_poly}")
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_data = pd.read_csv('ckd_feature_subset.csv')
print(ckd_data.head())

X = ckd_data.drop(columns=['Target_ckd'])  # All columns except the target
y = ckd_data['Target_ckd']

def evaluate_model(model, X, y):
    scores = cross_val_score(model, X, y, cv=5)
    return scores.mean(), scores.std()
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Support Vector Machine": SVC(random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(random_state=42, max_iter=1000)
}
results = {}

for model_name, model in models.items():
    mean_score, std_score = evaluate_model(model, X, y)
    results[model_name] = {'Mean Score': mean_score, 'Standard Deviation': std_score}

results_df = pd.DataFrame(results).T
print(results_df)
```

## Results and Conclusion for Classification Experiments

| Model                | Mean Score (Accuracy) | Standard Deviation |
|----------------------|-----------------------|---------------------|
| Logistic Regression  | 0.8561                | 0.0542              |
| Support Vector Machine (SVM) | 0.9282         | 0.0599              |
| k-Nearest Neighbors (k-NN)   | 0.9282         | 0.0319              |
| Neural Network       | 0.9413                | 0.0240              |

This shows that Neural Network is the best model because it has the highest mean score, 94.13% and lowest standard deviation, 0.0240. Logistic Regression had the lowest accuracy score. 

Experiments with Neural Network

```python
# Experiments with Neural Network.
def evaluate_nn(hidden_layer_sizes, activation, learning_rate_init, max_iter=1000):
    model = MLPClassifier(
        hidden_layer_sizes=hidden_layer_sizes,
        activation=activation,
        learning_rate_init=learning_rate_init,
        max_iter=max_iter,
        random_state=42
    )
    
    mean_score, std_score = evaluate_model(model, X, y)
    return mean_score, std_score


nn_configurations = [
    {"hidden_layer_sizes": (50,), "activation": "relu", "learning_rate_init": 0.001},
    {"hidden_layer_sizes": (100, 50), "activation": "tanh", "learning_rate_init": 0.001},
    {"hidden_layer_sizes": (100, 100), "activation": "relu", "learning_rate_init": 0.01},
    {"hidden_layer_sizes": (150,), "activation": "tanh", "learning_rate_init": 0.01},
    {"hidden_layer_sizes": (50, 50, 50), "activation": "relu", "learning_rate_init": 0.001},
]


nn_results = {}

for i, config in enumerate(nn_configurations, 1):
    mean_score, std_score = evaluate_nn(
        config["hidden_layer_sizes"], 
        config["activation"], 
        config["learning_rate_init"]
    )
    nn_results[f"NN Config {i}"] = {
        "Hidden Layers": config["hidden_layer_sizes"],
        "Activation": config["activation"],
        "Learning Rate": config["learning_rate_init"],
        "Mean Score": mean_score,
        "Standard Deviation": std_score
    }


nn_results_df = pd.DataFrame(nn_results).T

# Displaying the results clearly
from IPython.display import display  
print("Summary of Neural Network Configurations:\n")
display(nn_results_df)
```

## Results and Conclusion for Neural Network Experiments

The best model was NN Config 3. It has the highest accuracy score, 98.04%, and lowest standard deviation, 0.015. The most important parameter that made the biggest difference in configuration was the number of hidden layers and their size. Here, it had 2 hidden layers of size (100, 100).
