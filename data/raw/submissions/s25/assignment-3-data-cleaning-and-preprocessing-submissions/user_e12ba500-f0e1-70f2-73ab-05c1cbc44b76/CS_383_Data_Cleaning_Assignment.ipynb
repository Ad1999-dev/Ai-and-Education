````python
# Calculate the percentage of rows that contain atleast one missing value
the_percentage_of_missing_value_rows = combined_data_set_by_merging_QL.isnull().any(axis=1).mean()*(100)
# Print %
print(the_percentage_of_missing_value_rows,"%")
# Drop these rows from the dataset
combined_data_set_by_merging_QL.dropna(how='any',inplace=True)
combined_data_set_by_merging_QL.reset_index(drop = True, inplace = True)

# Print the Dataset
display('name of columns',combined_data_set_by_merging_QL.columns)
display(combined_data_set_by_merging_QL)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
encode_dum_columns_QL_merge = pd.get_dummies(combined_data_set_by_merging_QL,\
                            columns=categorical_columns,\
                            prefix=categorical_columns,dtype=int)
# Print the dataset
display('name of columns',encode_dum_columns_QL_merge.columns)
display(encode_dum_columns_QL_merge)
````

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
list_of_means = encode_dum_columns_QL_merge[numerical_columns].mean()
list_of_stds = encode_dum_columns_QL_merge[numerical_columns].std()
list_of_upper_bounds = list_of_means + 3*list_of_stds
list_of_lower_bounds = list_of_means - 3*list_of_stds
display(list_of_lower_bounds)
for value in numerical_columns:
    encode_dum_columns_QL_merge = encode_dum_columns_QL_merge[\
    (encode_dum_columns_QL_merge[value] >= list_of_lower_bounds[value])\
     & (encode_dum_columns_QL_merge[value] <= list_of_upper_bounds[value])].dropna()
# Print the dataset
encode_dum_columns_QL_merge.reset_index(drop = True, inplace = True)
display(encode_dum_columns_QL_merge)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
for value in numerical_columns:
    encode_dum_columns_QL_merge[value] = (encode_dum_columns_QL_merge[value] - encode_dum_columns_QL_merge[value].min())/\
                                        (encode_dum_columns_QL_merge[value].max() - encode_dum_columns_QL_merge[value].min())
# Print the dataset
display(encode_dum_columns_QL_merge)
````

````python
# Code to rename all the columns in the dataset

def change_to_full_name(column):
    for abbreviation in full_name_map:
        if re.match(rf'^{abbreviation}(\
        _normal|_abnormal|_yes|_no|_present|_notpresent|_good|_poor)?$', column):
            root_nae = full_name_map[abbreviation]
            if column.endswith('_yes'):
                return f"{root_nae} Yes"
            elif column.endswith('_no'):
                return f"{root_nae} No"
            elif column.endswith('_normal'):
                return f"{root_nae} Normal"
            elif column.endswith('_abnormal'):
                return f"{root_nae} Abnormal"
            elif column.endswith('_present'):
                return f"{root_nae} present"
            elif column.endswith('_notpresent'):
                return f"{root_nae} not present"
            elif column.endswith('_good'):
                return f"{root_nae} good"
            elif column.endswith('_poor'):
                return f"{root_nae} poor"
            else:
                return root_nae
    return column 

full_name_map = {
    'age': 'age',
    'bp': 'blood pressure',
    'sg': 'specific gravity',
    'al': 'albumin',
    'su': 'sugar',
    'rbc': 'red blood cells',
    'pc': 'pus cell',
    'pcc': 'pus cell clumps',
    'ba': 'bacteria',
    'bgr': 'blood glucose random',
    'bu': 'blood urea',
    'sc': 'serum creatinine',
    'sod': 'sodium',
    'pot': 'potassium',
    'hemo': 'hemoglobin',
    'pcv': 'packed cell volume',
    'wbcc': 'white blood cell count',
    'rc': 'red blood cell count',
    'htn': 'hypertension',
    'dm': 'diabetes mellitus',
    'cad': 'coronary artery disease',
    'appet': 'appetite',
    'pe': 'pedal edema',
    'ane': 'anemia',
    'class': 'class',
    'rbcc':'red blood cell count',
    'rbc': 'red blood cells',
    'ba': 'bacteria',
    'htn':'hypertension',
    
}

encode_dum_columns_QL_merge.rename(columns=\
{column: change_to_full_name(column) for column in encode_dum_columns_QL_merge.columns}
, inplace=True)
display(encode_dum_columns_QL_merge.columns)
display(encode_dum_columns_QL_merge)
````
