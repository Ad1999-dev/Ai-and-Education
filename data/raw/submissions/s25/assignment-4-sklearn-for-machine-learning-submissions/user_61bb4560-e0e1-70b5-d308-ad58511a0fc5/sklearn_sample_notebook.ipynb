```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

data = pd.read_csv('science_data_large.csv')
print(data.head(15))
print(data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X = data[['Temperature °C', 'Mols KCL']]
y = data['Size nm^3']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
score = model.score(X_test, y_test)
coefficients = model.coef_
intercept = model.intercept_
equation = f"Size nm^3 = {intercept:.5f} + {coefficients[0]:.5f} * Temperature + {coefficients[1]:.5f} * KCl"
print(f'Model Score (R^2): {score}')
print(f'Regression Equation: {equation}')

sample = pd.DataFrame([[500, 500]], columns=["Temperature °C", "Mols KCL"])
sample_prediction = model.predict(sample)
print(f"Prediction for sample input [500°C, 500 mols]: {sample_prediction[0]:.2f} nm^3")

```

Report on score and explain meaning: The linear regression model explains approximately 85.5% of the variation in slime size based on the temperature and amount of KCl indicating a strong model fit.

## 4. Use Cross Validation

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_data = pd.read_csv('ckd_feature_subset.csv')
print(ckd_data.head())
print(ckd_data.info())


```

```python
X = ckd_data.drop('Target_ckd', axis=1)
y = ckd_data['Target_ckd']  # Assuming Target_ckd is the label
X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
logistic_model = LogisticRegression(max_iter=1000)
svc_model = SVC(random_state=42)
knn_model = KNeighborsClassifier()

# List of models for easier looping
models = [("Logistic Regression", logistic_model), 
          ("SVC", svc_model), 
          ("KNN", knn_model)]

# Initialize a dictionary to hold scores
model_results = {}

for name, model in models:
    # Fit the model
    model.fit(X_train, y_train)
    
    # Evaluate the model
    score = model.score(X_test, y_test)  # Could also use cross_val_score for k-fold cross validation
    model_results[name] = score
    print(f"{name} Accuracy: {score:.2f}")

for name, model in models:
    cv_scores = cross_val_score(model, X, y, cv=5)
    print(f"{name} Cross-Validation Scores: {cv_scores}")
    print(f"{name} Mean CV Score: {cv_scores.mean():.2f}")


y_pred = logistic_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python


# Example configuration settings for neural network
models_nn = {
    "NN 1 Layer 5 Neurons": MLPClassifier(hidden_layer_sizes=(5,), random_state=42, max_iter=500),
    "NN 2 Layers 5 Neurons Each": MLPClassifier(hidden_layer_sizes=(5, 5), random_state=42, max_iter=500),
    "NN 1 Layer 10 Neurons": MLPClassifier(hidden_layer_sizes=(10,), random_state=42, max_iter=500)
}

for name, nn_model in models_nn.items():
    nn_model.fit(X_train, y_train)
    nn_cv_scores = cross_val_score(nn_model, X, y, cv=5)
    print(f"{name} Mean CV Score: {nn_cv_scores.mean():.2f}")
```
