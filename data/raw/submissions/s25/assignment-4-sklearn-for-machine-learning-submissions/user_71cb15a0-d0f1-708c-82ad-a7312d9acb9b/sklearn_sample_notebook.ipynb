```python
# Imports section
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
science = pd.read_csv("science_data_large.csv")
ckd = pd.read_csv("ckd_feature_subset.csv")


# Output the first 15 rows of the data
print(science.head(15))
print(ckd.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(science.info())
print(ckd.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = science.drop(columns=['Size nm^3'])
y = science['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# For grading consistency use random_state=42 
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = np.array([X_test.iloc[0]]).reshape(1, -1)
prediction = model.predict(sample_data)
print("Sample Prediction:", prediction)

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print("Linear Regression Score:", score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

Write the linear equation of a slime: (example equation: $E = mc^2$)


```python
print(f"Equation of Slime Growth: f(x) = {model.intercept_:.5f} + " + " + ".join([f"{coeff:.5f}*x{i+1}" for i, coeff in enumerate(model.coef_)]))
```

Report on score and explain meaning:

The score represents the coefficient of determination (or R^2 score) which measures how well the model explains the variance in the data.
A high R^2 value and the closer it is to 1 indicates that the model explains most of the variability in the response variable. Since our score is 0.855, it is rather close to 1 showing that the model explains most of the variability in the slime size data.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

cv_scores = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=cv_scores, scoring='r2')

# Report on their finding and their significance
print("cross validation scores:", scores)
print("mean CV score:", np.mean(scores))
print("standard deviation", np.std(scores))
```

Write findings here:

Cross validation helps test the stability of the model across different subsets of the data. The average of the cross validation score is 0.8597, which is pretty close to our R^2 score. This means that the model is generalizing well to new data and is consistent across different folds. The standard deviation is rather small which also shows our model is consistent.


## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
polynomial_model = make_pipeline(PolynomialFeatures(2), LinearRegression())
cv_scores_polynomial = cross_val_score(polynomial_model, X, y, cv=5)
print("polynomial regression mean CV score:", np.mean(cv_scores_polynomial))

# Report on the metrics and output the resultant equation as you did in Part 3.
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.


X = ckd.drop(columns=['Target_ckd'])  
y = ckd['Target_ckd']

X_train_ckd, X_test_ckd, y_train_ckd, y_test_ckd = train_test_split(X, y, test_size=0.1, random_state=42)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "SVM": SVC(kernel='rbf', random_state=42),
    "k-NN": KNeighborsClassifier(n_neighbors=5),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
}

results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    results[name] = {
        "mean accuracy": np.mean(scores),
        "standard deviation": np.std(scores)
    }


results_df = pd.DataFrame(results).T
print(results_df)
```

## Results and Conclusion for Classification Experiments

```python
# Experiments with Neural Network.
nn_configs = {
    "Small NN": MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42),
    "Medium NN": MLPClassifier(hidden_layer_sizes=(20, 20), max_iter=1000, random_state=42),
    "Large NN": MLPClassifier(hidden_layer_sizes=(50, 50, 50), max_iter=1000, random_state=42)
}


res = {}

for name, model in nn_configs.items():
    scores = cross_val_score(model, X_scaled, y, cv=5)
    res[name] = {
        "mean accuracy": np.mean(scores),
        "standard deviation": np.std(scores)
    }


res_df = pd.DataFrame(res).T
print(res_df)
```

## Results and Conclusion for Neural Network Experiments

According to the experiment results, Neural Networks is the most accurate model with a mean accuracy of 0.941 which was the highest of all 4. It performed the best with a medium sized neural network with a mean accuracy of 0.974. The worst performing model was the Logistic Regression with a mean accuracy of 0.8561.
