## 1. Loading the dataset

```python
import pandas as pd

# Using pandas load the dataset
data = pd.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
print (data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
data.info()
```

## 2. Splitting the dataset

```python
from sklearn.model_selection import train_test_split
# Take the pandas dataset and split it into our features (X) and label (y)
x = data[['Temperature °C', 'Mols KCL']]
y = data['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
# For grading consistency use random_state=42 
```

## 3. Perform a Linear Regression

```python
from sklearn.linear_model import LinearRegression
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(x_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
datapoint = [[132, 777]]
predicted = model.predict(datapoint)

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score = model.score(x_test, y_test)
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print(coefficients)
print(intercept)
print(predicted)
print(score)
```

## 4. Use Cross Validation

```python
from sklearn.model_selection import cross_val_score, KFold
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, x, y, cv=kf)

print(cv_scores)
# Report on their finding and their significance
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
# Load the dataset. Then train and evaluate the classification models.

kidney_data = pd.read_csv("ckd_feature_subset.csv")

x = kidney_data[['age', 'bp', 'wbcc', 'appet_poor', 'appet_good', 'rbcc']]
y = kidney_data['Target_ckd']

A_train, A_test, B_train, B_test = train_test_split(x, y, test_size=0.1, random_state=42)

log_reg = LogisticRegression()
log_reg.fit(A_train, B_train)

svm = SVC()
svm.fit(A_train, B_train)

knn = KNeighborsClassifier()
knn.fit(A_train, B_train)

nn = MLPClassifier()
nn.fit(A_train, B_train)

kfol = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores1 = cross_val_score(log_reg, x, y, cv=kfol)
cv_scores2 = cross_val_score(svm, x, y, cv=kfol)
cv_scores3 = cross_val_score(knn, x, y, cv=kfol)
cv_scores4 = cross_val_score(nn, x, y, cv=kfol)

print(cv_scores1)
print(cv_scores2)
print(cv_scores3)
print(cv_scores4)

print("Logistic Regression: Mean =", np.mean(cv_scores1), ", Std Dev =", np.std(cv_scores1))
print("SVM: Mean =", np.mean(cv_scores2), ", Std Dev =", np.std(cv_scores2))
print("KNN: Mean =", np.mean(cv_scores3), ", Std Dev =", np.std(cv_scores3))
print("Neural Network: Mean =", np.mean(cv_scores4), ", Std Dev =", np.std(cv_scores4))

```

## Results and Conclusion for Classification Experiments

| Model | Mean Accuracy | Standard Deviation |
|---------------------|---------------|---------------------|
| Logistic Regression | 0.8566 | 0.0663 |
| Support Vector Machine (SVM) | 0.9282 | 0.0476 |
| k-Nearest Neighbors (KNN) | 0.9280 | 0.0524 |
| Neural Network | 0.8566 | 0.0663 |

In this evaluation, the Support Vector Machine (SVM) model demonstrated the highest performance, achieving a mean accuracy of 0.928, which exceeds the accuracy of the other models assessed. This superior performance can be attributed to SVM's ability to effectively handle both linear and non-linear relationships, particularly in high-dimensional datasets. By utilizing kernel functions to maximize the margin between classes, SVM is capable of generalizing well, as evidenced by its low standard deviation of 0.0476, indicating consistent performance across different data splits.
In comparison, the k-Nearest Neighbors (KNN) model also performed admirably, with a mean accuracy of 0.9279 and a slightly higher standard deviation of 0.0524. This suggests that while KNN is quite effective, it may exhibit slightly greater variability in its performance across different subsets of the data.
Other models, such as Logistic Regression and Neural Networks, achieved lower accuracies, both with a mean of 0.8566 and higher standard deviations (0.0663), indicating less stability. The limitations in these models may stem from their inability to capture complex patterns inherent in the data or manage its high dimensionality effectively.

```python
# Experiments with Neural Network.
from sklearn.model_selection import KFold, cross_val_score
import numpy as np

nn1 = MLPClassifier(hidden_layer_sizes=(50,), activation='relu', max_iter=1000)
nn2 = MLPClassifier(hidden_layer_sizes=(50,), activation='logistic', max_iter=10000)
nn3 = MLPClassifier(hidden_layer_sizes=(500,), activation='logistic', max_iter=1000)

kfol = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_nn1 = cross_val_score(nn1, x, y, cv=kfol)
cv_scores_nn2 = cross_val_score(nn2, x, y, cv=kfol)
cv_scores_nn3 = cross_val_score(nn3, x, y, cv=kfol)

print(cv_scores_nn1)
print(cv_scores_nn2)
print(cv_scores_nn3)

print("NN1: Mean =", np.mean(cv_scores_nn1), ", Std Dev =", np.std(cv_scores_nn1))
print("NN2: Mean =", np.mean(cv_scores_nn2), ", Std Dev =", np.std(cv_scores_nn2))
print("NN3: Mean =", np.mean(cv_scores_nn3), ", Std Dev =", np.std(cv_scores_nn3))

```
