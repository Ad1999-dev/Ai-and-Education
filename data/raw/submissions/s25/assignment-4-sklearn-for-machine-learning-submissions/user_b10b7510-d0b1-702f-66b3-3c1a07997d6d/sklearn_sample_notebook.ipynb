```python
# Imports section
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

df = pd.read_csv('science_data_large.csv')

df.head(15)
```

```python
df.info()
```

```python
df.describe()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df[['Temperature °C', 'Mols KCL']]
y = df['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame({
    'Temperature °C': [40],
    'Mols KCL': [0.5]
})
y_pred = lin_model.predict(sample)

print(f"Predicted size {y_pred[0]:.2f} nm^3")
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
train_score = lin_model.score(X_train, y_train)
test_score = lin_model.score(X_test, y_test)

print(f"Training score: {train_score:.4f}")
print(f"Test score: {test_score:.4f}")

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

print(f"intercept: {lin_model.intercept_}")
print(f"coefficients: [Temperature, Mols KCL]: {lin_model.coef_}")
```

Report on score and explain meaning:

For this model, it has training score of 0.8610 and test score of 0.8552, which is the R² value of the model. This implies that about 86% of the variation in the slime size is explained by temperature and potassium chloride concentration. The similarity in training and test scores shows that the model generalizes well to new data without overfitting. With the coefficients is positive shows that temperature and concentration of KCl both positively have an impact and influence on the slime growth and the size.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

k_fold = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(lin_model, X, y, cv=k_fold)

# Report on their finding and their significance
print(f"Cross-validation: {cv_scores}")
print(f"Mean: {cv_scores.mean():.4f}")
print(f"Standard deviation: {cv_scores.std():.4f}")
```

Write findings here:
The cross validation shows that there is consistent level of performance from different splits of data with mean approximately 0.86 and standard deviation of 0.02. This shows that our model is not only stable but also reliable.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
X_poly_train, X_poly_test, y_poly_train, y_poly_test = train_test_split(X_poly, y, test_size=0.1, random_state=42)


poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_poly_train)


poly_train_score = poly_model.score(X_poly_train, y_poly_train)
poly_test_score = poly_model.score(X_poly_test, y_poly_test)
print(f"Training score: {poly_train_score:.4f}")
print(f"Test score: {poly_test_score:.4f}")

poly_cv_scores = cross_val_score(poly_model, X_poly, y, cv=k_fold)
print(f"Polynomial CV: {poly_cv_scores}")
print(f"Mean CV: {poly_cv_scores.mean():.4f}")
print(f"Standard deviation: {poly_cv_scores.std():.4f}")

c_poly = poly_model.coef_
i_poly = poly_model.intercept_
        
c_poly.sort()
c_poly = c_poly[::-1]

# Report on the metrics and output the resultant equation as you did in Part 3.
print(f"intercept: {i_poly:.5f}")
formatted_coefficients = ", ".join([f"{coef:.5f}" for coef in c_poly])
print(f"coefficients: [Temperature, Mols KCL]: [{formatted_coefficients}]")
```

## Results and Conclusion for Neural Network Experiments

```python
n_layer = {
    '1 layer, 10 neurons': MLPClassifier(hidden_layer_sizes=(10,), max_iter=2000, random_state=42),
    '2 layers, 10 neurons each': MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=42), 
    '3 layers, 10 neurons each': MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000, random_state=42),
    '1 layer, 50 neurons': MLPClassifier(hidden_layer_sizes=(50,), max_iter=2000, random_state=42)
}  
output_n = {}

for config_name, model in n_layer.items():
    cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='accuracy')
    output_n[config_name] = {
        'Mean Accuracy': cv_scores.mean(),
        'Standard Deviation': cv_scores.std()
    }

print(pd.DataFrame(output_n).T)

```

I experimented with various hidden layer structures and number of neurons with the hope of finding the best model structure. The two-layered network with 10 neurons in each of the two layers performed optimally at the maximum accuracy of 97.40%. A single-layered network with 50 neurons followed closely at 97.38%. The inclusion of a third layer reduced performance by a small amount to 96.09%. This result emphasizes that larger models do not always mean better predictive performance, and there is an optimal level of network complexity for this task of chronic kidney disease prediction.
