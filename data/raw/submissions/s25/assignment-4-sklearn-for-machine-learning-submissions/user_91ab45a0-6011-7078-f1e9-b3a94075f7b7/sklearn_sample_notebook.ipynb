```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
s = pd.read_csv("science_data_large.csv")
df = pd.DataFrame(s)
# Output the first 15 rows of the data
s.head(15)
```

```python
# Display a summary of the table information (number of datapoints, etc.)
s.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.drop("Size nm^3", axis = 1)
y = df["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
# For grading consistency use random_state=42 
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame({'Temperature °C': [450], 'Mols KCL': [500]})
predicted_output = model.predict(sample)
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
model.score(X, y)
```

```python
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_  
intercept = model.intercept_ 
print("coefficients:", coefficients)
print("intercept:", intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
cv_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print(cv_scores)
# Report on their finding and their significance
```

Write findings here: \
Since I chose the same scoring as the previous model, we can fairly compare the two models. Based on the $R^2$ values, 2 out of 5 of the scores are more than the linear regression model, meaning that the model performed better than the linear regression model 40% of the time and slightly worse 60% of the time. This means that the two models perform about the same.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd = pd.read_csv("ckd_feature_subset.csv")
df = pd.DataFrame(ckd)
X = df.drop('Target_ckd', axis=1)
y = df['Target_ckd']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 1. Logistic Regression

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_scores = cross_val_score(lr_model, X, y, cv=5, scoring='accuracy')
print(lr_scores)
```

## 2. Support Vector Machines

```python
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
svm_scores = cross_val_score(svm_model, X, y, cv=5, scoring='accuracy')
print(svm_scores)
```

## 3. k-Nearest Neighbors

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_scores = cross_val_score(knn_model, X, y, cv=5, scoring='accuracy')
print(knn_scores)
```

## 4. Neural Networks

```python
# Experiments with Neural Network.
nn_model_1 = MLPClassifier(max_iter=10000)
nn_model_1.fit(X_train, y_train)
nn_scores_1 = cross_val_score(nn_model_1, X, y, cv=5, scoring='accuracy')
print(nn_scores_1)
```

```python
nn_model_2 = MLPClassifier(hidden_layer_sizes=(8, 4), max_iter=10000, random_state=42)
nn_model_2.fit(X_train, y_train)
nn_scores_2 = cross_val_score(nn_model_2, X, y, cv=5, scoring='accuracy')
print(nn_scores_2)
```

```python
nn_model_3 = MLPClassifier(activation='identity', solver='lbfgs', learning_rate='adaptive', max_iter=10000, random_state=42)
nn_model_3.fit(X_train, y_train)
nn_scores_3 = cross_val_score(nn_model_3, X, y, cv=5, scoring='accuracy')
print(nn_scores_3)
```

## Results and Conclusion for Classification Experiments

```python
data = {
    'Array': ['Logistic Regression', 'Support Vector Machines', 'k-Nearest Neighbors'],
    'Mean': [np.mean(lr_scores), np.mean(svm_scores), np.mean(knn_scores)],
    'Standard Deviation': [np.std(lr_scores), np.std(svm_scores), np.std(knn_scores)]
}
results_table = pd.DataFrame(data)
print(results_table)
```

Conclusion:\
Logistic Regression performed the worst, while SVM and k-Nearest Neighbors performed surprisingly the exact same when you just look at the mean. However, since KNN has a smaller standard deviation, KNN overall performed the best. I believe this is the case because KNN models don't necessarily need a linear relationship between parameters as logistic regression does to perform better. Additionally, KNN models don't have as much of an impact with outliers as do logistic regression models do, which may explain why KNN has a lower standard deviation than the SVM model.
