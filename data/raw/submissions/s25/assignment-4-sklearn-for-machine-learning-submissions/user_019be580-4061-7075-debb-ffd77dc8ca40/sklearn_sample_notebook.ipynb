```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
slime = pd.read_csv("science_data_large.csv")
print(slime.head(15))
print(slime.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = slime.drop(columns=["Size nm^3"])
y = slime["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9,test_size=0.1, random_state=42)
# For grading consistency use random_state=42 
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
m = LinearRegression().fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
datap = X_train.head(1)
p_out = m.predict(datap)
print(p_out)
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
m_score = m.score(X_test, y_test)
print(m_score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = m.coef_
intercept = m.intercept_
print(coefficients)
print(intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
model_cross = LinearRegression()
cv = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model_cross, X, y, cv=cv)
# Report on their finding and their significance
print(scores)
```

Write findings here: [0.86152 0.82742 0.87195 0.88166 0.85609], after randomly splitting the orginal data 5 times into differnt sets, the number returned is R^2, or our score from the earlier question. Meaning this is how close the given set is to the mean and then meaning how accurate it is predicating a value. So the fact that our data is consinetly > 8 means the model is fairly accurate. 

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
#X_test_poly = poly.transform(X_test)
score = []

for train_i, test_i in cv.split(X_train):
    X_train_fold, X_test_fold = X_train.iloc[train_i], X_train.iloc[test_i]
    y_train_fold, y_test_fold = y_train.iloc[train_i], y_train.iloc[test_i]
    poly = PolynomialFeatures(degree=2)
    X_train_poly = poly.fit_transform(X_train_fold)
    X_test_poly = poly.transform(X_test_fold)
    model = LinearRegression().fit(X_train_poly, y_train_fold)
    score.append(model.score(X_test_poly,y_test_fold))
    
    
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
model = LinearRegression().fit(X_poly,y_train)

# Report on the metrics and output the resultant equation as you did in Part 3.
coefficients = model.coef_
intercept = model.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)
print("Cross-validated R² scores:", score)
print("Mean R² score:", np.mean(score))
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
clean_ckd = pd.read_csv("ckd_feature_subset.csv")
X = clean_ckd.drop(columns=["Target_ckd"])
y = clean_ckd["Target_ckd"]
kf = KFold(n_splits=5, shuffle=True, random_state=42)
# X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9,test_size=0.1, random_state=42)



# model = LogisticRegression().fit(X_train,y_train)
# lin_reg_score = cross_val_score(model, X_train, y_train, cv=kf)
model = LogisticRegression()
lin_reg_score = cross_val_score(model, X,y, cv=kf)
lin_reg_mean = lin_reg_score.mean()
lin_reg_sd = lin_reg_score.std()
print("LogisticRegression")
print(lin_reg_mean)
print(lin_reg_sd)


# model = SVC().fit(X_train,y_train)
# svc_score = cross_val_score(model, X_train, y_train, cv=kf)
model = SVC()
svc_score = cross_val_score(model, X, y, cv=kf)
svc_mean = svc_score.mean()
svc_sd = svc_score.std()
print("SVC")
print(svc_mean)
print(svc_sd)


# model = KNeighborsClassifier().fit(X_train,y_train)
# kneigh_score = cross_val_score(model, X_train, y_train, cv=kf)
model = KNeighborsClassifier()
kneigh_score = cross_val_score(model, X, y, cv=kf)
kneigh_score_mean = kneigh_score.mean()
kneigh_score_sd = kneigh_score.std()
print("KNeighborsClassifier")
print(kneigh_score_mean)
print(kneigh_score_sd)


# model = MLPClassifier(max_iter=1000).fit(X_train,y_train)
# mlp_score = cross_val_score(model, X_train, y_train, cv=kf)
model = MLPClassifier(max_iter=1000)
mlp_score = cross_val_score(model, X, y, cv=kf)
mlp_mean = mlp_score.mean()
mlp_sd = mlp_score.std()
print("MLPClassifier")
print(mlp_mean)
print(mlp_sd)
```

## Results and Conclusion for Classification Experiments

| Model Name | Mean | Std |

| SVC | 0.92827 | 0.047601 |

Interperting Meaning:
The best performing model was the MLP Classfier with the highest standard R^2 mean and the lowest standard devation among the scores. This means it as a high degree of accuracy and stability across any split of the data. Both SVC and KNeighbors both had a nearly identical mean and vaired with their STD. So both perforemd really well across multiple diffenrt folds. Logistic Regeression performed with the worst the lowest mean and high STD, so in comparsion to the other did was not as accurate and had a wider range of instaibtily. MLP may perform the best since it is a neural netowrk allowing it to capture more complex and non linear boundires within the dataset. 


```python
# Experiments with Neural Network.
# model = MLPClassifier(max_iter=1000,hidden_layer_sizes=(200,)).fit(X_train,y_train)
# mlp_score1 = cross_val_score(model, X_test, y_test, cv=kf)
model = MLPClassifier(max_iter=2000,hidden_layer_sizes=(200,))
mlp_score1 = cross_val_score(model, X, y, cv=kf)
mlp_mean = mlp_score1.mean()
mlp_sd = mlp_score1.std()
print("MLPClassifier1")
print(mlp_mean)
print(mlp_sd)

# model = MLPClassifier(max_iter=1000,activation='logistic').fit(X_train,y_train)
# mlp_score2 = cross_val_score(model, X_test, y_test, cv=kf)
model = MLPClassifier(max_iter=2000,activation='logistic')
mlp_score2 = cross_val_score(model, X, y, cv=kf)
mlp_mean = mlp_score2.mean()
mlp_sd = mlp_score2.std()
print("MLPClassifier2")
print(mlp_mean)
print(mlp_sd)


# model = MLPClassifier(max_iter=1000,solver='lbfgs').fit(X_train,y_train)
# mlp_score3 = cross_val_score(model, X_test, y_test, cv=kf)
model = MLPClassifier(max_iter=2000,solver='lbfgs')
mlp_score3 = cross_val_score(model, X, y, cv=kf)
mlp_mean = mlp_score3.mean()
mlp_sd = mlp_score3.std()
print("MLPClassifier3")
print(mlp_mean)
print(mlp_sd)
```
