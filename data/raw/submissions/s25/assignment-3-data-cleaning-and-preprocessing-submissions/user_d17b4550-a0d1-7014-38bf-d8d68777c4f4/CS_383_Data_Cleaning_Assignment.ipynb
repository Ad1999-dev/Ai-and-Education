## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value
rows_with_missing = combined_df.isna().any(axis='columns')
percentage_missing = (rows_with_missing.sum() / combined_df.shape[0]) * 100

# Print %
print(f"Percentage of rows with at least one missing value: {percentage_missing:}%")

# Drop these rows from the dataset
cleaned_df = combined_df.dropna()

# Print the Dataset
display(cleaned_df)
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
sorted_df = cleaned_df.sort_values(by='Target').reset_index(drop=True)
# Print the dataset
display(sorted_df)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
df_encoded = pd.get_dummies(sorted_df, columns=categorical_columns, dtype=int)

# Print the dataset
display(df_encoded)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: When using a single column for a binary variable, this would be equivalent in most downstream models. This is because model can still learn the same information and the only difference is the way of representation. However, for categorical variables with more than 2 values (like 5 types of condiments) or more, using a single column would not work because the model will misinterpret the order, and this will affect significantly to the performance and very likely gives wrong output. This is because the model misinterprets the information and it just like we use a different unexpected data to train the models. Also, some models may not able to process categorical type in a single column. 

In the sandwich rating prediction example, we can use one-hot encoding to tell the models the difference and independence of each condiment without accidentally establishing relationships (correlation) between them. Using a single column would make the model to think the condiments are ordinal while they are actually nominal, potentially missing the true connection between condiment type and sandwich rating, giving false results and bad predicting performance.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

mean = df_encoded[numerical_columns].mean()
std = df_encoded[numerical_columns].std()

lower_bound = mean - 3 * std
upper_bound = mean + 3 * std

mask = pd.DataFrame()
for col in numerical_columns:
    column_mask = (df_encoded[col] >= lower_bound[col]) & (df_encoded[col] <= upper_bound[col])
    mask[col] = column_mask

all_columns_mask = mask.all(axis='columns')
df_no_outliers = df_encoded[all_columns_mask]
print(f"Number of outliers removed: {df_encoded.shape[0] - df_no_outliers.shape[0]}")

# Print the dataset
display(df_no_outliers)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
min_values = df_no_outliers[numerical_columns].min()
max_values = df_no_outliers[numerical_columns].max()

df_normalized = df_no_outliers.copy()
for col in numerical_columns:
    df_normalized[col] = (df_no_outliers[col] - min_values[col]) / (max_values[col] - min_values[col])

# Print the dataset
display(df_normalized)
````

## Part 3.8: Remove Unnecessary columns

Are there any columns in this dataset which are not appropriate for modeling and predictions? Which column(s)? Justify their exclusion and remove them

Answer: I observe that the column `unique_id` is redundant and may introduce unexpected behavior of our models. Including identifier columns in learning models can lead to false correlation as the model may think there is a pattern and the ids are important parts of the data but actually they do not have any value in learning. So when the models predict, it may rely on the pattern of these arbitrary IDs rather than the medical features which are the real expected features we focus on. In addition, it will increase the complexity for the model unnecessarily as there are no purposes to present the IDs. In other words, this column is only useful for record identification, but adds no value for pattern recognition or prediction tasks. Therefore, it should be removed from the modeling dataset.

````python
#Remove that column
final_df = df_normalized.drop(columns='unique_id')
# Print the dataset
display(final_df)
````
