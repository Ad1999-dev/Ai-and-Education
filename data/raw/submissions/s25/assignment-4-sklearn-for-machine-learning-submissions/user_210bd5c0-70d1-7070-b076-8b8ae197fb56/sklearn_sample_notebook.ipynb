```python
# Imports section
import sklearn as skl
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, ShuffleSplit, KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import svm
import pandas as pd
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
SD = pd.read_csv("./science_data_large.csv")
# Output the first 15 rows of the data
print(SD.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(SD.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
x = SD[["Temperature °C", "Mols KCL"]]
y = SD["Size nm^3"]

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(x, y, train_size=0.9, random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
reg = LinearRegression().fit(X=x_train, y=y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame()
sample["Temperature °C"] = [550]
sample["Mols KCL"] = [550]
print(reg.predict(sample))
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print(reg.score(X=x_test, y=y_test))
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print(reg.coef_)
print(reg.intercept_)
```

## 5. Using Polynomial Regression

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
CKD = pd.read_csv("./ckd_feature_subset.csv")
x = CKD.drop("Target_ckd", inplace=False, axis=1)
y = CKD["Target_ckd"]
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(x,y, train_size=0.9, random_state=42)

results = pd.DataFrame(index=["Logistic Regression", "Support Vector Machines", "k-Nearest Neighbors"])
```

### Logistic Regression

```python
log_reg = LogisticRegression(random_state=42).fit(X=x_train, y=y_train)

log_kf_scores = cross_val_score(log_reg, X=x, y=y, cv=kf)

results.at["Logistic Regression", "scores"] = str(log_kf_scores)

results.at["Logistic Regression", "avg"] = log_kf_scores.mean()

results.at["Logistic Regression", "std"] = log_kf_scores.std()
```

### Support Vector Machines

```python
svm_model = svm.SVC(random_state=42).fit(X=x_train, y=y_train)

svm_kf_scores = cross_val_score(svm_model, X=x, y=y, cv=kf)

results.at["Support Vector Machines", "scores"] = str(svm_kf_scores)

results.at["Support Vector Machines", "avg"] = svm_kf_scores.mean()

results.at["Support Vector Machines", "std"] = svm_kf_scores.std()
```

### k-Nearest Neighbors

```python
model = Pipeline(
    steps=[("scaler", StandardScaler()), ("knn", KNeighborsClassifier())]
)

model.fit(X=x_train, y=y_train)

kn_kf_scores = cross_val_score(model, X=x, y=y, cv=kf)

results.at["k-Nearest Neighbors", "scores"] = str(kn_kf_scores)

results.at["k-Nearest Neighbors", "avg"] = kn_kf_scores.mean()

results.at["k-Nearest Neighbors", "std"] = kn_kf_scores.std()
```

## Results and Conclusion for Classification Experiments

```python
print(results)
```

The k-Nearest Neighbors model performed the best with an average score of 0.890323. I think this is the case because the other 2 models make boundary lines which could completely inaccurate for a certain section of the data space, while kNN takes the spatial situation of each point into account when classifying them rather than brushing with broad strokes with a boundary. 

### Neural Networks

```python
results = pd.DataFrame(columns=["scores", "avg", "std"])

def add_results(nn):
    nn.fit(X=x_train, y=y_train)
    row = []
    nn_scores = cross_val_score(nn, X=x, y=y, cv=kf)
    row.append(str(nn_scores))
    row.append(nn_scores.mean())
    row.append(nn_scores.std())
    results.loc[len(results)] = row
```

```python
# Experiments with Neural Network.
#Default

nn = MLPClassifier(random_state=42)

add_results(nn)
```

```python
#Increasing max_iterations
nn = MLPClassifier(max_iter=1000, random_state=42)

add_results(nn)
```

```python
#Increasing learning rate
nn = MLPClassifier(learning_rate_init=1, random_state=42)

add_results(nn)
```

```python
#Increasing learning rate and changing learning rate update schedule
nn = MLPClassifier(learning_rate="invscaling", learning_rate_init=1, random_state=42)

add_results(nn)
```

## Results and Conclusion for Neural Network Experiments

```python
print(results)
```

I increased max iterations because when I did the default (avg score 0.851613), I got a warning that said that the optimization of the parameters was not able to be completed with the default amount of iterations, so I added more iterations and got way better results (avg score 0.922581), most likely because the parameters were able to finish optimizing. I then tried increasing the learning rate which allowed the model to be trained way faster (the first and second times were 0.9 seconds and 2.5 seconds respectively, with the third being 0.1 seconds) but the average score was worse than both of my previous trials (0.825806). I tried changing the learning rate update schedule to inverse scaling rather than constant, but everything, the time and the average score was the same. What made the biggest difference in accuracy was the number of iterations, what made the most difference in speed was the learning rate.
