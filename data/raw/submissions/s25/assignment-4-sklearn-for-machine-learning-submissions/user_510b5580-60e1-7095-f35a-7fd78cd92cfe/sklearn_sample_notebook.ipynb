```python
# Imports section
import pandas as pd
import sklearn as sk
from numpy import ravel
from sklearn.model_selection import train_test_split, cross_val_score, ShuffleSplit
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

slime_data = pd.read_csv("science_data_large.csv")

slime_data.iloc[:15]

slime_data.describe()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

labels = slime_data[["Size nm^3"]]

features = slime_data.drop(columns=["Size nm^3"])

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42

features_train, features_test, labels_train, labels_test = train_test_split(
    features, labels, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model1 = LinearRegression().fit(features_train, labels_train)

# Create a sample datapoint and predict the output of that sample with the trained model
data_point = [[178, 372]]
print(model1.predict(data_point))

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print("\nScore:",model1.score(features_test, labels_test))

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
x0 = model1.coef_[0][0]
x1 = model1.coef_[0][1]
b = model1.intercept_
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
cross_val_score(model1, features, labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42))

# Report on their finding and their significance
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_data = pd.read_csv("ckd_feature_subset.csv")

ckd_labels = ravel(ckd_data[["Target_ckd"]])
ckd_features = ckd_data.drop(columns=["Target_ckd"])

ckd_features_train, ckd_features_test, ckd_labels_train, ckd_labels_test = train_test_split(
    ckd_features, ckd_labels, test_size=0.1, random_state=42)

LR_model = LogisticRegression(random_state=42).fit(ckd_features_train, ckd_labels_train)
SVC_model = svm.SVC(random_state=42).fit(ckd_features, ckd_labels)
KNN_model = KNeighborsClassifier().fit(ckd_features, ckd_labels)

print(cross_val_score(LR_model, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))
print(cross_val_score(SVC_model, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))
print(cross_val_score(KNN_model, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))

NN_model = MLPClassifier(random_state=42).fit(ckd_features, ckd_labels)
print(cross_val_score(NN_model, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))
```

The model with the best average $R^2$ score is the k-nearest neighbors model which also has the lowest deviation. By these metrics it is the best model but I am wary of k-nearest neighbor "finding trends" that do not exist.

```python
# Experiments with Neural Network.
NN_model1 = MLPClassifier(random_state=42, max_iter=1000).fit(ckd_features, ckd_labels)

NN_model2 = MLPClassifier(random_state=42, solver="sgd").fit(ckd_features, ckd_labels)

NN_model3 = MLPClassifier(random_state=42, learning_rate_init=0.005).fit(ckd_features, ckd_labels)

print(cross_val_score(NN_model1, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))
print(cross_val_score(NN_model2, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))
print(cross_val_score(NN_model3, ckd_features, ckd_labels, cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)))
```
