```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
alien_df = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(alien_df[:15])
# Display a summary of the table information (number of datapoints, etc.)
print(alien_df.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
alien_X = alien_df[['Mols KCL', 'Temperature °C']]
alien_y = alien_df['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
alien_X_train, alien_X_test, alien_y_train, alien_y_test = train_test_split(
    alien_X, alien_y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
alien_reg = LinearRegression().fit(alien_X_train, alien_y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
print(alien_reg.predict([[800, 100]]))
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print(alien_reg.score(alien_X_test, alien_y_test))
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print(alien_reg.coef_)
print(alien_reg.intercept_)
```

Write the linear equation of a slime: (example equation: $E = mc^2$)

$S = 1032.7M + 866.1T -409391.5$

Report on score and explain meaning:

Score uses the ratio of sum of squared residuals and the sum of squares (which measures variability in data) to give the model a score of 1 or below, where 1 is a perfect score. Our model got a score of 0.86, which is fairly close to a score of 1 and indicates that the model performs reasonably well.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)

alien_scores = cross_val_score(alien_reg, alien_X, alien_y, cv=kfoldcv)
# For grading consistency use n_splits=5 and random_state=42
# Report on their finding and their significance
print(alien_scores)
print(alien_scores.mean())
print(alien_scores.std())
```

Write findings here:

Linear regression consistently performs quite well on the data, acheiving an average score of 0.86 and standard deviation of 0.02. This is consistent with the results of the original model.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
alien_poly = poly.fit_transform(alien_X)

alien_poly_reg = LinearRegression().fit(alien_poly, alien_y)
# Perform k-fold cross validation (as above)
alien_poly_scores = cross_val_score(alien_poly_reg, alien_poly, alien_y, cv=kfoldcv)

# print(alien_poly_reg.predict([alien_poly[1]]))

# Report on the metrics and output the resultant equation as you did in Part 3.

print(alien_poly_reg.coef_)
print(alien_poly_reg.intercept_)

print(alien_poly_scores)
print(alien_poly_scores.mean())
print(alien_poly_scores.std())
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Using pandas load the dataset
ckd_df = pd.read_csv('ckd_feature_subset.csv')
# Output the first 15 rows of the data
print(ckd_df[:15])
# Display a summary of the table information (number of datapoints, etc.)
print(ckd_df.info())
```

```python
# Take the pandas dataset and split it into our features (X) and label (y)
ckd_X = ckd_df[['age', 'bp', 'wbcc', 'appet_poor', 'appet_good', 'rbcc']]
ckd_y = ckd_df['Target_ckd']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
ckd_X_train, ckd_X_test, ckd_y_train, ckd_y_test = train_test_split(
    ckd_X, ckd_y, test_size=0.1, random_state=42)
```

```python
# Use sklearn to train a model on the training set
ckd_reg = LogisticRegression().fit(ckd_X_train, ckd_y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print(ckd_reg.score(ckd_X_test, ckd_y_test))
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)

ckd_scores = cross_val_score(ckd_reg, ckd_X, ckd_y, cv=kfoldcv)
# For grading consistency use n_splits=5 and random_state=42
# Report on their finding and their significance
print(ckd_scores)
print(ckd_scores.mean())
print(ckd_scores.std())
```

```python
# 1. Train and test a Support Vector Machine (SVM) model
ckd_svm = SVC(random_state=42).fit(ckd_X_train, ckd_y_train)
print("SVM Test Score:", ckd_svm.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
svm_scores = cross_val_score(ckd_svm, ckd_X, ckd_y, cv=kfoldcv)
print("SVM Cross-Validation Scores:", svm_scores)
print("Mean Score:", svm_scores.mean())
print("Standard Deviation:", svm_scores.std())
```

```python
# 2. Train and test a K-Nearest Neighbors (KNN) model
ckd_knn = KNeighborsClassifier().fit(ckd_X_train, ckd_y_train)
print("KNN Test Score:", ckd_knn.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
knn_scores = cross_val_score(ckd_knn, ckd_X, ckd_y, cv=kfoldcv)
print("KNN Cross-Validation Scores:", knn_scores)
print("Mean Score:", knn_scores.mean())
print("Standard Deviation:", knn_scores.std())
```

```python
# 3. Train and test a Neural Network model
ckd_nn = MLPClassifier(random_state=42).fit(ckd_X_train, ckd_y_train)
print("Neural Network Test Score:", ckd_nn.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
nn_scores = cross_val_score(ckd_nn, ckd_X, ckd_y, cv=kfoldcv)
print("Neural Network Cross-Validation Scores:", nn_scores)
print("Mean Score:", nn_scores.mean())
print("Standard Deviation:", nn_scores.std())
```

## Results and Conclusion for Classification Experiments

| Model        | Mean Score | Std      |
|--------------|------------|----------|

| SVM          | 0.93       | 0.05     |

Logistic Regression:

The scores for the logistic regression model are a bit inconsistent. The mean score is a .86, which is good, but there is a standard deviation of 0.07

Support Vector Machines:

The scores using support vector machines were much more promising, with a mean score of 0.93 and a standard deviation of five. This means that the model performed more accurately and more consistently than regression,

k-Nearest Neighbors:

The KNN algorithm produced almost identical results to support vector machines.

Neural Network:

The neural network provided similar results to logistic regression. I did, however, receive a warning that the optimization has not converged after the maximum number of iterations was reached. This implies that the neural network may still be promising given a greater number of trainig epochs.

```python
ckd_nn = MLPClassifier(max_iter=2000, hidden_layer_sizes=(100, 50, 25,), learning_rate='adaptive', random_state=42).fit(ckd_X_train, ckd_y_train)
print("Neural Network Test Score:", ckd_nn.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
nn_scores = cross_val_score(ckd_nn, ckd_X, ckd_y, cv=kfoldcv)
print("Neural Network Cross-Validation Scores:", nn_scores)
print("Mean Score:", nn_scores.mean())
print("Standard Deviation:", nn_scores.std())
```

```python
ckd_nn = MLPClassifier(max_iter=2000, random_state=42).fit(ckd_X_train, ckd_y_train)
print('More Training Neural Net:')
print("Neural Network Test Score:", ckd_nn.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
nn_scores = cross_val_score(ckd_nn, ckd_X, ckd_y, cv=kfoldcv)
print("Neural Network Cross-Validation Scores:", nn_scores)
print("Mean Score:", nn_scores.mean())
print("Standard Deviation:", nn_scores.std())
```

```python
ckd_nn = MLPClassifier(learning_rate='adaptive', random_state=42).fit(ckd_X_train, ckd_y_train)
print('Adaptive Learning Neural Net:')
print("Neural Network Test Score:", ckd_nn.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
nn_scores = cross_val_score(ckd_nn, ckd_X, ckd_y, cv=kfoldcv)
print("Neural Network Cross-Validation Scores:", nn_scores)
print("Mean Score:", nn_scores.mean())
print("Standard Deviation:", nn_scores.std())
```

```python
ckd_nn = MLPClassifier(hidden_layer_sizes=(100, 50, 25,), random_state=42).fit(ckd_X_train, ckd_y_train)
print('Larger Neural Net:')
print("Neural Network Test Score:", ckd_nn.score(ckd_X_test, ckd_y_test))

kfoldcv = KFold(n_splits=5, random_state=42, shuffle=True)
nn_scores = cross_val_score(ckd_nn, ckd_X, ckd_y, cv=kfoldcv)
print("Neural Network Cross-Validation Scores:", nn_scores)
print("Mean Score:", nn_scores.mean())
print("Standard Deviation:", nn_scores.std())
```
