```python
# Imports section
import pandas as pd
import sklearn as sk
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
import numpy as np
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

df = pd.read_csv(r"C:\Users\sagea\OneDrive\CS 383\Assignment 4\science_data_large.csv")
print(df.head(15))
print("\n\n\nThe table information:")
df.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X = df.drop(columns=['Size nm^3']) 
y = df['Size nm^3']   
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.10, train_size=0.90, random_state=42, shuffle=True)
print(Xtrain)
print(ytrain)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression( fit_intercept=True, copy_X=True, n_jobs=None, positive=False)
model.fit(Xtrain, ytrain)
# Create a sample datapoint and predict the output of that sample with the trained model


ypred = model.predict(Xtest)

# Report the score (R-squared value)
# For Linear Regression, the score method returns R^2
score = model.score(Xtest, ytest)
print("The score is: ", score)
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

Report on score and explain meaning: The score is .85 which means that the linear regression model is performing fairly well and the model is fitting the data correctly and predicting the data about 85% correctly.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
# Use cross_val_score to evaluate the model
cvscores = cross_val_score(model, X, y, cv=kf)

print(cvscores)
# Report on their finding and their significance
```

Write findings here: The findings are all values in the .80-.90 range. The significance shows that the linear regression model is averaging pretty well when it comes to making accurate predictons based on random samples form the data.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

ckddf = pd.read_csv(r"C:\Users\sagea\OneDrive\CS 383\Assignment 4\ckd_feature_subset.csv")

print(ckddf.head())
Xckd = ckddf.drop(columns=['Target_ckd'])
yckd = ckddf['Target_ckd']  

Xc_train, Xc_test, yc_train, yc_test = train_test_split(Xckd, yckd, test_size=0.1, train_size=.90, random_state=42)
```

## Results and Conclusion for Classification Experiments

### Logistic Regression:

```python
LGmodel = LogisticRegression()
LGmodel.fit(Xc_train, yc_train)

gkf = KFold(n_splits=5, shuffle=True, random_state=42)
# Use cross_val_score to evaluate the model

LGcvscores = cross_val_score(LGmodel, Xckd, yckd, cv=gkf)
LGstddev = np.std(LGcvscores)
LGmean = np.mean(LGcvscores)
print(f"The cross validation scores for the logistic regression are: {LGcvscores}")
print(f"The standard deviation for the logistic regression CV scores is: {LGstddev}")
print(f"The the mean for the logistic regression CV scores is: {LGmean}")
# Report on their finding and their significance
```

### Support Vector Models

```python
SVMmodel = svm.SVC(kernel='linear')
SVMmodel.fit(Xc_train, yc_train)
gkf = KFold(n_splits=5, shuffle=True, random_state=42)

SVMcvscores = cross_val_score(SVMmodel, Xckd, yckd, cv=gkf)
SVMstddev = np.std(SVMcvscores)
SVMmean = np.mean(SVMcvscores)
print(f"The cross validation scores for the SVM are: {SVMcvscores}")
print(f"The standard deviation for the SVM CV scores is: {SVMstddev}")
print(f"The the mean for the SVM CV scores is: {SVMmean}")
# Report on their finding and their significance
```

### K-Nearest Neighbors

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(Xc_train)
X_test_scaled = scaler.transform(Xc_test)
knn = KNeighborsClassifier(n_neighbors=5) # Choose k=5 (or another appropriate value)
knn.fit(X_train_scaled, yc_train)

gkf = KFold(n_splits=5, shuffle=True, random_state=42)

knncvscores = cross_val_score(knn, Xckd, yckd, cv=gkf)
knnstddev = np.std(knncvscores)
knnmean = np.mean(knncvscores)
print(f"The cross validation scores for the KNN are: {knncvscores}")
print(f"The standard deviation for the KNN CV scores is: {knnstddev}")
print(f"The the mean for the KNN CV scores is: {knnmean}")
# Report on their finding and their significance
```

```python
NNmodel = MLPClassifier(hidden_layer_sizes=(10,), max_iter=2000, solver='adam', random_state=42)

# Train the model
NNmodel.fit(Xc_train, yc_train)

# Predict on the test set

gkf = KFold(n_splits=5, shuffle=True, random_state=42)
NNcvscores = cross_val_score(NNmodel, Xckd, yckd, cv=gkf)
NNstddev = np.std(NNcvscores)
NNmean = np.mean(NNcvscores)
print(f"The cross validation scores for the Neural Network are: {NNcvscores}")
print(f"The standard deviation for the Network CV scores is: {NNstddev}")
print(f"The the mean for the Neural Network CV scores is: {NNmean}\n")
```
