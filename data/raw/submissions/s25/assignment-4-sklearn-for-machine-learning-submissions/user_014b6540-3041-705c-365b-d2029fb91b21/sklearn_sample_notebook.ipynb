```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Imports section
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

data = pd.read_csv('science_data_large.csv')
display(data.head(15))
display(data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = data.drop('Size nm^3', axis=1)
y = data['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
# Create a sample datapoint and predict the output of that sample with the trained model
model = LinearRegression()
model.fit(X_train, y_train)

sample = X_test[0:1]
predict = model.predict(sample)
print(f"Predicted ouput: {predict}")
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print(f"Model score: {score:.5f}")
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

coeffs = model.coef_
intercept = model.intercept_

print("Coefficients:")
for name, coeff in zip(X.columns, coeffs):
	print(f"Coefficient for {name}: {coeff:.5f}")
print(f"Intercept: {intercept:.5f}")
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
from sklearn.model_selection import cross_val_score, KFold

# Define the cross-validation strategy
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

# Report on their finding and their significance
print(f"Cross validation scores: {scores}")
print(f"Mean score: {scores.mean():.5f}")
```

Write findings here:

The folding for the cross validation scores helps with in assesing how the model generalizes new data to prevent and identify overfitting or underfitting. The shuffling ensures that the data is shuffled in between folds and the random state adds extra variables so that it can be reproduced. The scoring is specifies R^2 which is the scoring system to evaluate the accuracy of the model. 

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=2)),
    ('model', LinearRegression())
])

kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(pipeline, X, y, cv=kf, scoring='r2')

print(f"Cross validation scores with polynomial features: {scores}")
print(f"Mean score: {scores.mean()}")

# Fit the pipeline on the entire dataset to get the coefficients
pipeline.fit(X, y)

model = pipeline.named_steps['model']

coeffs = model.coef_
intercept = model.intercept_

# Extract feature names for polynomial features
poly_feature_names = pipeline.named_steps['poly'].get_feature_names_out(X.columns)

print("Coefficients:")
for name, coeff in zip(poly_feature_names, coeffs):
    print(f"Coefficient for {name}: {coeff:.5f}")
print(f"Intercept: {intercept:.5f}")
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

data = pd.read_csv('ckd_feature_subset.csv')
display(data.head(15))
display(data.info())

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

models = {"Logistic Regression": LogisticRegression(),
    "Support Vector Classifier": SVC(),
    "k-Nearest Neighbors": KNeighborsClassifier(),
}

results = []
X = data.drop('Target_ckd', axis=1)
y = data['Target_ckd']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)


# Perform 5-fold cross-validation for each model
for model_name, model in models.items():
    cv_scores = cross_val_score(model, X, y, cv=5)
    results.append({
        "Model": model_name,
        "Average Accuracy": np.mean(cv_scores),
        "Standard Deviation": np.std(cv_scores)
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
print(results_df)


```

## Results and Conclusion for Classification Experiments

The results of the three models, are that out of the three, the logistic regression model had the worst average acuracy with a .05 std. The SVC and the k-nearest had the same exact acuraccy but the k-nearest was better because it had a smaller standard deviation. 

```python
# Experiments with Neural Network.

nn_configs = {
    "NN-Config 1": MLPClassifier(hidden_layer_sizes=(10,),activation='relu', max_iter=500, random_state=42),
    "NN-Config 2": MLPClassifier(hidden_layer_sizes=(10,),activation='logistic', max_iter=500, random_state=42),
    "NN-Config 3": MLPClassifier(hidden_layer_sizes=(10, 10), activation='relu', max_iter=500, random_state=42),
    "NN-Config 4": MLPClassifier(hidden_layer_sizes=(10, 10), activation='logistic', max_iter=500, random_state=42),}

results = []
# Perform 5-fold cross-validation for each Neural Network configuration
for config_name, nn_model in nn_configs.items():
    cv_scores = cross_val_score(nn_model, X, y, cv=5)
    results.append({
        "Model": config_name,
        "Average Accuracy": np.mean(cv_scores),
        "Standard Deviation": np.std(cv_scores)
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Print the final results including NN configurations
print(results_df)
```

## Results and Conclusion for Neural Network Experiments

I ran 4 different models, two with relu activation and two with logistic activation. The relu activations had better accuracy than the logistic activations. I also changed the layer sizes. NN1 and NN3 only had 1 single hidden layer with 10 neurons, and NN2 and NN4 had 2 single layers with 10 neurons each. With each layer, the model becomes more accurate as shown above. As well as relu activation is more accurate than logistic activation for this data set. The relu had more accuracy with more layers, but the logistic one had less acuraccy with more layers but a smaller standard deviation. Overall the relu activation had the best accuracy.
