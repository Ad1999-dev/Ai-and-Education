# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_data = pd.read_csv('ckd_feature_subset.csv')
print(ckd_data.head(15))
ckd_data.info()
```

## Results and Conclusion for Classification Experiments

```python
# Step 1: Split into features and label
X_ckd = ckd_data.drop(columns=['Target_ckd'])
y_ckd = ckd_data['Target_ckd']

# Step 2: Use a fixed random seed of 42
X_train_ckd, X_test_ckd, y_train_ckd, y_test_ckd = train_test_split(X_ckd, y_ckd, test_size=0.1, random_state=42)

# Step 3: Scale the features
scaler = StandardScaler()
X_train_ckd = scaler.fit_transform(X_train_ckd)
X_test_ckd = scaler.transform(X_test_ckd)

# Step 4: Define the models
models = {
    'Logistic Regression': LogisticRegression(max_iter=2500),  # Increased max_iter
    'SVC': SVC(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'MLPClassifier': MLPClassifier(max_iter=2500)  # Increased max_iter
}

results = {}

# Step 5: Evaluate the models using cross-validation
for name, model in models.items():
    cv_scores = cross_val_score(model, X_train_ckd, y_train_ckd, cv=5)
    results[name] = {
        'mean_score': np.mean(cv_scores),
        'std_score': np.std(cv_scores)
    }
    print(f"{name} Cross-validation scores: {cv_scores}")

# Step 6: Store results in a DataFrame
results_df = pd.DataFrame(results).T
print(results_df)

# Step 7: Example of exploring different configurations of the MLPClassifier
nn_configs = [
    {'hidden_layer_sizes': (5,), 'activation': 'tanh'},
    {'hidden_layer_sizes': (10,), 'activation': 'relu'},
    {'hidden_layer_sizes': (15, 5), 'activation': 'logistic'}  # This is acceptable if 'logistic' is recognized
]

nn_results = {}

# Step 8: Evaluate different neural network configurations
for config in nn_configs:
    model = MLPClassifier(hidden_layer_sizes=config['hidden_layer_sizes'], 
                          activation=config['activation'], 
                          max_iter=2500, 
                          verbose=False) 
    cv_scores = cross_val_score(model, X_train_ckd, y_train_ckd, cv=5)
    nn_results[f"{config['hidden_layer_sizes']}, {config['activation']}"] = {
        'mean_score': np.mean(cv_scores),
        'std_score': np.std(cv_scores)
    }

# Step 9: Store neural network results in a DataFrame
nn_results_df = pd.DataFrame(nn_results).T
print(nn_results_df)
```

## Results and Conclusion for Neural Network Experiments

The evaluation of various classification models on the Chronic Kidney Disease (CKD) dataset yielded insightful results regarding their performance. Starting with Logistic Regression, this model achieved cross-validation scores ranging from approximately 0.889 to 0.964, which culminated in a mean score of 0.926720 with a standard deviation of 0.033438. These results indicate that Logistic Regression can effectively discriminate between classes, although the variability shows it may struggle with particular subsets of the data. The Support Vector Classifier (SVC) outperformed the other models, achieving perfect scores in three of the five folds. This resulted in a mean score of 0.962963 and a standard deviation of 0.040572, suggesting a robust ability to classify CKD instances accurately.


Additionally, the KNeighbors Classifier exhibited more variability in its scores, with a mean of 0.897090 and a higher standard deviation of 0.072288, indicating potential sensitivity to the choice of neighbors and the presence of noise in the data. Lastly, the MLP Classifier demonstrated strong performance across the board, with cross-validation scores averaging 0.955556 and a standard deviation of 0.043192. Notably, while the MLP Classifier's performance was commendable, it is important to consider the configurations tested for hyperparameter tuning. The configurations with different hidden layers yielded several interesting results; the MLP with 15 hidden nodes and logistic activation reached a mean score of 0.963228. This indicates that while neural networks can adequately capture complex patterns in the data, the choice of architecture significantly impacts performance stability. Overall, the SVC and MLP models emerged as the most effective for this classification task, with logistic regression providing a strong baseline performance. Further tuning, especially of the MLP Classifier, may reveal even better model fits for the CKD dataset.
