```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print("First 15 Rows of the Dataset:")
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print("\nSummary of the Dataset:")
print(data.info())
print(data.describe())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = data[['Temperature °C', 'Mols KCL']]  # Adjust the feature columns as necessary
y = data['Size nm^3']  # Adjust the label column name accordingly

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)

# For grading consistency use random_state=42 

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model

sample_datapoint = [[25, 0.5]]  # 25 °C, 0.5 Mols KCL
predicted_size = model.predict(sample_datapoint)

print(f"Predicted Size for sample datapoint (Temperature: 25°C, Mols KCL: 0.5): {predicted_size[0]} nm^3")

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

score = model.score(X_test, y_test)
print(f"Model Score: {score}")

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

coefficients = model.coef_
intercept = model.intercept_

# For example, using coefficients list: [coef_temp, coef_kcl]
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")

# LaTeX equation for the model
equation = f"h(x) = {intercept:.5f} + ({coefficients[0]:.5f} * Temperature) + ({coefficients[1]:.5f} * Mols KCL)"
print(f"Equation: {equation}")
```

## **Linear Equation for Slime Growth**  

From our trained model, the equation predicting slime size based on temperature and KCL concentration is:  

\[
h(x) = -409391.47958 + (866.14641 *Temperature) + (1032.69507 * Mols KCL)
\]

##### **The General Equation**: h(x)= β0 + β1(Temperature ∘C) + β2(Mols KCL)

This equation suggests that for each additional degree Celsius, the predicted slime size increases by **866.15 nm³**, and for each additional **1 mol** of KCL, the slime size increases by **1032.70 nm³**.  

---

## **Model Score and Explanation**  

The model achieved a score of **0.855**, which refers to the **coefficient of determination** (\( R^2 \) score). This metric measures how well the model's predictions match the actual data.  

- An \( R^2 \) score of **1.0** indicates a perfect fit, meaning the model explains **100%** of the variance in slime growth.  
- An \( R^2 \) score of **0** means the model performs no better than predicting the average slime size.  
- Our score of **0.855** means that about **85.5% of the variance** in slime size is explained by the model. This suggests a **strong fit**, but there is still some room for improvement.

## 5. Using Polynomial Regression

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold


# Step 1: Augment the original feature set using degree 2 polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 2: Create a Linear Regression model
model = LinearRegression()

# Step 3: Perform 5-fold cross validation with shuffling for reproducibility
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores_poly = cross_val_score(model, X_poly, y, cv=kf)

# Step 4: Report the mean and standard deviation of the cross-validation scores
mean_score_poly = scores_poly.mean()
std_dev_score_poly = scores_poly.std()
print(f"Cross-Validation Scores (Polynomial Features): {scores_poly}")
print(f"Mean Score (Polynomial Features): {mean_score_poly:.4f}")
print(f"Standard Deviation of Scores (Polynomial Features): {std_dev_score_poly:.4f}")

# Step 5: Fit the model to the full dataset of polynomial features
model.fit(X_poly, y)

# Step 6: Extract coefficients and intercept from the trained model
intercept_poly = model.intercept_
coefficients_poly = model.coef_
print(f"Intercept: {intercept_poly}")
print(f"Coefficients: {coefficients_poly}")

# Generate the equation in LaTeX
# If X is a DataFrame, we can extract the original feature names
feature_names = poly.get_feature_names_out(X.columns)
equation = f"h(x) = {intercept_poly:.5f}"
for name, coef in zip(feature_names, coefficients_poly):
    equation += f" + ({coef:.5f} \\times {name})"

print("LaTeX Equation:")
print(equation)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
data = pd.read_csv('ckd_feature_subset.csv')

print(data.head(15))
print("\nSummary of the Dataset:")
print(data.info())
```

```python
# Define features (X) and target labels (y)
X = data.drop('Target_ckd', axis=1)  # Features
y = data['Target_ckd']                 # Target variable

# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)

# Confirm the sizes of train and test sets
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

## Results and Conclusion for Classification Experiments

```python
# Import necessary libraries for classification
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score

# Initialize models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(max_iter=1000, random_state=42)
}

# Evaluate models using 5-fold cross-validation
results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5)
    results[name] = {
        "mean_score": scores.mean(),
        "std_dev": scores.std()
    }

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results).T
results_df
```

### Findings:

- The **Neural Network** model achieved the highest mean score of **0.948**, indicating it is the most effective classifier for this dataset among those evaluated. The low standard deviation of **0.038** suggests that the model is stable across different folds.
- The **Support Vector Machine (SVM)** model closely follows with a mean score of **0.919**. This model also performed very well, suggesting it can effectively delineate the decision boundaries between the classes. However, it was slightly less effective than the neural network.
- The **K-Nearest Neighbors** model had a mean score of **0.9116**. While still a robust classifier, it was outperformed by both the SVM and the Neural Network. The relatively high variance (standard deviation of **0.07296**) suggests that KNN may be sensitive to the specific partition of the data used in cross-validation.
- The **Logistic Regression** model achieved the lowest mean score of **0.8677** among the models evaluated. Despite being a simpler model, it still produced reasonable results. However, its performance indicates that it may not capture the complexities of the dataset as effectively as the other three models. The standard deviation of 0.0612 suggests a stable performance across the folds, though lower overall accuracy.


Overall, the Neural Network is concluded to be the most suitable model for this classification task based on the feature set available.

## Results and Conclusion for Neural Network Experiments

```python
# Experimenting with different settings for the Neural Network
nn_configs = [
    {'hidden_layer_sizes': (10,), 'activation': 'logistic'},
    {'hidden_layer_sizes': (10,), 'activation': 'relu'},
    {'hidden_layer_sizes': (20,), 'activation': 'tanh'},
]

# Reporting findings for different configurations
nn_results = {}
for config in nn_configs:
    nn_model = MLPClassifier(hidden_layer_sizes=config['hidden_layer_sizes'], activation=config['activation'], max_iter=2000, random_state=42)
    scores = cross_val_score(nn_model, X_train, y_train, cv=5)
    nn_results[f"Layers: {config['hidden_layer_sizes']}, Activation: {config['activation']}"] = {
        "mean_score": scores.mean(),
        "std_dev": scores.std()
    }

nn_results_df = pd.DataFrame(nn_results).T
nn_results_df
```

### Summary of Findings

1. **Performance**:
   - The configuration with **20 hidden neurons and Tanh activation** achieved the highest mean score of **0.933862**. This indicates that it is the most effective at predicting Chronic Kidney Disease among the tested configurations.

2. **Effect of Activation Functions**:
   - Both configurations with **10 hidden neurons**, regardless of whether logistic or ReLU activations were used, produced identical mean scores of **0.911905**. This suggests that adding more complexity (in terms of hidden layers or neurons) can improve model performance, since the single-layer configurations are less effective.

3. **Standard Deviations**:
   - The standard deviation for the Tanh configuration is the lowest at **0.036724**, indicating that this model is providing consistent predictions across different data folds. In contrast, the standard deviations for the logistic and ReLU configurations are higher (**0.055830**), which suggests slightly more variability in performance.
