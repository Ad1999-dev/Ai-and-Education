```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
sciDat = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
pd.DataFrame(sciDat.head(15))

```

```python
# Display a summary of the table information (number of datapoints, etc.)
sciDat.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
feat = sciDat[["Temperature °C","Mols KCL"]]
lab = sciDat["Size nm^3"]

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
xTr, xTe, yTr, yTe = train_test_split(feat, lab, test_size=.1, train_size=.9, random_state=42)
xTr.shape, xTe.shape, yTr.shape, yTe.shape
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
reg = LinearRegression().fit(xTr, yTr)

# Create a sample datapoint and predict the output of that sample with the trained model
sInd = np.random.randint(0, len(xTe))
print("Sample Index: " + str(sInd))
sample = xTe.iloc[[sInd]]
print("Sample: " + str(sample))
sampleO = reg.predict(sample)
print("Prediction: " + str(sampleO[0]))
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print("Score: " + str(reg.score(xTe, yTe)))

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print("Coefficient 1: " + str(reg.coef_[0]))
print("Coefficient 2: " + str(reg.coef_[1]))
print("Intercept: " + str(reg.intercept_))
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, random_state=42, shuffle=True)
cvs = cross_val_score(reg, feat, lab, cv = kf)

# Report on their finding and their significance
print("CVS Mean: " + str(cvs.mean()))
print("CVS Standard Deviation: " + str(cvs.std()))
print("CVS Scores: " + str(cvs))
```

Write findings here:

Our cross validation scores for 5 folds are roughly 0.86152, 0.82742, 0.87195, 0.88166, and 0.85609 which have a mean of about .85973 which is quite close to our test score of about 0.85525. But, our standard deviation makes up for this difference, and thus, our original test falls well within the bounds of what was expected. Therefore we can conclude our model is not over/under fit and makes relatively reliable predictions.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import sklearn.neural_network as nn
```

```python
# Load the dataset. Then train and evaluate the classification models.
# Using pandas load the dataset
ckdDat = pd.read_csv('ckd_feature_subset.csv')
feat = ckdDat[["age","bp","wbcc","appet_poor","appet_good","rbcc"]]
lab = ckdDat["Target_ckd"]

models = {
"Logistic Regression" : LogisticRegression(random_state=42),
"Support Vector Machines" : SVC(random_state=42),
"k-Nearest Neighbors" : KNeighborsClassifier(),
"Neural Networks" : nn.MLPClassifier(max_iter=1000, random_state=42)
}

results = []

for modelN, model in models.items():
    cvs = cross_val_score(model, feat, lab, cv = kf)
    results.append([modelN,str(cvs.mean()), str(cvs.std())])

pd.DataFrame(results, columns = ["Model", "Mean Accuracy", "Standard Deviation"])
    
```

## Results and Conclusion for Classification Experiments

The performance of four different machine learning models was evaluated, with results indicating that Neural Networks achieved the highest mean accuracy of approximately 93.51%, followed closely by Support Vector Machines at 92.82% and k-Nearest Neighbors at 92.80%. Logistic Regression provided a solid baseline with an accuracy of around 85.66%, though it lagged behind the others. Notably, Neural Networks exhibited the lowest standard deviation (0.0405), suggesting the most consistent performance across cross-validation folds. Overall, this analysis highlights the superiority of complex models like Neural Networks in classification tasks while emphasizing the importance of stability in model performance.

```python
# Experiments with Neural Network.

def evaluate_model(model): 
    cvs = cross_val_score(model, feat, lab, cv=kf)
    return cvs.mean(), cvs.std()

layer_variations = {
    'Single Layer 100': nn.MLPClassifier(hidden_layer_sizes=(100,), max_iter=2000, random_state=42),
    'Two Layers 100, 200': nn.MLPClassifier(hidden_layer_sizes=(100, 200), max_iter=2000, random_state=42),
}

solver_variations = {
    'Adam': nn.MLPClassifier(solver='adam', max_iter=2000, random_state=42),
    'SGD': nn.MLPClassifier(solver='sgd', max_iter=2000, random_state=42),
    'LBFGS': nn.MLPClassifier(solver='lbfgs', max_iter=2000, random_state=42),
}

activation_variations = {
    'ReLU': nn.MLPClassifier(activation='relu', max_iter=2000, random_state=42),
    'Tanh': nn.MLPClassifier(activation='tanh', max_iter=2000, random_state=42),
    'Logistic': nn.MLPClassifier(activation='logistic', max_iter=2000, random_state=42),
}

results = {}
for variation in [layer_variations, solver_variations, activation_variations]:
    for name, model in variation.items():
        results[name] = evaluate_model(model)

pd.DataFrame(results, index=['Mean Accuracy', 'Standard Deviation']).T
```

## Results and Conclusion for Neural Network Experiments

I did a pretty exhaustive test of the possible variations you can create using the SKLearn neural net. In my findings, i discovered that when you increase the complexity, or neuron count, as one may expect the accuracy goes up. When it comes to solvers, I found that the LBFGS optimization algorithm provided the most accurate outputs. This may be because it is made specifically for smaller datasets (which this one very much is) so that definitely makes sense. Finally, with activation functions Tanh performed the best of the 3 provided.
