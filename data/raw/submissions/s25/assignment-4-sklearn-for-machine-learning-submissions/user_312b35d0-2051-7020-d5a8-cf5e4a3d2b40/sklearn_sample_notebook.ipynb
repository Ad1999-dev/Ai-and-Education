```python
# Imports section
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data.drop('Size nm^3',axis=1)
y = data['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample = data.loc[[0]].drop(columns=['Size nm^3'])
sample_pred = model.predict(sample)
print("Sample actual size: ", data.iloc[0]['Size nm^3'])
print("Sample predicted size: ", sample_pred[0])

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print("Model score: ", score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_
print(coefficients, intercept)
```

Report on score and explain meaning: The $R^2$ score describes how well the linear regression model predicted the labels based off of how far the predicted value was from the actual value (in other words, the residuals). More mathemtically and precisely, it is equal to $(1-(\frac{u}{v}))$, where $u$ is the residual sum of squares and $v$ is the total sum of squares.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kfold = KFold(n_splits = 5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=kfold)

# Report on their finding and their significance
print("Scores:", scores)
print("Average score: ", np.mean(scores))
print("Standard deviation: ", np.std(scores))
```

Write findings here: The mean accuracy of the cross validation accuracy is 0.85973 and the standard deviation is 0.01839. We can see that the cross validation accuracy is very close to the R2 value reported by model.score, and the standard deviation is quite small. This is significant because it shows that our model gives pretty consistent results across different folds, and that these results are decently high.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
pr = PolynomialFeatures(degree=2)
X_poly = pr.fit_transform(X)

# Perform k-fold cross validation (as above)
kfold = KFold(n_splits = 5, shuffle=True, random_state=42)
scores_pr = cross_val_score(model, X_poly, y, cv=kfold, scoring='r2')

# Report on the metrics and output the resultant equation as you did in Part 3.
print("Scores:", scores_pr)
print("Average score: ", np.mean(scores_pr))
print("Standard deviation: ", np.std(scores_pr))

coefficients = model.coef_
intercept = model.intercept_
print(coefficients, intercept)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_data = pd.read_csv('ckd_feature_subset.csv')
ckd_data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

X_ckd = ckd_data.drop('Target_ckd', axis=1)
y_ckd = ckd_data['Target_ckd']
X_train_ckd, X_test_ckd, y_train_ckd, y_test_ckd = train_test_split(X_ckd, y_ckd, test_size=0.1, random_state=42)

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train_ckd, y_train_ckd)
print('Logistic Regression Score:', lr_model.score(X_test_ckd, y_test_ckd))

# Support Vector Machines
svm_model = SVC()
svm_model.fit(X_train_ckd, y_train_ckd)
print('SVM Score:', svm_model.score(X_test_ckd, y_test_ckd))

# k-Nearest Neighbors
knn_model = KNeighborsClassifier()
knn_model.fit(X_train_ckd, y_train_ckd)
print('k-Nearest Neighbors Score:', knn_model.score(X_test_ckd, y_test_ckd))

# Neural Networks
nn_model = MLPClassifier(learning_rate_init=0.1)
nn_model.fit(X_train_ckd, y_train_ckd)
print('Neural Networks Score:', nn_model.score(X_test_ckd, y_test_ckd))
```

## Results and Conclusion for Classification Experiments

```python
# Experiments with Neural Network.
nn_model_1 = MLPClassifier(solver='sgd', learning_rate_init=0.1)
nn_model_1.fit(X_train_ckd, y_train_ckd)
print('Neural Networks Score 1:', nn_model_1.score(X_test_ckd, y_test_ckd))

nn_model_2 = MLPClassifier(solver='sgd', activation='logistic',learning_rate_init=0.01)
nn_model_2.fit(X_train_ckd, y_train_ckd)
print('Neural Networks Score 2:', nn_model_2.score(X_test_ckd, y_test_ckd))

nn_model_3 = MLPClassifier(activation='tanh',learning_rate_init=0.01) # default solver='adam'
nn_model_3.fit(X_train_ckd, y_train_ckd)
print('Neural Networks Score 3:', nn_model_3.score(X_test_ckd, y_test_ckd))

nn_model_4 = MLPClassifier(activation='identity',learning_rate_init=0.1) 
nn_model_4.fit(X_train_ckd, y_train_ckd)
print('Neural Networks Score 4:', nn_model_4.score(X_test_ckd, y_test_ckd))
```

## Results and Conclusion for Neural Network Experiments

The best result from the experiments was using the default 'adam' solver (a stochastic gradient-based optimizer), default 'relu' activation function, and an initial learning rate of 0.1 to get an accuracy of 0.9375. From the other experiments, the worst result was from using an sgd solver (stochastic gradient descent), logistic activation function, and initial learning rate of 0.01. Other different configurations of activation functions and use of sgd solver resulted in the same accuracy of 0.875.
