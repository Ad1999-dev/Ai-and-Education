```python
# Imports section
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
dfsci = pd.read_csv("science_data_large.csv")
print(dfsci)
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X = dfsci.drop(columns=['Size nm^3'])  
y = dfsci[['Size nm^3']]           

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(X_train)
#print(y_train)
print(X_test)
#print(y_test)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

model = LinearRegression()
model.fit(X_train, y_train)
score = model.score(X_test, y_test)

print(score)

coefficients = model.coef_
intercept = model.intercept_

print(f"coefficients: {coefficients}")
print(f"intercept: {intercept}")

#sample datapoint 
a = 100
b = 541
prediction = 866.14641*a +1032.69507*b - 409391.47958 
print(prediction)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
shuffled = KFold(n_splits=5, random_state=42, shuffle=True)
scores = cross_val_score(model, X_train, y_train, cv = shuffled)
# Report on their finding and their significance
print(f"scores: {scores}")
print(f"standard deviation = {scores.std()}")
print(f"average = {scores.mean()}", )
```

Write findings here: Even when the data is shuffled it still generally looks like a score of 0.84 to 0.88 so not a huge variance with the way kfold is moving things around. This tells us that the model is pretty stable and not under or over fitting too much. 

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
# Report on the metrics and output the resultant equation as you did in Part 3.

polymodel = LinearRegression()
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
X_polytest = poly.transform(X_test)
polymodel.fit(X_poly, y_train)

scores_poly = cross_val_score(polymodel, X_poly, y_train, cv = shuffled)
score_poly = polymodel.score(X_polytest, y_test)
polycoefficients = polymodel.coef_
polyintercept = polymodel.intercept_

print(f"coefficients: {polycoefficients}")
print(f"intercept: {polyintercept}")
print(f"cross validation scores = {scores_poly}")
print(f"standard deviation = {scores_poly.std()}")
print(f"average = {scores_poly.mean()}", )

foo = np.array([[541, 100]])
foe = poly.transform(foo)
pred = polymodel.predict(foe)
print(f"prediction = {pred}")
```

# Part 2. Chronic Kidney Disease Prediction via Classification

```python
# Load the dataset. Then train and evaluate the classification models.

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

# Load the dataset
ckd_data = pd.read_csv('ckd_feature_subset.csv')
X = ckd_data.drop(columns='Target_ckd')
y = ckd_data['Target_ckd']

# Initialize models
logistic_model = LogisticRegression(max_iter=1000)
svc_model = SVC()
knn_model = KNeighborsClassifier()

def evaluate_model(model, X, y, shuffled, name=None):
    name = model.__class__.__name__
    scores = cross_val_score(model, X, y, cv=shuffled, scoring='accuracy')
    
    print(f"Model: {name}")
    print(f"Cross-validation scores: {scores}")
    print(f"Average accuracy: {scores.mean():.5f}")
    print(f"Standard deviation: {scores.std():.5f}")
    print('-' * 80)

# Evaluate models
scores_logreg = evaluate_model(logistic_model, X, y, shuffled)
scores_svc = evaluate_model(svc_model, X, y, shuffled)
scores_knn = evaluate_model(knn_model, X, y, shuffled)
```
