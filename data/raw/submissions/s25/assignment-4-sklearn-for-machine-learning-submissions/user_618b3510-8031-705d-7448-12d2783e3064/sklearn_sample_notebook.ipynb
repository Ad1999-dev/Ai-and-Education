```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

slime_data = pd.read_csv("science_data_large.csv")
print(slime_data.head(15))
slime_data.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X = slime_data[['Temperature °C', 'Mols KCL']]
y = slime_data['Size nm^3']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
model = LinearRegression()
model.fit(X_train, y_train)

sample_data = pd.DataFrame({
    "Temperature °C": [25],
    "Mols KCL": [0.5]
})

prediction = model.predict(sample_data)
print("Prediction of the output of that sample with the trained model:", prediction)

score = model.score(X_test, y_test)
print("Model score:", score)

coeffs = model.coef_
intercept = model.intercept_
print("Coefficients:", coeffs)
print("Intercept:", intercept)
```

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Report on their finding and their significance
from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=5, random_state=42, shuffle=True)
scores = cross_val_score(model, X, y, cv=kf)
# print("Cross Validation scores for each fold:", scores)
# print("Average Cross Validation score:", scores.mean())
```

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_model = LinearRegression()

kf = KFold(n_splits=5, random_state=42, shuffle=True)

scores_poly = cross_val_score(poly_model, X_train_poly, y_train, cv=kf)
print("Cross validation scores for each fold:", scores_poly)
print("Average cross validation score:", scores_poly.mean())

poly_model.fit(X_train_poly, y_train)

coeffs_poly = poly_model.coef_
intercept_poly = poly_model.intercept_
print("Polynomial Regression Coefficients:", coeffs_poly)
print("Polynomial Regression Intercept:", intercept_poly)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

ckd_data = pd.read_csv("ckd_feature_subset.csv")

X = ckd_data.drop("Target_ckd", axis=1)
y = ckd_data["Target_ckd"]

print(ckd_data.head())
```

## Training and evaluating the classification models

```python
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "SVC": SVC(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(random_state=42, max_iter=1000)
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# use cross validation and store results
results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
    results[name] = scores
    # print(f"{name} Accuracy scores: {scores}")
    # print(f"{name} Mean Accuracy: {scores.mean():.4f}, Std: {scores.std():.4f}\n")

# create summary table
results_table = pd.DataFrame({
    "Model": list(results.keys()),
    "Mean Accuracy": [np.mean(scores) for scores in results.values()],
    "Std Accuracy": [np.std(scores) for scores in results.values()]
})
print(results_table)
```

Based on the cross validation results, the neural network performed the best. It achieved a mean accuracy of approximately 94.75% with a standard deviation of 3.94%, which is higher than the other models. Logistic regression had a 86.24% mean accuracy, SVC had a 93.40% mean accuracy, and KNN had a 93.40% mean accuracy. The neural network is more effective at correctly classifying data and also more stable 

The higher mean accuracy shows that the neural network is more effective at correctly classifying the CKD data, and its relatively low standard deviation suggests that its performance is stable across different folds of the data.

## Results and Conclusion for Classification Experiments

```python
configs = {
    "Single layer 50": MLPClassifier(random_state=42, max_iter=1000, hidden_layer_sizes=(50,)),
    "Single layer 100": MLPClassifier(random_state=42, max_iter=1000, hidden_layer_sizes=(100,)),
    "Two layers (100,50)": MLPClassifier(random_state=42, max_iter=1000, hidden_layer_sizes=(100, 50))
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {}
for name, model in configs.items():
    scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
    results[name] = scores
    # print(f"{name} Accuracy scores: {scores}")
    # print(f"{name} Mean Accuracy: {scores.mean():.4f}, Std: {scores.std():.4f}\n")

results_table = pd.DataFrame({
    "Configuration": list(results.keys()),
    "Mean accuracy": [np.mean(scores) for scores in results.values()],
    "Std accuracy": [np.std(scores) for scores in results.values()]
})
print(results_table)
```

## Results and Conclusion for Neural Network Experiments

Looking at the single hidden layer to the two layer architecture, the two layer architecture resulted in a significant improvement in mean accuracy (94-95% to 97.38%). The standard deviation also dropped. We discover that the two layer models are more accurate and more consistent across different folds. 

Increasing the number of neurons in a single layer from 50 to 100 gave a slight improvment in performance but not as much as adding an additional layer. These results tell us that adding an extra hidden layer provides the most impact on improving classification performance.
