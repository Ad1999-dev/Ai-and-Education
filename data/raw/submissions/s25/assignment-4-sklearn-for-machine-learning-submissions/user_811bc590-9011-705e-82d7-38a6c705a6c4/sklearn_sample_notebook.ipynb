## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

science_data = pd.read_csv("science_data_large.csv")

display(science_data.head(15))
science_data.describe()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = science_data[['Temperature °C', 'Mols KCL']]
y = science_data['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Use sklearn to train a model on the training set
reg = LinearRegression()
reg.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = pd.DataFrame({'Temperature °C': [450], 'Mols KCL': [500]})
prediction = reg.predict(sample_data)
print("Size prediction based off sample data:", prediction[0])

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score = reg.score(X_test, y_test)
print("Score of model:", score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coeff = reg.coef_
intercept = reg.intercept_
print("Coefficient:", coeff)
print("Intercept:", intercept)
```

## 4. Use Cross Validation

```python
from sklearn.model_selection import cross_val_score, ShuffleSplit
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

n_samples = X.shape[0]
cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=42)
cv_scores = cross_val_score(reg, X, y, cv=cv)

# Report on their finding and their significance
print("Cross-validated scores for each fold:", cv_scores)
print("Mean cross-validated score:", cv_scores.mean())
```

Write findings here:

The scores across 5 folds are: [0.86151889 0.82742341 0.87195173 0.88166206 0.85609101] and the mean score is: 0.8597294202684646.
This mean score is slightly higher than the score found above with just the regular linear regression. Since the score is high and close to 1.0 it means the model will most likely perform well on unseen data meaning it is good at generalization and is not to overfit to the training data. Showing the model has done well to capture underlying patterns in the data.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
# Load the dataset. Then train and evaluate the classification models.

random_seed = 42

ckd_data = pd.read_csv("ckd_feature_subset.csv")

display(ckd_data.head())


X_ckd = ckd_data.drop(columns=["Target_ckd"])
y_ckd = ckd_data["Target_ckd"]

X_ckd_train, X_ckd_test, y_ckd_train, y_ckd_test = sk.model_selection.train_test_split(X_ckd, y_ckd, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

models = {
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Neural Network": MLPClassifier()
}

n_samples = X_ckd.shape[0]
cv = ShuffleSplit(n_splits=5, test_size=0.1, random_state=random_seed)

model_results = {}
for name, model in models.items():
    model.fit(X_ckd_train, y_ckd_train)
    scores = cross_val_score(model, X_ckd, y_ckd, cv=cv)
    model_results[name] = {
        'Average Accuracy': np.average(scores),
        'Standard Deviation': np.std(scores)
    }


results_df = pd.DataFrame(model_results).T
print("Model Performance (5-Fold Cross-Validation):")
display(results_df)
```

## Results and Conclusion for Classification Experiments

KNN performed the best of all the models with the highest average accuracy of 97.5% and the lowest standard deviation of .031. This high accuracy shows that KNN classified the highest proportion of samples and had consistent performance across the different folds of cross-validation shown by its low std. KNN most likely performed well because of the simplicity of its algorithm which makes predictions based on the majority class among the nearest neighbors.

SVM also performed well most likely because of its ability to handle high-dimensional spaces and find complex decision boundaries.

Neural Network performed a little worse than SVM and Logistic Regression performed the worst most likely because it is linear it might struggle with complex non-linear relationships in the data.

```python
# Experiments with Neural Network.

# Experiment with Neural Network configurations
nn_configurations = {
    'Default Configuration': MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=random_seed),
    'More Hidden Layers': MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=random_seed),
    'Different Learning Rate': MLPClassifier(hidden_layer_sizes=(100,), learning_rate_init=0.01, max_iter=1000, random_state=random_seed)
}

nn_results = {}
for config_name, model in nn_configurations.items():
    scores = cross_val_score(model, X_ckd, y_ckd, cv=cv)
    nn_results[config_name] = {
        'Average Accuracy': scores.mean(),
        'Standard Deviation': scores.std()
    }

# Create a results DataFrame for Neural Networks
nn_results_df = pd.DataFrame(nn_results).T
display(nn_results_df)
```

## Results and Conclusion for Neural Network Experiments

The Default Configuration with a single hidden layer had an average accuracy of 96.25% with a std of .05 which is a pretty good performance.

The second configuration with an additional hidden layer had an increase in performance to 98.75% and a decrease in std to .025 showing that adding more hidden layers allowed the model to learn more complex relations in the data increasing performance.

The final configuration with a different learning rate of .01 was able to fit the data perfectly showing that learning rate is very important in improving the models accuracy. The perfect score could potentially hint at an over complexity in this configuration and some overfitting to the data occurring.

From the testing the number of hidden layers and learning rate of a large affect on the accuracy and performance of the neural network model. More hidden layers allow the model allows the model to pick up on more complex relationships in the data. Adjusting the learning rate can lead to quick convergence and better results. It is important to keep a balance in the complexity to prevent the risk overfitting and maintaining a model that has good generalization capabilities.
