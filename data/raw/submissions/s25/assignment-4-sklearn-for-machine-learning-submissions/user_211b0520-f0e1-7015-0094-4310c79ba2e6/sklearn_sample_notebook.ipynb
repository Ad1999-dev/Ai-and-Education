```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

# Using pandas to load the dataset
data = pd.read_csv('science_data_large.csv')  # Ensure the CSV file path is correct

# Output the first 15 rows of the data
print(data.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(data.info())
print(data.describe())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

# Splitting the dataset into features (X) and label (y)
X = data[['Mols KCL', 'Temperature °C']]  # Adjust column names as needed
y = data['Size nm^3']  # Adjust based on how the target is defined

# Use sklearn to split the features and labels into a training/test set (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = np.array([[403, 694]])  # Example: KCl: 500, Temperature: 700
predicted_size = model.predict(sample_data)
print(f'The predicted size for KCl: 500 and Temperature: 700 is: {predicted_size[0]}')

# Report the score for that model (R^2 score)
score = model.score(X_test, y_test)
print(f'Model score (R^2): {score}')

# Extract the coefficients and intercept from the model
coefficients = model.coef_
intercept = model.intercept_
print(f'Coefficients: {coefficients}, Intercept: {intercept}')
```

Report on score and explain meaning:

The R^2 score is a measure of the correlation between what the model predicts and the actual results. An R^2 score of 1 means the model always perfectly predicts the output, while an R^2 score of 0 means the model is not well fit. ## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Report on their finding and their significance

# Create the KFold object with shuffling enabled
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation using the KFold instance 
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')
print(f'Cross-validation scores: {cv_scores}')
print(f'Average CV score: {np.mean(cv_scores)}, Standard deviation: {np.std(cv_scores)}')
```

Write findings here:

The scores from the cross validation are all somewhat similar, with a standard deviation of only ~0.0184. Since the mean value is close to 1, and the standard deviation is close to 0, this means that the model should be fairly consistient at making predictions across a viriety of data sets.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.

# Using the PolynomialFeatures library to create an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)  # Transforming the feature set

# Initialize the polynomial regression model
model_poly = LinearRegression()

# Perform K-Fold cross-validation with the polynomial features
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Collect cross-validation scores
cv_scores_poly = cross_val_score(model_poly, X_poly, y, cv=kf, scoring='r2')
print(f'Polynomial Regression Cross-validation scores: {cv_scores_poly}')
print(f'Average Polynomial CV score: {np.mean(cv_scores_poly):.4f}, Standard deviation: {np.std(cv_scores_poly):.4f}')

# Fit the polynomial model on the entire dataset to derive the final coefficients
model_poly.fit(X_poly, y)

# Extract coefficients and intercept for polynomial regression
poly_coefficients = model_poly.coef_
poly_intercept = model_poly.intercept_

print(f'Coefficients: {poly_coefficients}, Intercept: {poly_intercept}')
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the cleaned CKD dataset
ckd_data = pd.read_csv('ckd_feature_subset.csv')  # Ensure the CSV file path is correct
print(ckd_data.head())

# Features (X) and Labels (y)
X_ckd = ckd_data.drop('Target_ckd', axis=1)  # Ensure this matches your target column name
y_ckd = ckd_data['Target_ckd']  # This is the target variable

# Split the dataset into training/testing (90% train, 10% test)
X_train_ckd, X_test_ckd, y_train_ckd, y_test_ckd = train_test_split(
    X_ckd, y_ckd, test_size=0.1, random_state=42
)

# Set up for 5-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize model scores dictionary
model_scores = {
    'Model': [],
    'Average Accuracy': [],
    'Standard Deviation': []
}

# 1. Logistic Regression
log_reg = LogisticRegression(max_iter=1000)  # Increased iterations for convergence if needed
cv_scores_log_reg = cross_val_score(log_reg, X_ckd, y_ckd, cv=kf, scoring='accuracy')
model_scores['Model'].append('Logistic Regression')
model_scores['Average Accuracy'].append(np.mean(cv_scores_log_reg))
model_scores['Standard Deviation'].append(np.std(cv_scores_log_reg))
print(f'Logistic Regression - CV scores: {cv_scores_log_reg}')

# 2. Support Vector Machines
svc_model = SVC()
cv_scores_svc = cross_val_score(svc_model, X_ckd, y_ckd, cv=kf, scoring='accuracy')
model_scores['Model'].append('Support Vector Machines')
model_scores['Average Accuracy'].append(np.mean(cv_scores_svc))
model_scores['Standard Deviation'].append(np.std(cv_scores_svc))
print(f'Support Vector Machines - CV scores: {cv_scores_svc}')

# 3. k-Nearest Neighbors
knn_model = KNeighborsClassifier()
cv_scores_knn = cross_val_score(knn_model, X_ckd, y_ckd, cv=kf, scoring='accuracy')
model_scores['Model'].append('k-Nearest Neighbors')
model_scores['Average Accuracy'].append(np.mean(cv_scores_knn))
model_scores['Standard Deviation'].append(np.std(cv_scores_knn))
print(f'k-Nearest Neighbors - CV scores: {cv_scores_knn}')

# Create a DataFrame for easier visualization of scores
scores_df = pd.DataFrame(model_scores)

# Display the scores
print("\nModel Performance Summary:")
print(scores_df)
```

Conclusion:
Overall the model with the best accuracy and the lowest standard deviation seems to be the Support Vector Machines model. However, the K-Nearest Neighbors had similar accuarcy and standard deviation. The Logistic Regression model seems to be clearly the worst with a signifigantly lower accuracy, and the highest standard deviation, indicating that logistic regression is the least accurate model, and also the least generalized model.

```python
# 4. Neural Networks
# with 3 different configurations

# Initialize model scores dictionary
model_scores2 = {
    'Model': [],
    'Average Accuracy': [],
    'Standard Deviation': []
}

configs = [
    {'hidden_layer_sizes': (10,), 'activation': 'relu'},
    {'hidden_layer_sizes': (10,), 'activation': 'tanh'},
    {'hidden_layer_sizes': (20,), 'activation': 'relu'},
    {'hidden_layer_sizes': (20,), 'activation': 'tanh'},
    {'hidden_layer_sizes': (10, 10), 'activation': 'relu'},
    {'hidden_layer_sizes': (10, 10), 'activation': 'tanh'},
    {'hidden_layer_sizes': (20, 20), 'activation': 'relu'},
    {'hidden_layer_sizes': (20, 20), 'activation': 'tanh'}
]
   
for config in configs:
    nn_model = MLPClassifier(hidden_layer_sizes=config['hidden_layer_sizes'], activation=config['activation'], max_iter=3000)
    cv_scores_nn = cross_val_score(nn_model, X_ckd, y_ckd, cv=kf, scoring='accuracy')
    
    model_scores2['Model'].append(f'Neural Network {config["hidden_layer_sizes"]} {config["activation"]}')
    model_scores2['Average Accuracy'].append(np.mean(cv_scores_nn))
    model_scores2['Standard Deviation'].append(np.std(cv_scores_nn))
    print(f'Neural Network Configuration {config}: CV scores: {cv_scores_nn}')

# Create a DataFrame for easier visualization of scores
scores_df2 = pd.DataFrame(model_scores2)

# Display the scores
print("\nModel Performance Summary:")
print(scores_df2)
```
