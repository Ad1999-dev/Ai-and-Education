## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
categorised_data = pd.get_dummies(sort_data, columns=categorical_columns, drop_first=False, dtype=int)
# Print the dataset
print(categorised_data)
````

## Answer

For binary categorical variables (0/1):

- **Single Column Encoding**: Efficient and straightforward, but not necessary for binary variables.
- **One-hot Encoding**: Also equivalent but adds complexity without much benefit for just two categories.

For **binary categorical variables**, using a single column (0/1) is typically the most efficient approach.

## Categorical Variables with More Than Two Values

### Example: Sandwich Condiments

Let's consider 5 different condiments: Mustard, Mayo, Ketchup, Hot Sauce, and Relish.

### Single Column Encoding

In the single column encoding method, we assign integer values to each condiment:

| Condiment  | Encoding |
|------------|----------|
| Mustard    | 1        |
| Mayo       | 2        |
| Ketchup    | 3        |
| Hot Sauce  | 4        |
| Relish     | 5        |

If we have a dataset with various sandwiches and their corresponding condiment choices, it would look like this in a **single column** encoding:

| Sandwich ID | Condiment Encoding |
|-------------|--------------------|
| 1           | 1                  |  (Mustard)       |
| 2           | 3                  |  (Ketchup)       |
| 3           | 2                  |  (Mayo)          |
| 4           | 5                  |  (Relish)        |
| 5           | 4                  |  (Hot Sauce)     |

This approach creates an **artificial ordinal relationship** between the condiments, which can mislead the model. For instance, the model may interpret Mustard (1) as being closer to Mayo (2) than Hot Sauce (4).

### One-hot Encoding

In the one-hot encoding method, each condiment is represented by a separate binary column:

| Mustard | Mayo | Ketchup | Hot Sauce | Relish |
|---------|------|---------|-----------|--------|
| 1       | 0    | 0       | 0         | 0      |  (Mustard)      |
| 0       | 1    | 0       | 0         | 0      |  (Mayo)         |
| 0       | 0    | 1       | 0         | 0      |  (Ketchup)      |
| 0       | 0    | 0       | 1         | 0      |  (Hot Sauce)    |
| 0       | 0    | 0       | 0         | 1      |  (Relish)       |

### Comparison of Encoding Methods

- **Single Column Encoding**: Imposes an artificial ordinal relationship, misleading the model into thinking that some condiments are numerically related.
- **One-hot Encoding**: Properly represents each condiment as independent, allowing the model to learn the true impact of each condiment type on sandwich ratings without imposing a false ordinal relationship.

### Conclusion

For the sandwich condiment example:

- **One-hot Encoding is Necessary**: To avoid misleading the model into making incorrect assumptions about the relationships between different condiment types, which could reduce prediction accuracy.


## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
no_outliers = categorised_data.copy()
for i in numerical_columns:
    mean = no_outliers[i].mean()
    std = no_outliers[i].std() 
    
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std
    
    no_outliers = no_outliers[(no_outliers[i] >= lower_bound) & 
                                       (no_outliers[i] <= upper_bound)]

# Print the dataset    
print(no_outliers)
print(no_outliers.shape)    
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
normalized_data = no_outliers.copy()
scaler = MinMaxScaler()
normalized_data[numerical_columns] = scaler.fit_transform(normalized_data[numerical_columns])


# Print the dataset
print(normalized_data)
````

### 2. Redundant Binary Columns
Several binary columns were removed because they represent the negative case of binary variables that are already captured by their positive counterparts. The following columns were removed:

- **`rbc_normal`**: Redundant with `rbc_abnormal`
- **`pc_normal`**: Redundant with `pc_abnormal`

### Rationale for Removal
These redundant binary columns were created during one-hot encoding. Since each pair is perfectly negatively correlated (when one is 1, the other is always 0), retaining both columns introduces multicollinearity and does not add any new information. Removing one column from each binary pair helps maintain all the necessary information while reducing dimensionality. This reduces the risk of overfitting and improves model efficiency.


````python
#Remove that column
redundant_columns = ['unique_id', 'rbc_normal', 'pc_normal', 'pcc_notpresent', 
                    'ba_notpresent', 'htn_no', 'dm_no', 'cad_no', 
                    'appet_good', 'pe_no', 'ane_no', 'Target_notckd']

final_data = normalized_data.drop(columns=redundant_columns)


# final_data = normalized_data.to_csv('drop.csv', index=False)
# Print the dataset
print(final_data)
print(final_data.shape)
````
