## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicatedRowNumerical = numericalData.duplicated().sum()
print("Number of Duplicate rows in numerical data:", duplicatedRowNumerical)

# Drop these duplicate rows
numericalData = numericalData.drop_duplicates()

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicatedRowCategorical = categoricalData.duplicated().sum()
print("Number of Duplicate rows in categorical data:", duplicatedRowCategorical)
display(categoricalData[categoricalData.duplicated(keep=False)].groupby(list(categoricalData.columns)).size())
categoricalData = categoricalData.drop_duplicates()
````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

**Answer:** All the unique IDs that are present in the combined dataset need to exist in both datasets before combining. Because I chose the joining method to be 'inner', it specifies that any row that will be present needs to also be present in both of the original datasets.

````python
# Merge the two given numerical and categorical datasets based on their unique_ID.
data = pd.merge(numericalData, categoricalData, on='unique_id', how='inner')
#Print the combined dataset
display(data)
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value
percentNaN = data.isna().any(axis=1).sum() / data.shape[0] * 100
# Print %
print(str(percentNaN) + "% of row contains NaN")
# Drop these rows from the dataset
data = data.dropna().reset_index(drop=True)
# Print the Dataset
display(data)
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
data = data.sort_values(by='Target').reset_index(drop=True)
# Print the dataset
display(data)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

data = pd.get_dummies(data, columns=categorical_columns, prefix=categorical_columns, dtype=int)
# Print the dataset
display(data)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

**Answer:** I believe that if you have a binary feature, performing one-hot encoding (like what we did above) or label encoding (having one column and representing them with zero or one) will not make a difference to the model, as they are both interpreted as either yes or no. However, if that feature has more than 2 values (let's say five, for example), doing label encoding (ex. one column with zero 0-4) might indirectly tell the model that this feature has an ordered relationship with each other (one type of condiment is "more" than the other in some way), which is not accurate because they could be distinct values such that there are no orders among them at all.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

colMean = data[numerical_columns].mean(axis=0)
colSTD = data[numerical_columns].std(axis=0)

filter = (data[numerical_columns] <= colMean + 3 * colSTD) & (data[numerical_columns] >= colMean - 3 * colSTD)

filter = filter.all(axis=1)
data = data[filter].reset_index(drop=True)

display(data)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
for col in numerical_columns:
    min = data[col].min()
    max = data[col].max()
    data[col] = (data[col] - min) / (max - min)

# Print the dataset
display(data)
````

## Part 3.8: Remove Unnecessary columns

Are there any columns in this dataset which are not appropriate for modeling and predictions? Which column(s)? Justify their exclusion and remove them

**Answer:** I think one feature that should be removed from the dataset is the `unique_id` because it does not provide any relevant information regarding CKD diagnosis of the patient but rather just a unique identifier for each record. Furthermore, including `unique_id` can contribute to overfitting if the data is not properly shuffled before training. For instance, if the dataset is organized such that all records of patients with CKD appear consecutively at the beginning, the model may learn patterns solely based on the order of the records rather than on features that actually contribute to the CKD development.

````python
#Remove that column
data = data.drop(["unique_id"], axis=1)
# Print the dataset
display(data)
````

````python
# Code to rename all the columns in the dataset

column_mapping = {
    'age': 'age',
    'bp': 'blood_pressure',
    'sg': 'specific_gravity',
    'al': 'albumin',
    'su': 'sugar',
    'rbc': 'red_blood_cells',
    'pc': 'pus_cell',
    'pcc': 'pus_cell_clumps',
    'ba': 'bacteria',
    'bgr': 'blood_glucose_random',
    'bu': 'blood_urea',
    'sc': 'serum_creatinine',
    'sod': 'sodium',
    'pot': 'potassium',
    'hemo': 'hemoglobin',
    'pcv': 'packed_cell_volume',
    'wc': 'white_blood_cell_count',
    'rc': 'red_blood_cell_count',
    'htn': 'hypertension',
    'dm': 'diabetes_mellitus',
    'cad': 'coronary_artery_disease',
    'appet': 'appetite',
    'pe': 'pedal_edema',
    'ane': 'anemia'
}

def rename_columns(columns, mapping):
    new_columns = []
    for col in columns:
        base_name = col
        for abbr in mapping.keys():
            if col.startswith(abbr):
                base_name = mapping[abbr] + '_' + col[len(abbr):] if len(col) > len(abbr) else mapping[abbr]
                break
        new_columns.append(base_name)
    return new_columns

# Rename columns in the DataFrame
data.columns = rename_columns(data.columns, column_mapping)
data
````
