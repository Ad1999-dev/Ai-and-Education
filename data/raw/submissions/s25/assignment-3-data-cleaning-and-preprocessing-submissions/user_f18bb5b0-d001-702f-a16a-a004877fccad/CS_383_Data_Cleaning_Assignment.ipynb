## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicate_num = numerical_df.duplicated(subset=['unique_id'], keep='first').sum()
print(f"Number of duplicate rows in numerical dataset: {duplicate_num}")

# Drop these duplicate rows
if duplicate_num > 0:
    numerical_df = numerical_df.drop_duplicates(subset=['unique_id'], keep='first')
    print(f"After removing duplicates, numerical dataset has {len(numerical_df)} rows")
else:
    print("No duplicates found in numerical dataset")

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicate_cat = categorical_df.duplicated(subset=['unique_id'], keep='first').sum()
print(f"Number of duplicate rows in categorical dataset: {duplicate_cat}")

if duplicate_cat > 0:
    categorical_df = categorical_df.drop_duplicates(subset=['unique_id'], keep='first')
    print(f"After removing duplicates, categorical dataset has {len(categorical_df)} rows")
else:
    print("No duplicates found in categorical dataset")
````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why <redacted> you think so? If not, what <redacted> the rows that are missing from one of the datasets look like in the combined table?

Answer:

Not all unique IDs are present in both datasets. This is because when examining the data csv files, the categorical dataset has 400 rows while the numerical dataset has 390 row (after removing duplicate rows). This means the categorical dataset contains 10 IDs that don't appear in the numerical dataset.

An outer join preserves all rows from both datasets, filling missing values with NaN when a match isn't found. So, in the combined table, rows missing from numerical dataset would contain categorical data but has NaN values for numerical columns.

````python
# Merge the two given numerical and categorical datasets based on their unique_ID.
combined_df = pd.merge(numerical_df, categorical_df, on='unique_id', how='outer')

# Print the combined dataset
print("\nCombined dataset (showing all rows):")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(combined_df)
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain at least one missing value
rows_with_na = combined_df.isna().any(axis=1).sum()
total_rows = len(combined_df)
na_percentage = (rows_with_na / total_rows) * 100

# Print %
print(f"Percentage of rows with at least one missing value: {na_percentage:.2f}%")

# Drop these rows from the dataset
cleaned_df = combined_df.dropna()
print(f"Original dataset shape: {combined_df.shape}")
print(f"Dataset shape after removing rows with missing values: {cleaned_df.shape}")

# Print the entire cleaned dataset
print("\nCleaned dataset:")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(cleaned_df)
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column
sorted_df = cleaned_df.sort_values(by='Target')

# Reset the indices after sorting
sorted_df = sorted_df.reset_index(drop=True)

# Print the dataset
print("Sorted dataset by Target value:")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(sorted_df)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to <redacted> this **

````python
# Create dummy variables for all categorical columns
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
encoded_df = pd.get_dummies(sorted_df, columns=categorical_columns)

# Print the entire dataset
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(encoded_df)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: 

Using a single column for binary variables works the same as dummy variables in modeling. However, for categorical variables with more than two values (like five condiment types), a single column would wrongly suggest that there is some sort of order between the values (such as suggesting that condiment type 5 is "more" than type 1, which is wrong). Using dummy variables for multi-class categorical features is better as it creates separate binary columns for each category. The downside is that we'll have more columns in our dataset.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

df_no_outliers = encoded_df.copy()

for col in numerical_columns:
    mean = df_no_outliers[col].mean()
    std = df_no_outliers[col].std()
    
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std
    
    mask = (df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)
    
    df_no_outliers = df_no_outliers[mask]

print(f"Original dataset shape: {encoded_df.shape}")
print(f"Dataset shape after removing outliers: {df_no_outliers.shape}")
print(f"Number of rows removed: {encoded_df.shape[0] - df_no_outliers.shape[0]}")

# Print the cleaned dataset without outliers
print("\nCleaned dataset (without outliers):")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df_no_outliers)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize all Numerical Attributes in the dataset.
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

normalized_df = df_no_outliers.copy()

for column in numerical_columns:
    min_val = df_no_outliers[column].min()
    max_val = df_no_outliers[column].max()
    normalized_df[column] = (df_no_outliers[column] - min_val) / (max_val - min_val)

# Print the normalized dataset
print("\nNormalized dataset:")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(normalized_df)
````

## Part 3.8: Remove Unnecessary columns

Are there any columns in this dataset which are not appropriate for modeling and predictions? Which column(s)? Justify their exclusion and remove them

Answer:

Since unique_id is just a record identifier and has no relationship with the target variable or other features, it should be removed from the dataset to prevent it from interfering with the model's predictions.

````python
# Remove unique_id column
normalized_df = normalized_df.drop('unique_id', axis=1)

# Print the dataset without the unique_id column
print("\nDataset without unique_id column:")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(normalized_df)
````
