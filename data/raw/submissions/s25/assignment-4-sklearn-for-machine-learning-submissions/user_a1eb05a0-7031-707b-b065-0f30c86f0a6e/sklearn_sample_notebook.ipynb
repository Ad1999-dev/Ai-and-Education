## 1. Loading the dataset

```python
# Using pandas load the dataset
data_set = pd.read_csv(r'C:\Users\deshn\Downloads\assignment-4-sklearn-for-machine-learning-<redacted>\assignment-4-sklearn-for-machine-learning-<redacted>\science_data_large.csv')
# Output the first 15 rows of the data
print(data_set.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(data_set.describe())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data_set[['Temperature °C','Mols KCL']]
Y = data_set['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train,X_test,Y_train,Y_test = sklearn.model_selection.train_test_split(X,Y,test_size=0.1,random_state=42)
# For grading consistency use random_state=42 
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
regression =LinearRegression().fit(X_train,Y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample_point = [[405, 723]]
prediction=regression.predict(sample_point)
print('Predicted output for sample point: ', prediction)

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score= regression.score(X_test,Y_test)
print('Score for the model: ', score)
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coff= regression.coef_
inter= regression.intercept_
print('Coefficients: ', coff)
print('Intercept: ', inter)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

k_fold= KFold(n_splits=5, random_state=42, shuffle=True)
validation_score = cross_val_score(regression,X,Y,cv=k_fold)
print('Scores',validation_score)

# Report on their finding and their significance
mean_score = validation_score.mean()
std_score = validation_score.std()
print('Mean score:',mean_score)
print('Standard deviation:',std_score)

n_trials = 10  # Repeat the cross-validation 10 times
all_scores = []

for trial in range(n_trials):
    # Use cross_val_score to evaluate the model
    scores = cross_val_score(regression, X, Y, cv=k_fold)
    all_scores.append(scores)

# Convert scores to a NumPy array for easier analysis
all_scores = np.array(all_scores)

# Print scores for each trial
for i, scores in enumerate(all_scores):
    print(f'Trial {i + 1} Scores: {scores}')

# Calculate mean and standard deviation across trials
mean_scores = all_scores.mean(axis=0)
std_scores = all_scores.std(axis=0)

# Displaying results in a table format
results_df = pd.DataFrame({
    'Fold': np.arange(1, 5 + 1),
    'Mean Score': mean_scores,
    'Standard Deviation': std_scores
})

print("\nCross-validation Results:")
print(results_df)
```

Result and findings conclusion about variability and mean and score
the mean score of 0.86 from cross validation means that acrosss all the different sets the model is able to explain 86% of the variance in the target variable. It is generally considered as good performance.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

CKD_dataset = pd.read_csv(r'C:\Users\deshn\Downloads\assignment-4-sklearn-for-machine-learning-<redacted>\assignment-4-sklearn-for-machine-learning-<redacted>\ckd_feature_subset.csv')
print(CKD_dataset.head(15))
# Split the dataset into training and testing sets
XC= CKD_dataset.drop('Target_ckd', axis=1)
YC = CKD_dataset['Target_ckd']
X_trainC, X_testC, Y_trainC, Y_testC = sklearn.model_selection.train_test_split(XC,YC, test_size=0.1, random_state=42)
```

# Logistic Regression

```python
log_reg= LogisticRegression(max_iter= 200)
log_reg.fit(X_trainC, Y_trainC)
sample_log= X_testC.iloc[0].values.reshape(1,-1)
sample_log_pred= log_reg.predict_proba(sample_log)
print('Probability of the sample is: ', sample_log_pred)
Score_log= log_reg.score(X_testC, Y_testC)
print('Score of the model is: ', Score_log)
coef_log= log_reg.coef_
print('Coefficients of the model are: ', coef_log)
intercept_log= log_reg.intercept_
print('Intercept of the model is: ', intercept_log)
```

# Support Vector Machines

```python
mod_svc= SVC(probability=True)
mod_svc.fit(X_trainC, Y_trainC)
sample_svc= X_testC.iloc[0].values.reshape(1,-1)
sample_svc_pred= mod_svc.predict_proba(sample_svc)
print('Probability of sample is: ', sample_svc_pred)
score_svc= mod_svc.score(X_testC, Y_testC)
print('Score of SVC is: ', score_svc)
```

# K-Nearest Neighbors

```python
K_mod= KNeighborsClassifier(n_neighbors=5)
K_mod.fit(X_trainC, Y_trainC)
sample_k= X_testC.iloc[0].values.reshape(1,-1)
prediction_k = K_mod.predict_proba(sample_k)
print('Prediction for sample: ', prediction_k)
score_k= K_mod.score(X_testC, Y_testC)
print('Score for sample: ', score_k)
```

```python
neural_mod= MLPClassifier(hidden_layer_sizes= (10,10), activation='relu', solver='adam', max_iter=1000, random_state=42)
neural_mod.fit(X_trainC, Y_trainC)
sample_neu= X_testC.iloc[0].values.reshape(1,-1)
pro =neural_mod.predict_proba(sample_neu)
score_neu= neural_mod.score(X_testC, Y_testC)
print('Score',score_neu)
print('Probability',pro)
```

```python
k_fold= KFold(n_splits=5, random_state=42, shuffle=True)
classifications = {'Logistic Regression':log_reg, 'Support Vector Machines':mod_svc,'K-Nearest Neighbors': K_mod, 'Neural Network':neural_mod}
table={}
for name, classification in classifications.items():
    validation_score = cross_val_score(classification,XC,YC,cv=k_fold)
    print('Scores',validation_score)
    mean_score = validation_score.mean()
    std_score = validation_score.std()
    table[name]={'Average':mean_score,'Standard Deviation':std_score}

the_table = pd.DataFrame(table).T
the_table.columns = ['Average','Standard Deviation']
print()
print(the_table)
```
