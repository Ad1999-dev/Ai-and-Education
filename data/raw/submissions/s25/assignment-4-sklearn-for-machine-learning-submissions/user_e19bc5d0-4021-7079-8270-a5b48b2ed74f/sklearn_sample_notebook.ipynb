```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
slime_csv = pd.read_csv("science_data_large.csv")
print(slime_csv.head(15))
print(slime_csv.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = slime_csv.drop(columns=["Size nm^3"])
Y = slime_csv["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
print(X_train, X_test, Y_train, Y_test)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
reg = LinearRegression().fit(X_train, Y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame([[500, 700]], columns=X_train.columns)
print(reg.predict(sample))
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print(reg.score(X_test, Y_test))
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print(reg.coef_)
print(reg.intercept_)
```

Report on score and explain meaning: 0.85525. The R^2 score indicates how well the concentration of KCl and temperature in the model explain the variance in the slime's growth. ## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kfolds = KFold(n_splits=5, shuffle=True, random_state=42)
print(cross_val_score(reg, X, Y, cv=kfolds))
# Report on their finding and their significance
```

Write findings here: In k-fold cross validation with 5 splits, the model is trained on 4 of the folds and tested on the other one, repeated 5 times with each fold being the test fold one time. The scores of 0.86152, 0.82742, 0.87195, 0.88166, and 0.85609 suggest that the model is stable since all of scores are relatively close and that the model generalizes well since all of the scores are relatively high even when training and testing with different sets.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
model = Pipeline([('poly', PolynomialFeatures(degree=2)), ('linear', LinearRegression())])
model.fit(X,Y)
print(cross_val_score(model, X, Y, cv=KFold(n_splits=5, shuffle=True, random_state=42)))
# Report on the metrics and output the resultant equation as you did in Part 3.
print(model.score(X_test, Y_test))
print(model.named_steps['linear'].coef_)
print(model.named_steps['linear'].intercept_)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_dataset = pd.read_csv("ckd_feature_subset.csv")
ckd_X = ckd_dataset.drop(columns=["Target_ckd"])
ckd_Y = ckd_dataset["Target_ckd"]

ckd_kfolds = KFold(n_splits=5, shuffle=True, random_state=42)
models = [LogisticRegression(random_state=42), SVC(random_state=42), KNeighborsClassifier(), MLPClassifier(random_state=42, max_iter=1000)]
for model in models:
    cross_val_scores = cross_val_score(model, ckd_X, ckd_Y, cv=ckd_kfolds)
    print(np.mean(cross_val_scores))
    print(np.std(cross_val_scores))
```

## Results and Conclusion for Classification Experiments

| Model               |      Average     | Standard Deviation |
|---------------------|------------------|---------------------|
| Logistic Regression | 0.85656          | 0.06627             |
| SVM                 | 0.92817          | 0.04760             |
| k-Nearest Neighbors | 0.92796          | 0.05244             |
| Neural Network      | 0.93505          | 0.04047             |

Based on these results, a Neural Network performed the best as it has the highest average with the lowest standard deviation followed by SVM and k-Nearest Neighbors close behind. The logistic regression performed by far the worst. The neural network probably performed the best because predicting CKD most likely has very complex and nonlinear relationships between all the variables. Neural networks can do a better job capturing complex patterns.

```python
# Experiments with Neural Network.
neural_net_configs = [
    MLPClassifier(random_state=42, hidden_layer_sizes=(100, 50, 30), max_iter=1000), 
    MLPClassifier(random_state=42, solver="lbfgs"), 
    MLPClassifier(random_state=42, activation="logistic", max_iter=1200),
    MLPClassifier(random_state=42, hidden_layer_sizes=(100, 50, 30), solver="lbfgs", activation="relu")
]
for config in neural_net_configs:
    cross_val_scores = cross_val_score(config, ckd_X, ckd_Y, cv=ckd_kfolds)
    print(np.mean(cross_val_scores))
    print(np.std(cross_val_scores))
```

## Results and Conclusion for Neural Network Experiments

| Model | Average | Standard Deviation |
|-------|------------------|---------------------|
| Hidden Layers (100,50,30)     | 0.96753          | 0.03534             |
|   lbfgs Solver   | 0.97398          | 0.02420             |
| Logistic Activation     | 0.91527          | 0.04824             |
| Hidden Layers (100,50,30) &  lbfgs Solver & relu activation   | 0.93505          | 0.04047             |

The lbfgs solver alone had the best performance with the highest average and lowest standard deviation. Following that, just hidden layers with size (100, 50, 30) also performed very well being the second best. This is followed by the combination of Hidden Layers (100,50,30) &  lbfgs Solver & relu activation and then logistic activation performed the worst. The choice of solver and hidden layer sizes made the biggest differences on the performance. The activation had little impact other than the average was much lower for logistic. Interestingly, combining all of the best performing options I tried for solver, hidden layers, and activation made the model perform worse.
