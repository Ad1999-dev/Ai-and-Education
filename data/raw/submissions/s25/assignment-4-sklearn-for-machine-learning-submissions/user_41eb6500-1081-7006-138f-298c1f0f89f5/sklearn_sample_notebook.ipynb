```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
df = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df[['Temperature °C', 'Mols KCL']] 
Y = df['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, Y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
data_test = pd.DataFrame([['500','500']])
test_prediction = model.predict(data_test)
print(test_prediction)
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
model_score = model.score(X_test, Y_test)
print(model_score)
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coef = model.coef_
intercept = model.intercept_
print('Coefficients -->', coef)
print('Intercept -->', intercept)
```

Report on score and explain meaning:
score of 0.860 means that 86% of the variation in Size nm^3 is explained by Temperature °C and Mols KCL, meaning the model captures most of the important patterns in the data. Since this is a high score, it indicates the model makes accurate predictions and should perform well on new data.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X, Y, cv=cv, scoring='r2')
print('Cross-validation scores:', cv_scores)
# Report on their finding and their significance
```

Write findings here: The cross-validation scores show that the model makes good predictions, as they range from 82.7% to 88.2%. Since the scores are similar across different data splits, it means the model is reliable and consistent, and will work well on new data

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
data = pd.read_csv('ckd_feature_subset.csv')

X_2 = data.drop(columns=["Target_ckd"])  # Features
Y_2 = data["Target_ckd"]  # Target variable

# Standardize features
scaler = StandardScaler()
X_2_scaled = scaler.fit_transform(X_2)

# Define cross-validation strategy
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Define classification models
class_models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "SVM (SVC)": SVC(kernel="rbf", random_state=42),
    "k-NN": KNeighborsClassifier(n_neighbors=5),
}

results = {}

for model_name, model in class_models.items():
    scores = cross_val_score(model, X_2_scaled, Y_2, cv=cv)
    results[model_name] = {"Mean Accuracy": np.mean(scores), "Std Dev": np.std(scores)}

print(results)
```

## Results and Conclusion for Classification Experiments
The SVM model performed the best, achieving the highest mean accuracy (96.7%) with a low standard deviation (2.9%), indicating both strong performance and stability. Logistic Regression followed closely with 94.8% accuracy, while k-NN had the lowest accuracy (92.2%) and the highest variability (6.0%), suggesting it may be more sensitive to data distribution. Overall, SVM is the best choice for this dataset due to its ability to capture complex patterns while maintaining consistent performance across different data splits.

```python
neural_network_model = MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, random_state=42)

nn_scores = cross_val_score(neural_network_model, X_2_scaled, Y_2, cv=cv, scoring="accuracy")

nn_results = {
    "Mean Accuracy": np.mean(nn_scores),
    "Std Dev": np.std(nn_scores)
}

print(nn_results)
```

## Results and Conclusion for Neural Network Experiments
The Neural Network achieved a mean accuracy of 96.1%, showing strong predictive performance. The standard deviation of 3.2% indicates consistent results across different data splits, meaning the model generalizes well. Compared to the Classification models, the performance of the Neural Network is only slightly worse than the SVC (best Classification).
