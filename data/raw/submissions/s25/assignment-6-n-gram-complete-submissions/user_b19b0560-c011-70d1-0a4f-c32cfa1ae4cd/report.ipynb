#### `create_frequency_tables(document, n)`

### Code analysis

The `create_frequency_tables` function builds n frequency tables, where each table corresponds to an n-gram size from 1 to n. The function iterates through the provided document character by character to extract all possible n-grams of sizes from 1 to n. It counts the occurrences of each n-gram and populates the respective frequency tables. When the end index exceeds the document's length, it skips that n-gram extraction to avoid index errors.

## `calculate_probability(sequence, char, tables)`

### Formula
The formula for sequence likelihood as described in section 2 can be expressed as follows:

$$P(X_1=x_1, X_2=x_2, \dots, X_t=x_t) = P(x_1) \cdot \prod_{i=2}^{t} P(x_i | x_{max(1,i-n+1)}, \dots, x_{i-1})$$

### Code analysis

In the `calculate_probability` function, probability calculations leverage the frequency tables built earlier to estimate conditional probabilities for the next character based on the context of the preceding characters. The implementation includes a recursive backoff mechanism that allows the model to reduce its context size sequentially when a probability cannot be estimated due to the unavailability of certain n-grams.
### Your Calculations

Using the probability tables for the sequence "aa" and the next possible characters in the table:

1. $P(X_1=a, X_2=a, X_3=a)$
   - $P(a|aa) = \frac{f(aaa)}{f(aa)} = \frac{2}{5}$

2. $P(X_1=a, X_2=a, X_3=b)$
   - $P(b|aa) = \frac{f(aab)}{f(aa)} = \frac{2}{5}$

3. $P(X_1=a, X_2=a, X_3=c)$
   - $P(c|aa) = \frac{f(aac)}{f(aa)} = \frac{1}{5}$

## `predict_next_char(sequence, tables, vocabulary)`

### Code analysis

The `predict_next_char` function computes the predicted next character by evaluating the probabilities of all characters in the vocabulary based on the current input sequence. The character with the highest computed probability from the `calculate_probability` function is chosen as the most likely next character. This function efficiently handles the trimming of the sequence to match the n-gram size restrictions.

### So what should be the next character in the sequence?

Based on the probability distribution obtained above for all the next possible characters after "aa":
- P(a|aa) = 2/5
- P(b|aa) = 2/5
- P(c|aa) = 1/5

As both 'a' and 'b' share the highest probability (2/5), either character can be selected as the next one. In the implementation, the first maximum value encountered is returned, which would be 'a' if the vocabulary is processed in alphabetical order.
