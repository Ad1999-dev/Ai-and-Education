```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
raw_data = pd.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
print(raw_data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(raw_data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
raw_x = raw_data[['Temperature °C', 'Mols KCL']]
raw_y = raw_data[['Size nm^3']]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
train_x, test_x, train_y, test_y = train_test_split(raw_x, raw_y, test_size=0.1, random_state=42)
# For grading consistency use random_state=42 
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model_linear_regression = LinearRegression()
model_linear_regression.fit(train_x, train_y)

# Create a sample datapoint and predict the output of that sample with the trained model
sample = [[300, 900]]
print(model_linear_regression.predict(sample))

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
print("\nTrain Score:",model_linear_regression.score(train_x, train_y))
print("\nTest Score:",model_linear_regression.score(test_x, test_y))

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
intercept = model_linear_regression.intercept_[0] 
coefficients = model_linear_regression.coef_[0]

print("\nIntercept:", intercept)
print("\nCoefficients:", coefficients)
```

Report on score and explain meaning:  
The $R^2$ for the training dataset $R^2_{train}=0.861$ and the test dataset is $R^2_{test}=0.855$, this indicates that the model is able to explain 86.1% of the variance in the training dataset and 85.5% of the variance in the test dataset, a good result for a linear model. Furthermore, the difference in the $R^2$ scores between the test and training is 0.006, this indicates that the model is not overfitting the data.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
from sklearn.model_selection import cross_val_score, KFold

kfold = KFold(n_splits=5, random_state=42, shuffle=True)

scores_linear = cross_val_score(model_linear_regression, raw_x, raw_y, cv=kfold)

# Report on their finding and their significance
print("Cross Validation Scores:", scores_linear)
print("Mean CV Score:", scores_linear.mean())
print("Standard Deviation of CV Score:", scores_linear.std())
```

Write findings here:   
The cross validation scores provide a robust measure of our model’s performance across different subsets of the data. 
By using 5-fold cross validation with a fixed random state, we ensure reproducibility and a fair evaluation. For instance, 
the CV scores are [0.86151889 0.82742341 0.87195173 0.88166206 0.85609101], the mean CV score is approximately 0.86 with a 
low standard deviation. This indicates that our linear regression model consistently explains about 86% of the variance in 
the slime size across different data splits. Such a high and consistent score suggests that our model generalizes well to new, unseen data.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
raw_x_poly = poly.fit_transform(raw_x)

feature_names = poly.get_feature_names_out(['Temperature °C', 'Mols KCL'])
print("Polynomial feature names:", feature_names)

# Split the polynomial features into train and test sets
train_x_poly, test_x_poly, train_y, test_y = train_test_split(raw_x_poly, raw_y, test_size=0.1, random_state=42)

# Train a new Linear Regression model on the polynomial features
poly_model = LinearRegression()
poly_model.fit(train_x_poly, train_y)
# Report on the metrics and output the resultant equation as you did in Part 3.

# Compute cross validation scores on the polynomial features
cv_scores_poly = cross_val_score(poly_model, raw_x_poly, raw_y, cv=kfold)
intercept_poly = poly_model.intercept_[0] 
coefficients_poly = poly_model.coef_[0]

print("\nPolynomial Regression CV Scores:", cv_scores_poly)
print("\nMean CV Score:", cv_scores_poly.mean())
print("\nStandard Deviation of CV Score:", cv_scores_poly.std())
print("\nIntercept:", intercept_poly)
print("\nCoefficients:", coefficients_poly)


```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

ckd_data = pd.read_csv("ckd_feature_subset.csv")

print("Dataset columns:", ckd_data.columns)
target_column = 'Target_ckd'

# Separate the features (X) and the target (y)
X = ckd_data.drop(columns=[target_column])
y = ckd_data[target_column]

from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Set up a 5-fold cross validation configuration
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Define the classification models + NN
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Support Vector Machines": SVC(random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(random_state=42, max_iter=1000)
}

# Dictionary to store results
results = {}

# Evaluate each model using cross validation
for name, model in models.items():
    cv_scores = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
    results[name] = {"Mean Accuracy": cv_scores.mean(), "Std Accuracy": cv_scores.std()}

# Display the results in a table
results_df = pd.DataFrame(results).T
print("Classification Model Performance:")
print(results_df)
```

## Results and Conclusion for Classification Experiments

The following Classification models were ran and to gauge performance of the models cross validation was used:

- Logistic Regression ${[Mean_{accuracy}: 0.86, Std_{accuracy}: 0.066]}$
- Support Vector Machines ${[Mean_{accuracy}: 0.93, Std_{accuracy}: 0.048]}$
- k-Nearest Neighbors ${[Mean_{accuracy}: 0.93, Std_{accuracy}: 0.052]}$
- Neural Network ${[Mean_{accuracy}: 0.94, Std_{accuracy}: 0.040]}$

Based on these results the best performing classification was provided by neural networks (94%), followed by KNN(93%) and Support Vector Machines(93%); with Logistic regression(86%) performing the worst. All the classification models had a std of ≈ 0 indicating they perform well throughout all the data, however, the performance of the polynomial regression is better than these methods. 



```python
# Experiments with Neural Network.

# Define different neural network configurations
nn_configs = {
    "Config 1": MLPClassifier(hidden_layer_sizes=(50,), activation='relu', random_state=42, max_iter=1000),
    "Config 2": MLPClassifier(hidden_layer_sizes=(100,), activation='tanh', random_state=42, max_iter=1000),
    "Config 3": MLPClassifier(hidden_layer_sizes=(50, 50), activation='relu', random_state=42, max_iter=1000),
    "Config 4": MLPClassifier(hidden_layer_sizes=(50, 50, 50), activation='relu', random_state=42, max_iter=1000),
    "Config 5": MLPClassifier(hidden_layer_sizes=(100, 50, 25), activation='relu', random_state=42, max_iter=1000),
    "Config 6": MLPClassifier(hidden_layer_sizes=(256, 128, 64, 32, 16, 8), activation='relu', random_state=42, max_iter=1000),
    "Config 7": MLPClassifier(hidden_layer_sizes=(512, 256, 128, 64, 32, 16, 8), activation='relu', random_state=42, max_iter=1000),
    "Config 8": MLPClassifier(hidden_layer_sizes=(1000, 500, 100), activation='relu', random_state=42, max_iter=1000),
}

nn_results = {}

# Evaluate each neural network configuration using cross validation
for name, model in nn_configs.items():
    cv_scores = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
    nn_results[name] = {"Mean Accuracy": cv_scores.mean(), "Std Accuracy": cv_scores.std()}

# Convert the neural network results to a DataFrame and Print
nn_results_df = pd.DataFrame(nn_results).T
print("Neural Network Configurations Performance:")
print(nn_results_df)
```

## Results and Conclusion for Neural Network Experiments

8 configs were tested with varying no. of layers and nodes per layer, ReLu activation was used for all configs with max iters. capped at 1000 (more iters were tested but yielded no accuracy differences). Based on the results [output of above cell] performance increased with more no. of layers and increased no. of nodes from config 1 - config 5, however, after increasing no. of layers from 3 there were no improvements to accuracy performance. Therefore, the best config. is a NN with 3 hidden layers of size: (>100, >50, >25) yielding a performance of 97% accuracy with a std. of ≈ 0.03.
