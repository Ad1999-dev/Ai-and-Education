```python
# Imports section
import pandas as pd 
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
file_path = 'science_data_large.csv' 
data = pd.read_csv(file_path)

print(data.head(15))

print(data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X = data[['Mols KCL', 'Temperature °C']]  # Selecting features
y = data['Size nm^3']  # Assuming Temperature is the label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

model = LinearRegression()
model.fit(X_train, y_train)

sample_datapoint = pd.DataFrame({'Mols KCL': [600], 'Temperature °C': [400]})  # Use representative values
predicted_output = model.predict(sample_datapoint)

# Report the model's score
model_score = model.score(X_test, y_test)

# Extract coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

# Printing the results (if needed)
print(f"Predicted output: {predicted_output}")
print(f"Model score: {model_score}")
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Report on their finding and their significance
scores = cross_val_score(model, X, y, cv=5, scoring='r2', n_jobs=-1)

# Report the findings
print(f"Cross-validated R² scores: {scores}")
print(f"Mean R² score: {scores.mean()}")
print(f"Standard deviation of R² scores: {scores.std()}")

# Calculate the mean and standard deviation of the scores
mean_score = scores.mean()
std_dev_score = scores.std()
```

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.
X = data[['Mols KCL', 'Temperature °C']]
y = data['Size nm^3']

# Create polynomial features of degree 2
poly = PolynomialFeatures(degree=2, include_bias=False)  # do not include bias (constant term)
X_poly = poly.fit_transform(X)

# Initialize the polynomial regression model
model_poly = LinearRegression()

# Create a KFold object with shuffling
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform k-fold cross-validation with shuffling
scores_poly = cross_val_score(model_poly, X_poly, y, cv=kf, scoring='r2', n_jobs=-1)

# Report the findings
print(f"Cross-validated R² scores for polynomial regression: {scores_poly}")
print(f"Mean R² score for polynomial regression: {scores_poly.mean()}")
print(f"Standard deviation of R² scores for polynomial regression: {scores_poly.std()}")

# Fit the model on the entire dataset
model_poly.fit(X_poly, y)

# Extract coefficients and intercept
coefficients_poly = model_poly.coef_
intercept_poly = model_poly.intercept_

# Output resultant equation
# Generate feature names for the polynomial features
feature_names = poly.get_feature_names_out(input_features=['Mols KCL', 'Temperature °C'])
equation_terms = [f"{coeff} * {name}" for coeff, name in zip(coefficients_poly, feature_names)]
equation = " + ".join(equation_terms)

print(f"Resultant polynomial equation:\nSize nm^3 = {intercept_poly} + {equation}")
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Load the CKD dataset
file_path = 'ckd_feature_subset.csv'  # Ensure this matches the filename you have
data = pd.read_csv(file_path)

# Display the first few rows of the dataset and info for inspection
print(data.head())
print(data.info())

# Separating features and target variable
# Adjust 'Target_ckd' to the actual name in your dataset if necessary
X = data.drop('Target_ckd', axis=1)  # Replace 'Target_ckd' with the appropriate target column name
y = data['Target_ckd']

# Check for any missing values in the features
print(X.isnull().sum())  # Check for null values

# Define the classification models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Support Vector Machine': SVC(),
    'k-Nearest Neighbors': KNeighborsClassifier(),
}

# Initialize a list to store results
results = []

# Random seed for reproducibility
random_state = 42

# Create a KFold object with shuffling
kf = KFold(n_splits=5, shuffle=True, random_state=random_state)

# Train and evaluate the models
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')  # Use 'accuracy' as scoring
    results.append({
        'Model': name,
        'Mean Accuracy': scores.mean(),
        'Standard Deviation': scores.std()
    })

# Define neural network configurations to test
nn_configs = [
    ('NN_1', MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=random_state)),
    ('NN_2', MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=random_state)),
    ('NN_3', MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=1000, random_state=random_state))  # 2 hidden layers
]

# Evaluate each neural network configuration
for name, nn_model in nn_configs:
    nn_scores = cross_val_score(nn_model, X, y, cv=kf, scoring='accuracy')
    results.append({
        'Model': name,
        'Mean Accuracy': nn_scores.mean(),
        'Standard Deviation': nn_scores.std()
    })

# Compile results into a DataFrame
results_df = pd.DataFrame(results)

# Display the results
print(results_df)

# Optional: Exporting to a CSV for reference
results_df.to_csv('ckd_classification_results.csv', index=False)
```

## Results and Conclusion for Classification Experiments
The classification experiments indicated that the best performing model was the neural network configuration NN_3, achieving a mean accuracy of 0.9546 with a standard deviation of 0.0437. This model, featuring two hidden layers with 50 neurons each, effectively captured complex patterns in the CKD dataset, showcasing its robust predictive capability for classifying Chronic Kidney Disease. Other models, such as the Support Vector Machine (mean accuracy of 0.9282) and k-Nearest Neighbors (mean accuracy of 0.9280), also demonstrated strong performance, indicating their suitability for this task. In contrast, the Logistic Regression model (mean accuracy of 0.8566) performed significantly worse, highlighting its limitations in capturing the non-linear relationships present in the data.

## Results and Conclusion for Neural Network Experiments
The analysis of the neural network configurations revealed that the choice of architecture significantly impacted classification performance. Specifically, the number of hidden layers and neurons per layer were critical factors; NN_3 outperformed the simpler NN_1 (with a single hidden layer of 10 neurons) and NN_2 (with a single layer of 100 neurons). Additionally, the activation function and the number of training iterations  are essential parameters that can affect model training and convergence. These findings suggest that utilizing a deeper architecture and optimizing training configurations can substantially enhance the model’s ability to classify chronic kidney disease effectively.
