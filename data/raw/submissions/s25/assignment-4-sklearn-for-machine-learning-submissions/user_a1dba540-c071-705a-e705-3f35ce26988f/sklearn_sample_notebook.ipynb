```python
# Imports section
import pandas as pd
import sklearn as sk
import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data[['Temperature °C','Mols KCL']]
data.drop(columns=['Temperature °C','Mols KCL'])
y = data['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1, train_size=0.9, random_state=42)
# For grading consistency use random_state=42 
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = sk.linear_model.LinearRegression()
model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample_datap = pd.DataFrame({'Temperature °C':[469],'Mols KCL': [649]})
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
predicted_output = model.predict(sample_datap)
score = model.score(X_test, y_test)
print("Score of Model: ", score)
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coef = model.coef_
intercept = model.intercept_
print("Coefficient: " ,coef,"\nIntercept: ", intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
scores = sk.model_selection.cross_val_score(model, X, y, cv=5)
print("Cross Validation Scores: ", scores)
# Report on their finding and their significance
```

Write findings here:
The scores output by the cross validation function show a difference of roughly only +-2, which shows a very tight cluster even across different subsets of data. This low score range indicates that the model is reproducible and robust, further validating the score of 0.855 I recieved earlier.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
# Polynomial regression (degree=2) with k-fold cross-validation
poly = sk.preprocessing.PolynomialFeatures(degree=2)

# Fit only to training data, then transform training and test sets
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)  # <-- fix applied here

# Initialize k-fold cross-validation on polynomial training data
kfold = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
poly_model = sk.linear_model.LinearRegression()

# Cross-validation scores (use polynomial training data here!)
poly_cv_scores = sk.model_selection.cross_val_score(poly_model, X_train_poly, y_train, cv=kfold)
print("Polynomial CV Scores:", poly_cv_scores)

# Train polynomial model fully on training set
poly_model.fit(X_train_poly, y_train)

# Predict using polynomial model on test set to check metrics realistically
y_test_pred = poly_model.predict(X_test_poly)

poly_features = poly.get_feature_names_out(['Temperature °C', 'Mols KCL'])
poly_coef = poly_model.coef_
poly_intercept = poly_model.intercept_
print("Polynomial Coefficients: ", poly_coef)
print("Polynomial Intercept: ", poly_intercept)
print("Polynomial Features: ", poly_features)

# Report on the metrics and output the resultant equation as you did in Part 3.
```
