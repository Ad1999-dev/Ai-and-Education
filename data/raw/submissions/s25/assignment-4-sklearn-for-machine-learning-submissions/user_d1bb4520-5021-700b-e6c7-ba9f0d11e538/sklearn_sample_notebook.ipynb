```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

data = pd.read_csv('science_data_large.csv')
print(data.head(15))
print(data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42

X = data[['Temperature °C', 'Mols KCL']]  # Independent variables
y = data['Size nm^3']  # Dependent variable

# Split into training (90%) and testing (10%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Output dataset shapes
print(f"Training Features: {X_train.shape}, Training Labels: {y_train.shape}")
print(f"Testing Features: {X_test.shape}, Testing Labels: {y_test.shape}")
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

linear = LinearRegression()
linear.fit(X_train, y_train)
sample = [[500.5,459.5]]
prediction = linear.predict(sample)


print(f"Predicted Size (nm³) for sample input: {prediction[0]:.5f}")

model_score = linear.score(X_test, y_test)
print(f"Model Score (R²): {model_score:.4f}")


coefficients = linear.coef_
intercept = linear.intercept_

print(f"Intercept: {intercept}")
print(f"Coefficients: {coefficients}")
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Report on their finding and their significance
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(linear, X, y, cv=kf, scoring='r2')

# Print results
print(f"Cross-Validation R² Scores: {cv_scores}")
print(f"Mean R² Score: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation of R²: {np.std(cv_scores):.4f}")
```

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Perform regression
model = LinearRegression()
model.fit(X_poly, y)


k = 5  
cv_scores = cross_val_score(model, X_poly, y, cv=k, scoring='neg_mean_squared_error')
mean_cv_score = -np.mean(cv_scores)


y_pred = model.predict(X_poly)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

coefficients = model.coef_
intercept = model.intercept_
equation_terms = [f"{coeff:.3f}*{term}" for coeff, term in zip(coefficients, poly.get_feature_names_out())]
equation = f"{intercept:.3f} + " + " + ".join(equation_terms[1:])

print(f"Mean CV MSE: {mean_cv_score:.3f}")
print(f"Training MSE: {mse:.3f}")
print(f"R^2 Score: {r2:.3f}")
print(f"Resultant Equation: {equation}")
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
data = pd.read_csv('ckd_feature_subset.csv')


X = data.drop('Target_ckd', axis=1)
y = data['Target_ckd']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Support Vector Machine': SVC(random_state=42),
    'k-Nearest Neighbors': KNeighborsClassifier(),
    'Neural Network': MLPClassifier(random_state=42, max_iter=1000)
}


results = {}
for name, model in models.items():

    cv_scores = cross_val_score(model, X_scaled, y, cv=5)
    results[name] = {
        'mean_accuracy': cv_scores.mean(),
        'std_accuracy': cv_scores.std()
    }

print("Classification Model Performance:")
print("Model | Mean Accuracy | Standard Deviation")
print("------|--------------|-------------------")
for name, metrics in results.items():
    print(f"{name} | {metrics['mean_accuracy']:.4f} | {metrics['std_accuracy']:.4f}")

nn_configs = [
    {
        'hidden_layer_sizes': (10,),
        'activation': 'relu',
        'solver': 'adam'
    },
    {
        'hidden_layer_sizes': (50, 25),
        'activation': 'tanh',
        'solver': 'sgd'
    },
    {
        'hidden_layer_sizes': (100, 50, 25),
        'activation': 'logistic',
        'solver': 'lbfgs'
    }
]
```

## Results and Conclusion for Classification Experiments

In the classification analysis of Chronic Kidney Disease prediction, we evaluated four machine learning models using cross-validation. The models demonstrated very high performance, with accuracies ranging between 94% and 97%. The Neural Network emerged as the top performer, achieving a mean accuracy of 96.73% with the most consistent results across different data subsets. Support Vector Machines closely followed with a 95.42% accuracy.
The performance suggests that the indicators & their corrospondign values that are present in the dataset are accurate predictors of CKD, and therefore valuable to have as data points. The high accuracy across all models indicates that machine learning techniques can potentially provide powerful tools for early disease detection, offering clinicians a reliable method to screen for CDK using a small set of key physiological measurements through ML & AI.

```python
nn_results = {}
for i, config in enumerate(nn_configs, 1):
    nn = MLPClassifier(
        hidden_layer_sizes=config['hidden_layer_sizes'],
        activation=config['activation'],
        solver=config['solver'],
        random_state=42,
        max_iter=1000
    )
    
    cv_scores = cross_val_score(nn, X_scaled, y, cv=5)
    nn_results[f'Config {i}'] = {
        'config': config,
        'mean_accuracy': cv_scores.mean(),
        'std_accuracy': cv_scores.std()
    }

print("\nNeural Network Configuration Experiments:")
print("Config | Hidden Layers | Activation | Solver | Mean Accuracy | Std Deviation")
print("-------|--------------|------------|--------|--------------|---------------")
for name, metrics in nn_results.items():
    config = metrics['config']
    print(f"{name} | {config['hidden_layer_sizes']} | {config['activation']} | {config['solver']} | {metrics['mean_accuracy']:.4f} | {metrics['std_accuracy']:.4f}")
```

The neural network experiments, the configurations varied in hidden layer structure, activation functions, and optimization solvers, based on the advice of 383GPT. The most sophisticated model (again according to 383GPT), featuring three hidden layers with 100, 50, and 25 neurons and using the logistic activation function with the LBFGS solver, achieved the highest performance. This configuration reached 98.02% accuracy with the lowest variability across different data splits. The other configurations did almost just as well, with only a percent or two difference between them.
