```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold, StratifiedKFold
import warnings
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
slime_df = pd.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
display(slime_df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
display(slime_df.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = slime_df[["Temperature °C", "Mols KCL"]]
y = slime_df["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
linear = LinearRegression().fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = X_test.head(1)
print("Predicted: " + str(linear.predict(sample)))
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
print("Coefficient of determination: " + str(linear.score(X_test, y_test)))
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print("Coefficients: " + str(linear.coef_))
print("Intercept: " + str(linear.intercept_))
```

Report on score and explain meaning: This model has an $R^2$ score (coefficient of determination) of 0.855. The coefficient of determination is the proportion of variability in the labels that is accounted for by the features. A value of 0.855 is high, and means that 85.5% of the variability in the size is accounted for by the features.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kfold = KFold(n_splits=5, random_state=42, shuffle=True)
linear_cv = cross_val_score(LinearRegression(), X, y, cv=kfold)
print(linear_cv)
print("Average: " + str(linear_cv.mean()) + " Standard deviation: " + str(linear_cv.std()))
# Report on their finding and their significance
```

Write findings here: The average $R^2$ score is 0.859 across the 5 splits, with a standard deviation of 0.018. The average is very close to the score from part 3, meaning we can interpret it the same way - the score is high, and across the 5 splits an average of 85.9% of the variability in the size was accounted for by the features. The standard deviation is low, meaning that the $R^2$ score was fairly consistent across the 5 splits.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
poly_X = PolynomialFeatures(degree=2).fit_transform(X)
poly_cv = cross_val_score(LinearRegression(), poly_X, y, cv=kfold)
print(poly_cv)
print("Average: " + str(poly_cv.mean()) + " Standard deviation: " + str(poly_cv.std()))
# Report on the metrics and output the resultant equation as you did in Part 3.
poly_X_train, poly_X_test, poly_y_train, poly_y_test = train_test_split(poly_X, y, test_size=0.1, random_state=42)
poly = LinearRegression().fit(poly_X_train, poly_y_train)
print("Coefficient of determination: " + str(poly.score(poly_X_test, poly_y_test)))
print("Coefficients: " + str(poly.coef_))
print("Intercept: " + str(poly.intercept_))
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_df = pd.read_csv("ckd_feature_subset.csv")

X = ckd_df[["age", "bp", "wbcc", "appet_poor", "appet_good", "rbcc"]]
y = ckd_df["Target_ckd"]

skfold = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

ckd_df
```

```python
#Logistic Regression
logistic_cv = cross_val_score(LogisticRegression(), X, y, cv=skfold)
print(logistic_cv)
print("Logistic Regression Average: " + str(logistic_cv.mean()) + " Standard deviation: " + str(logistic_cv.std()))
```

```python
#Support Vector Machines
SVM_cv = cross_val_score(SVC(), X, y, cv=skfold)
print(SVM_cv)
print("Support Vector Machines Average: " + str(SVM_cv.mean()) + " Standard deviation: " + str(SVM_cv.std()))
```

```python
#k-Nearest Neighbors
KNN_cv = cross_val_score(KNeighborsClassifier(), X, y, cv=skfold)
print(KNN_cv)
print("k-Nearest Neighbors Average: " + str(KNN_cv.mean()) + " Standard deviation: " + str(KNN_cv.std()))
```

```python
#Neural Network
base_NN_cv = cross_val_score(MLPClassifier(random_state=42), X, y, cv=skfold)
print(base_NN_cv)
print("Neural Network Average: " + str(base_NN_cv.mean()) + " Standard deviation: " + str(base_NN_cv.std()))
```

## Results and Conclusion for Classification Experiments

| Model | Average | Standard Deviation |
|---|---|---|
| Logistic Regression | 0.8623655913978496 | 0.0338393776563801 |
| Support Vector Machines | 0.9339784946236559 | 0.04730009629719232 |
| k-Nearest Neighbors | 0.9339784946236559 | 0.04730009629719232 |
| Neural Network | 0.8819354838709679 | 0.040986787546499565 |

All models used the default settings. Support vector machines and k-nearest neighbors tied for the highest average accuracy at 0.93, and also had the same standard deviation of 0.047. None of the models performed very poorly, with the worst being logistic regression with an average accuracy of 0.86. Support vector machines and k-nearest neighbors are both better at handling non-linear relationships than logistic regression, likely explaining why those two performed better. Neural networks perform better with large datasets (especially with the solver='adam' default parameter, as explained in the sklearn documentation for MLPClassifier), and this dataset is small, likely explaining why the support vector machines and k-nearest neighbors models did better in this case.

```python
# Experiments with Neural Network.
warnings.filterwarnings("ignore")

for solver in ['adam', "lbfgs", "sgd"]:
    for activation in ['identity', 'logistic', 'tanh', 'relu']:
        print("\nSettings: solver=" + solver + ", activation=" + activation)
        new_NN_cv = cross_val_score(MLPClassifier(solver=solver, activation=activation, random_state=42), X, y, cv=skfold)
        print("Average: " + str(new_NN_cv.mean()) + " Standard deviation: " + str(new_NN_cv.std()))
```
