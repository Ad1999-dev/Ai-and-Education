```python
# Imports section
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

dataset = pd.read_csv('science_data_large.csv')

print("First 15 lines in data set:\n")
print(dataset.head(15))

print("\nInfo about data set:\n")
print(dataset.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

# Step 2: Define features (X) and label (y)
# Assuming the last column is the label, adjust accordingly if it’s different
X = dataset.iloc[:, :-1]  # All columns except the last one
y = dataset.iloc[:, -1]   # The last column

# Step 3: Split the dataset into a training set and a testing set (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Optionally, you can print the shapes of the resulting datasets
print(f"Training set shape: {X_train.shape}, {y_train.shape}")
print(f"Test set shape: {X_test.shape}, {y_test.shape}")
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX


# Step 1: Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 2: Create a sample data point for prediction
# Here, we will create a sample point. Adjust the values as per your features.
# Replace [feature1_value, feature2_value] with actual numeric values.

# Randomly select an index from the training set
random_index = np.random.randint(0, len(X_train))  # Random index within the bounds of X_train

# Select the sample datapoint from the training set

sample_datapoint = pd.DataFrame(X_train.iloc[random_index]).T
predicted_output = model.predict(sample_datapoint)

# Step 3: Report the score of the model
score = model.score(X_test, y_test)

print(f"Predicted output for the sample data point: {predicted_output[0]}")
print(f"Model score (R²): {score}")

# Step 4: Extract coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

# Resulting equation form: y = intercept + coefficient1 * x1 + coefficient2 * x2
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

### Model Score (R²)

The R² (R-squared) score of our linear regression model is approximately **0.8552**.

### Explanation of the Score

The R² score is a statistical measure that represents the proportion of the variance for the dependent variable that is explained by the independent variables in the model. 

1. **Range of R²**: 
   R² values range from 0 to 1:
   - An R² of **1.0** indicates that the model perfectly explains the variance in the dependent variable.
   - An R² of **0.0** means that the model does not improve predictions and is equivalent to using the mean of the dependent variable as a prediction.

2. **Interpretation of 0.8552**: 
   - An R² score of **0.8552** signifies that approximately **85.52%** of the variance in the dependent variable is accounted for by the independent variables.
   - This indicates a strong fit, suggesting that the model explains a significant amount of variability in the outcome. 

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Report on their finding and their significance

# Create a linear regression model instance
model_cv = LinearRegression()

# Step 1: Use cross_val_score to perform 5-fold cross-validation
# Setting cv=5 for grading consistency
scores = cross_val_score(model_cv, dataset.iloc[:, :-1], dataset.iloc[:, -1], cv=5, scoring='r2')

# Print the scores for each fold
print(f"Cross-validation R² scores for each fold: {scores}")
print(f"Mean cross-validation R² score: {scores.mean()}")
```

Write findings here:

- **Mean R² Score**: 0.85682

### Explanation of Findings

   - The scores range from about **0.8392** to **0.8720**, indicating that the model performs consistently well across all the folds.

2. **Mean R² Score**: The average R² score of **0.8568** means that, on average, the model explains about **85.68%** of the variance in the dependent variable. This is a strong indication that the model is effective in making predictions.

3. **Model Performance**: Comparing this mean cross-validation score to previous scores can help us understand if the model is reliable. If this score is similar to what we saw earlier, it suggests that the model is likely not overfitting and can generalize well to new data.

4. **Overall Conclusion**: Cross-validation helps us confirm that the model works well on different parts of the data, giving us more confidence that it will perform well when applied to unseen data.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)

# Report on the metrics and output the resultant equation as you did in Part 3.

# Step 1: Augment the dataset with polynomial features of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(dataset.iloc[:, :-1])  # Create polynomial features

# Step 2: Create a new linear regression model instance
model_poly = LinearRegression()

# Step 3: Perform k-fold cross-validation
scores_poly = cross_val_score(model_poly, X_poly, dataset.iloc[:, -1], cv=5, scoring='r2')

# Print the scores for each fold
print(f"Cross-validation R² scores for polynomial regression: {scores_poly}")
print(f"Mean cross-validation R² score for polynomial regression: {scores_poly.mean()}")

# Step 4: Fit the model and extract coefficients and intercept
model_poly.fit(X_poly, dataset.iloc[:, -1])
coefficients_poly = model_poly.coef_
intercept_poly = model_poly.intercept_

# Output the coefficients and intercept
print(f"Coefficients: {coefficients_poly}")
print(f"Intercept: {intercept_poly}")
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

ckdset = pd.read_csv('ckd_feature_subset.csv')

print(ckdset.head())


# print(ckdset.info())
# print(ckdset.describe())
```

```python
X = ckdset.drop(columns=['Target_ckd'])  # Features
y = ckdset['Target_ckd']  # Target variable


# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a Logistic Regression model
logistic_model = LogisticRegression(random_state=42)

# Perform 5-fold cross-validation
logistic_scores = cross_val_score(logistic_model, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"Logistic Regression Cross-validation Scores: {logistic_scores}")
print(f"Mean Accuracy: {logistic_scores.mean():.5f}")
print(f"Standard Deviation: {logistic_scores.std():.5f}")
```

```python
# Create a Support Vector Classifier model
svc_model = SVC(random_state=42)

# Perform 5-fold cross-validation
svc_scores = cross_val_score(svc_model, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"SVC Cross-validation Scores: {svc_scores}")
print(f"Mean Accuracy: {svc_scores.mean():.5f}")
print(f"Standard Deviation: {svc_scores.std():.5f}")
```

```python
# Create a k-Nearest Neighbors classifier model
knn_model = KNeighborsClassifier()

# Perform 5-fold cross-validation
knn_scores = cross_val_score(knn_model, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"KNN Cross-validation Scores: {knn_scores}")
print(f"Mean Accuracy: {knn_scores.mean():.5f}")
print(f"Standard Deviation: {knn_scores.std():.5f}")
```

```python
# Create a Neural Network model (default settings with one hidden layer)
nn_model = MLPClassifier(random_state=42, max_iter=1000)

# Perform 5-fold cross-validation
nn_scores = cross_val_score(nn_model, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"Neural Network Cross-validation Scores: {nn_scores}")
print(f"Mean Accuracy: {nn_scores.mean():.5f}")
print(f"Standard Deviation: {nn_scores.std():.5f}")
```

### Summary of Results

After evaluating four different classification models on the CKD dataset, we find the following results:

- The **Neural Network** model achieved the highest mean accuracy of **96.73%**, demonstrating its effectiveness in classifying Chronic Kidney Disease cases based on the features provided. At first, 500 iterations was not enough to reach fully optimal convergence, and while it run properly, multiple ConvergenceWarnings were thrown. This was fixed by increasing the number of iterations from 500 to 1000.
- The **Support Vector Machine** model also performed well with a mean accuracy of **95.42%**, indicating strong predictive ability.
- The **Logistic Regression** and **k-Nearest Neighbors** models showed similar performance, both achieving mean accuracies around **94.13%** and **94.11%** respectively.

### Discussion on Performance
The neural network performed best likely due to its ability to capture complex relationships and patterns in the data through multiple layers and non-linear activation functions. The other models, while effective, might not capture such intricate interactions in feature space, particularly with a dataset that may have non-linear relationships.

Given these findings, it would be beneficial to further explore different hyperparameters or configurations for the neural network to see if additional gains in performance can be achieved.

## Results and Conclusion for Classification Experiments

```python
# Experiments with Neural Network.

# Configuration 1: Single Hidden Layer with ReLU Activation

nn_model1 = MLPClassifier(hidden_layer_sizes=(10,), activation='relu', random_state=42, max_iter=2000)
nn_scores1 = cross_val_score(nn_model1, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"Neural Network Cross-validation Scores: {nn_scores1}")
print(f"Mean Accuracy: {nn_scores1.mean():.5f}")
print(f"Standard Deviation: {nn_scores1.std():.5f}")
```

```python
# Configuration 2: Two Hidden Layers with ReLU Activation

nn_model2 = MLPClassifier(hidden_layer_sizes=(50, 25), activation='relu', random_state=42, max_iter=2000)
nn_scores2 = cross_val_score(nn_model2, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"Neural Network Cross-validation Scores: {nn_scores2}")
print(f"Mean Accuracy: {nn_scores2.mean():.5f}")
print(f"Standard Deviation: {nn_scores2.std():.5f}")
```

```python
# Configuration 3: More Complex Network with ReLU Activation

nn_model3 = MLPClassifier(hidden_layer_sizes=(100, 50, 25), activation='relu', learning_rate_init=0.001, random_state=42, max_iter=1000)
nn_scores3 = cross_val_score(nn_model3, X_scaled, y, cv=5, scoring='accuracy')

# Print results
print(f"Neural Network Cross-validation Scores: {nn_scores3}")
print(f"Mean Accuracy: {nn_scores3.mean():.5f}")
print(f"Standard Deviation: {nn_scores3.std():.5f}")
```
