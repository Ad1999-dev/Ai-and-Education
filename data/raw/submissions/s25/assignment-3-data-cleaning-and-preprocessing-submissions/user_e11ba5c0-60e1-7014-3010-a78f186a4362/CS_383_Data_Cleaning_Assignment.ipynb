## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value

isna = merged_data.isna()
count = sum(isna.sum(axis=1) == 0)

missing_rows_percent = count / len(merged_data)

# Print %

print(f"{100 - missing_rows_percent * 100}%")

# Drop these rows from the dataset

nomissing_data = merged_data.dropna()

# Print the Dataset

display(nomissing_data)
nomissing_data.columns
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting

label_sorted_data = nomissing_data.sort_values(by="Target")
label_sorted_data.reset_index(inplace=True, drop=True)

# Print the dataset

label_sorted_data
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

onehot_sorted_data = pd.get_dummies(label_sorted_data, columns=categorical_columns)

# Print the dataset
onehot_sorted_data
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: If we used a single column for the binary variable (instead of two onehot columns), the downstream modeling would be the equivalent. This is because the onehot columns have the same number of possible values [0,1] and [1,0], vs [0,] and [1,], so the model would be able to learn the same. 

This changes if there are more than 2 values. If the type of condiment can take 5 different values, and we encode each to 0, 1, 2, 3, and 4 respectively, then anything a machine learning model learns about condiment 4 might be partially applied to condiment 3 since 3 is close to 4, and vice versa, even if condiment 3 and condiment 4 might not be related. In fact, anything a model learns about any of the condiments might erroneously affect its beliefs about the other condiments as well. 

In other words, we are trying to combine 5 mutually orthogonal directions (the "ketchup", "mustard", "ranch", etc directions) into only 1 dimension, which is impossible. 

However, if the condiments are all very similar but are ordered in spiciness from 0 to 4, representing this as a single numerical variable might be a good idea. Then the model might learn that more spicy means one thing while less spicy means another thing; then what it learns about 4 can be correctly partially applied to 3. In other words, there are only 2 directions ("spicy" vs "not spicy"), just quantized, which DOES fit in 1 dimension.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

num_col_means = onehot_sorted_data[numerical_columns].mean(axis=0)
num_col_stdev = onehot_sorted_data[numerical_columns].std(axis=0)

# a mask on EVERY numerical value, checking if it is within 3 sigma of the mean.
mask = (onehot_sorted_data[numerical_columns] <= num_col_means + 3 * num_col_stdev) & \
       (onehot_sorted_data[numerical_columns] >= num_col_means - 3 * num_col_stdev)

# display(mask)
# mask on the _rows_ of the df, each True row has ALL values within 3 sigma of the mean.
mask = mask.all(axis=1)
# display(mask)

no_outliers_data = onehot_sorted_data[mask].reset_index(drop=True)

# Print the dataset
no_outliers_data
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.

# num_col_means2 = no_outliers_data[numerical_columns].mean(axis=0)
# num_col_stdev2 = np.sqrt(no_outliers_data[numerical_columns].var(axis=0))

# normalized_data = no_outliers_data.copy()
# normalized_data[numerical_columns] = (normalized_data[numerical_columns] - num_col_means2) / num_col_stdev2
# normalized_data = normalized_data.reset_index(drop=True)

row_mins = no_outliers_data[numerical_columns].min(axis=0)
row_maxs = no_outliers_data[numerical_columns].max(axis=0)

normalized_data = no_outliers_data.copy()
normalized_data[numerical_columns] = (normalized_data[numerical_columns] - row_mins) / (row_maxs - row_mins)
normalized_data = normalized_data


# Print the dataset
normalized_data
````
