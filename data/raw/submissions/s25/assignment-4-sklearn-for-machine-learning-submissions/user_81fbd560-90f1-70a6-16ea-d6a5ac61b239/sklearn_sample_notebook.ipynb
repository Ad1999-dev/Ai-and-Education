```python
# Imports section
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn import datasets
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
dataframe = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(dataframe.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(dataframe.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = dataframe[['Temperature °C', 'Mols KCL']]
Y = dataframe['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.9, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(x_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = pd.DataFrame([[200, 700]])
predicted_size = model.predict(sample_data)
print(predicted_size)

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print(model.score(x_train, y_train))

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
data = pd.read_csv('ckd_feature_subset.csv')

X = data.drop(columns=['Target_ckd']) 
y = data['Target_ckd']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=42)

def evaluate_model(model, X, y):
    scores = cross_val_score(model, X, y, cv=5)
    return scores.mean(), scores.std()

# Initialize classifiers
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Support Vector Machines": SVC(),
    "k-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(max_iter=1000)
}

# Evaluate each model
results = {}
for model_name, model in models.items():
    mean_accuracy, std_dev = evaluate_model(model, X, y)
    results[model_name] = (mean_accuracy, std_dev)

# Create a DataFrame for results
results_df = pd.DataFrame(results, index=["Mean Accuracy", "Standard Deviation"]).T
print(results_df)
```

## Results and Conclusion for Classification Experiments

Overall, each model performed pretty well with the lowest having a mean accuracy of 0.856. The Logistic Regression model performed the worst which could be because this model is linear in nature and does not capture the complexities in the data. The SVC and KNN models had the same mean accuracy but with different standard deviations. This suggests that SVC could vary across different subsets of data showing more sensitivity than the KNN model. The best model was the neural network likely due to its capacity to learn complex patterns within the features. It had the highest accuracy and lowest standard deviation indicating it is extremely accurate and robust to variations in the dataset.

```python
# Experiments with Neural Network. Report of 3 minimum
nn_configurations = {
    "Config 1": MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000),
    "Config 2": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000),
    "Config 3": MLPClassifier(hidden_layer_sizes=(50, 25), max_iter=1000)
}

nn_results = {}
for config_name, model in nn_configurations.items():
    mean_accuracy, std_dev = evaluate_model(model, X, y)
    nn_results[config_name] = (mean_accuracy, std_dev)

# Create a DataFrame for Neural Network results
nn_results_df = pd.DataFrame(nn_results, index=["Mean Accuracy", "Standard Deviation"]).T
print(nn_results_df)
```

## Results and Conclusion for Neural Network Experiments

Overall every configuration had a very high mean accuracy suggesting how reliable neural networks are. There is a significant improvement in accuracy from Config 1 to Config 3 indicating that adding a second layer (and the additional neurons) allows the model to capture more complex patterns in the data. Increasing the number of neurons from 50 (Config 1) to 100 (Config 2) appears to positively affect the model with more consistency and accuracy. Overall, the parameter of number of layers in each configuration made the biggest difference in which it would significantly enhance the classification of the model for this dataset. Neural networks are very effective and the more parameters, the better they perform.
