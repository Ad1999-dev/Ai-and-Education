```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
load_science_dataset_QL = pd.read_csv("science_data_large.csv")
display(load_science_dataset_QL.head(15))
display(load_science_dataset_QL.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_features = load_science_dataset_QL.iloc[:,:-1]
Y_label = load_science_dataset_QL.iloc[:,-1]
X_set_train, X_set_test, Y_set_train, Y_set_test = train_test_split(X_features,\
                            Y_label,test_size=0.1,train_size= 0.9,random_state=42)
display(X_set_test)
display(Y_set_test)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
dummi_data_sample = pd.DataFrame({"Temperature °C": [90], "Mols KCL" : [661]})
applied_Linear_Regression_model = LinearRegression()
applied_Linear_Regression_model.fit(X_set_train,Y_set_train)
linear_re_score = applied_Linear_Regression_model.score(X_set_test,Y_set_test)
predict_sample_point = applied_Linear_Regression_model.predict(dummi_data_sample).round(5)
coefficientd_values = applied_Linear_Regression_model.coef_.round(5)
intercept_value = applied_Linear_Regression_model.intercept_.round(5)
linear_re_data_frame = pd.DataFrame({
    "Sample" : [dummi_data_sample],
    'Predict sample output': [predict_sample_point],
    'Intercept point': [intercept_value],
    'coefficientd values': [coefficientd_values],
    'Linear Score': [linear_re_score],
})
display(linear_re_data_frame)
```

Equation : $$h(x) = \hat{y}(w, x) = -409391.47958 + 866.14641*temperature + 1032.69507* Mols KCL$$

Report on score and explain meaning: The score is 0.8552477276095, and is residual sum of squares. This score measures how well the model's predictions align with the actual values. A score of 1 indicates that the model's predictions perfectly match the true values, while a score of 0 means no alignment. Based on the returned score, the Linear Regression model demonstrates relatively good accuracy about 85% matching.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
# Report on their finding and their significance
cross_valid_QL = KFold(n_splits=5, random_state=42, shuffle=True)
cross_scores = cross_val_score(applied_Linear_Regression_model,X_features,Y_label,cv=cross_valid_QL)
mean_of_folds = cross_scores.mean()
stdard_QLoc_diation = cross_scores.std()

folds_score_BR_data_frame = pd.DataFrame({
    'Fold': np.arange(1, len(cross_scores) + 1),
    'Score': cross_scores
})
metrics = pd.DataFrame({
    'Mean' : [mean_of_folds],
    'standard Deviation': [stdard_QLoc_diation]
})
display(folds_score_BR_data_frame)
display(metrics) 
```

Write findings here: The `cross_val_score` function returns five different values (0.861519, 0.827423, 0.871952, 0.881662, 0.856091) corresponding to the five splits. The scores do not vary significantly, as the standard deviation is approximately 0.018. The mean accuracy is about 0.86, indicating that the model has performed quite well during training. After rerun multiple times, those values remain unchanged

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
poly_regeresion_form = PolynomialFeatures(degree=2, include_bias= False)
X_features_in_poly_form = poly_regeresion_form.fit_transform(X_features)

applied_Linear_Regression_model.fit(X_features_in_poly_form,Y_label)

coefficientd_values_after_poly_transform = applied_Linear_Regression_model.coef_

intercept_value_after_poly_transform = applied_Linear_Regression_model.intercept_

scores_on_poly_cross_validation = cross_val_score(applied_Linear_Regression_model,X_features_in_poly_form,\
                                                  Y_label,cv=cross_valid_QL)

poly_re_mean = scores_on_poly_cross_validation.mean()
poly_re_std_dev = scores_on_poly_cross_validation.std()
dummi_data_sample_after_poly = poly_regeresion_form.fit_transform(dummi_data_sample)

poly_re_data_fr = pd.DataFrame({
    "predict sample output": [applied_Linear_Regression_model.predict(dummi_data_sample_after_poly)],
    "coefficents" : [coefficientd_values_after_poly_transform.round(5)],
    "intercept": [intercept_value_after_poly_transform.round(5)],
    "scores": [scores_on_poly_cross_validation],
    "mean": [poly_re_mean],
    "standard deviation": [poly_re_std_dev],
})
display(poly_re_data_fr)

# Report on the metrics and output the resultant equation as you did in Part 3.
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
# Imports section
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPClassifier

load_ckd_feature_set = pd.read_csv("ckd_feature_subset.csv")
cross_valid_QL = KFold(n_splits=5, random_state=42, shuffle=True)
X_features_ckd_feature = load_ckd_feature_set.iloc[:,:-1]
Y_label_ckd_feature = load_ckd_feature_set.iloc[:,-1]
display(load_ckd_feature_set.head(15))
display(load_ckd_feature_set.info())


classification_training_models = {
    "logistic_re_model" : LogisticRegression(),
    "support_vactor_mac_modl" : svm.SVC(),
    "K_near_neighbor_model" : KNeighborsClassifier(),
    "Neural Network": MLPClassifier()
}
```

```python
def Test_Classification_Models (classification_training_models):
    for key, learning_model_training in classification_training_models.items():
        learning_model_score = cross_val_score(learning_model_training,X_features_ckd_feature,\
                                           Y_label_ckd_feature,cv=cross_valid_QL,scoring='accuracy') 
        mean_of_folds_of_model = learning_model_score.mean()
        std_dev_of_models = learning_model_score.std()
Test_Classification_Models(classification_training_models)
```

```python
# Experiments with Neural Network.
#1 change hidden layer size:
classification_training_models ["Neural Network Change Hidden Layer Size to 500"]\
      = MLPClassifier(hidden_layer_sizes= [500,500,500])
#1.1 change hidden layer size 1:
classification_training_models ["Neural Network Change Hidden Layer Size to 10"]\
      = MLPClassifier(hidden_layer_sizes= 10)
#2 change activation:
classification_training_models ["Neural Network Change from relu to tanh"]\
        = MLPClassifier( activation="tanh")
#3 change solver:
classification_training_models ["Neural Network Change adam -> sgd"] \
        = MLPClassifier(solver="sgd")
Test_Classification_Models(classification_training_models)
```
