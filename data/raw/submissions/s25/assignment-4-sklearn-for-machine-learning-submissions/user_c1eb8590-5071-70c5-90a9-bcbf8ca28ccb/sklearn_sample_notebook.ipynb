```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split , cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
science_df = pd.read_csv("science_data_large.csv")
print(science_df.head(15))
print(science_df.describe())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = science_df[['Mols KCL', 'Temperature °C']]  # Features
y = science_df['Size nm^3']              # Label

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame({'Mols KCL': [500], 'Temperature °C': [500]})
prediction = reg_model.predict(sample)
print("Prediction:", prediction)
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
model_score = reg_model.score(X_test, y_test)
print("Model score:", model_score)
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coeff = reg_model.coef_
intercept = reg_model.intercept_

print("Coeffecients :" ,coeff,"Intercepts :", intercept )
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
# Load the dataset. Then train and evaluate the classification models.


ckd_df = pd.read_csv("ckd_feature_subset.csv")
X = ckd_df[['age', 'bp', 'wbcc', 'appet_poor', 'appet_good', 'rbcc']]
y = ckd_df['Target_ckd']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.1, random_state=42)

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Support Vector Machines': SVC(max_iter=1000),  
    'k-Nearest Neighbors': KNeighborsClassifier(),
    'Neural Network': MLPClassifier(max_iter=1000)
}

cv_results = pd.DataFrame()

# Perform 5-fold cross-validation for each model and collect accuracy scores
for model_name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    cv_results[model_name] = scores  # Add the scores to the DataFrame

# Display the results
print("Cross-Validation Accuracy Scores:")
print(cv_results)

#Holistic score result 
mean_scores = cv_results.mean().rename('Mean Accuracy')
std_scores = cv_results.std().rename('Standard Deviation')
summary = pd.concat([mean_scores, std_scores], axis=1)
print("\nSummary of Cross-Validation Results:")
print(summary)
```

## Results and Conclusion for Classification Experiments

1. Logistic Regression - The accuracy for logistic regression across the 5 folds ranged from .7667 -.9333. The model had a wide range when it came to accuracy, showing that even though it can distinguish between classes, it may struggle within several subdatasets. 
2. SVM - The SVM showed strong and consistent performance within the 5 folds as it ranged from from .8709678 - 1.000. 2 of the folds had 1.0000, which show that there is a chance the model is being overfit on the data, that it is not able to build proper patterns and is following hte noise in the model.
3. K-NN - KNN also had strong and consistent scores within the 5 folds as it ranged from .90 - .967742. The model was thus able to properly distinguise between the classes and does not show unusual signs of being overfit, making it a strong contender for best model in this list. 
4. Neural Network ( MLP CLassifier) - The Neural Network had the best scores as it ranged from .9032 - .9677, showing lower variability than KNN while also not being overfit to the given data. These scores show that the model is capable of generalizing and it is able to keep this high accuracy without much variance. 

```python
# Experiments with Neural Network.
# Configuration 1: Simple network
nn_configs ={
'Config 1':MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42),

# Configuration 2: Deeper network
'Config 2' :MLPClassifier(hidden_layer_sizes=(50, 25), max_iter=1000, random_state=42),
# Configuration 3: Single layer network with tanh activation
'Config 3' : MLPClassifier(hidden_layer_sizes=(100,), activation='tanh', max_iter=1000, solver='adam', early_stopping=True, random_state=42)
}

nn_results = pd.DataFrame()

for config_name, model in nn_configs.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    nn_results[config_name] = scores  


print("Neural Network Configuration Accuracy Scores:")
print(nn_results)

```
