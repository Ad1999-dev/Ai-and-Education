```python
# Imports section
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
science_dataset = pd.read_csv('science_data_large.csv')
display(science_dataset.head(15))
display(science_dataset.describe())
science_dataset.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = science_dataset[['Temperature °C', 'Mols KCL']]  # Features
y = science_dataset['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_datapoint = pd.DataFrame({'Temperature °C': [123], 'Mols KCL': [683]})
predicted_output = lr_model.predict(sample_datapoint)
print("Predicted Output:", predicted_output)


# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
model_score = lr_model.score(X_test, y_test)
print("Model Score:", model_score)


# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = lr_model.coef_
print("Coefficients:", coefficients)
intercept = lr_model.intercept_
print("Intercept:", intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

cv = KFold(n_splits=5, random_state=42, shuffle=True)
scores = cross_val_score(lr_model, X, y, cv=cv)
print("Scores:", scores)
mean = np.mean(scores)
print("Mean:", mean)

# Report on their finding and their significance
```

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
print("Columns:", poly.get_feature_names_out())

X_train, X_test = train_test_split(X_poly, test_size=0.1, random_state=42)

lr_model_poly = LinearRegression(fit_intercept=False)
lr_model_poly.fit(X_train, y_train)

sample_datapoint = poly.fit_transform(pd.DataFrame({'Temperature °C': [123], 'Mols KCL': [683]}))
predicted_output = lr_model_poly.predict(sample_datapoint)
print("Predicted Output:", predicted_output)

model_score = lr_model_poly.score(X_train, y_train)
print("Model Score:", model_score)

coefficients = lr_model_poly.coef_
print("Coefficients:", coefficients)
intercept = lr_model_poly.intercept_
print("Intercept:", intercept)

cv = KFold(n_splits=5, random_state=42, shuffle=True)
scores = cross_val_score(lr_model_poly, X_poly, y, cv=cv)
print("Cross Validation Scores:", scores)
mean = np.mean(scores)
print("Cross Validation Mean:", mean)

# Report on the metrics and output the resultant equation as you did in Part 3.
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_dataset = pd.read_csv('ckd_feature_subset.csv')
display(ckd_dataset.head(15))
ckd_dataset.info()
```

```python
X = ckd_dataset.drop('Target_ckd', axis=1)  # Adjust 'class' if your target column is named differently
y = ckd_dataset['Target_ckd'] 

models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Support Vector Machine": SVC(random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(max_iter=1500, random_state=42)
}

# Perform 5-fold cross validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store the results
results = {}

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=kf)
    results[name] = {
        'Mean Accuracy': np.mean(scores),
        'Standard Deviation': np.std(scores)
    }

# Convert results to a DataFrame for better visualization
results_df = pd.DataFrame(results).T
results_df
```

## Results and Conclusion for Classification Experiments

### Model Performance Summary

| Model                  | Mean Accuracy | Standard Deviation |
|-----------------------|---------------|---------------------|
| Logistic Regression    | 0.856559      | 0.066269            |
| Support Vector Machine | 0.928172      | 0.047601            |
| k-Nearest Neighbors    | 0.927957      | 0.052440            |
| Neural Network         | 0.935054      | 0.040466            |

The logistic regression model had the worst accuracy of 0.856559 and the highest standard deviation of 0.066269. This means it had the worst result of all the models and that its results were less consistent than other models.

The SVM performed significantly better than the Logistic Regression and more in line with the other models with an accuracy of 0.928172 and a standard deviation of 0.047601. This means it is both very accurate and also consistent with its results.

The kNN algorithm was marginally less accurate than the SVM with a mean of 0.927957 and it had a slightly higher standard deviation of 0.052440. This means that the model's accuracy is very similar to the SVM but is is less consistent than the SVM algorithm.

Finally, the neural network was the most accurate with a mean accuracy of 0.927957 and the most consistent with a standard deviation of 0.040466.

The neural network is the best performing model in this experiment, achieving the highest mean accuracy with the lowest standard deviation. This suggests that it not only correctly classifies a high percentage of instances but also maintains consistency across the different subsets of the dataset. The results suggest that the chronic kidney disease classification task benefits from models that can capture the underlying complexity in the data. The neural network's  performance implies it can model intricate relationships between features, which may be critical for accurately predicting CKD.

```python
# Experiments with Neural Network.
configurations = [
    {'hidden_layer_sizes': (10), 'max_iter': 100},
    {'hidden_layer_sizes': (10), 'max_iter': 1000},
    {'hidden_layer_sizes': (50), 'max_iter': 100},
    {'hidden_layer_sizes': (50), 'max_iter': 1000}
]

# Store results for each configuration
nn_results = {}

for config in configurations:
    model = MLPClassifier(hidden_layer_sizes=config['hidden_layer_sizes'], activation='relu', max_iter=config['max_iter'], random_state=42)
    scores = cross_val_score(model, X, y, cv=kf)
    nn_results[f"Layer Size: {config['hidden_layer_sizes']}, Max Iterations: {config['max_iter']}"] = {
        'Mean Accuracy': np.mean(scores),
        'Standard Deviation': np.std(scores)
    }

# Convert results to a DataFrame for better visualization
nn_results_df = pd.DataFrame(nn_results).T
nn_results_df
```

## Results and Conclusion for Neural Network Experiments

Results Analysis
1. Layer Size: 10, Max Iterations: 100
    - This configuration yielded the lowest mean accuracy of 82.4% with a standard deviation of 7.4%. The model's performance suggests that a smaller hidden layer with fewer iterations may not capture the complexities of the data adequately.
2. Layer Size: 10, Max Iterations: 1000
    - Increasing the iterations to 1000 improved accuracy significantly to 90.9%, with a standard deviation of 5.2%. This indicates that allowing more training time leads to better learning from the data, even with a small hidden layer.
3. Layer Size: 50, Max Iterations: 100
    - With a hidden layer size of 50 and 100 iterations, the accuracy slightly improved to 84.3% with a standard deviation of 6.5%. This suggests that increasing the layer size helps but still requires more iterations for optimal performance.
4. Layer Size: 50, Max Iterations: 1000
    - The best performance came from this configuration, achieving a mean accuracy of 93.5% and a standard deviation of 4.0%. This result indicates that both a larger hidden layer and higher iteration count work synergistically to capture complex patterns in the data effectively.

Looking at all these results, we can say that increasing the number of iterations has a greater impact than increasing the hidden layer size, especially when the number of iterations is below the limit for the model to converge. We might be able to get better results, after increasing the number of iterations till it converges, by both increasing the hidden layer size and the number of hidden layers beyond one.
