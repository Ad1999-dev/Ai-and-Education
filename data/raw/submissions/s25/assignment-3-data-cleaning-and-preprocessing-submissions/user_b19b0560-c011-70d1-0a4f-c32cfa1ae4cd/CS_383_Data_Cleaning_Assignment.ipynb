Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
# Define the categorical columns to be transformed
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

# Step 1: Create dummy variables explicitly for 'rbc' and other binary columns
for col in categorical_columns:
    if col in cleaned_combined_dataset.columns and cleaned_combined_dataset[col].dtype == 'object':
        # Create dummy variables with drop_first=False to keep both categories
        dummies = pd.get_dummies(cleaned_combined_dataset[col], prefix=col, drop_first=False)
        
        # Convert boolean values to integers
        dummies = dummies.astype(int)
        
        # Combine the new dummy columns with the original dataset
        cleaned_combined_dataset = pd.concat([cleaned_combined_dataset, dummies], axis=1)

# Step 2: Remove the original categorical columns if they are still in the DataFrame
# Check if the columns exist before dropping
existing_columns_to_drop = [col for col in categorical_columns if col in cleaned_combined_dataset.columns]
cleaned_combined_dataset.drop(columns=existing_columns_to_drop, inplace=True)

# Step 3: Print the transformed dataset
print("\nDataset after converting categorical columns to dummy variables:")
print(cleaned_combined_dataset)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: When dealing with binary variables, using a single column to indicate the presence (1) or absence (0) of a category is equivalent to creating two separate dummy variables. This approach simplifies the model without losing information. However, for categorical features with more than two values, such as "type of condiment" that can take five values, it is crucial to use dummy variables (one-hot encoding) instead of a single numeric column. This ensures that the model accurately interprets the categorical variable without imposing an artificial order. Using dummy variables allows the model to treat each category independently, enhancing its ability to learn and make predictions.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

# Step 1: Create a copy of the dataset to work with
cleaned_dataset_no_outliers = cleaned_combined_dataset.copy()

# Step 2: Loop through each numerical column to identify and drop outliers
for column in numerical_columns:
    if column in cleaned_dataset_no_outliers.columns:
        mean = cleaned_dataset_no_outliers[column].mean()
        std_dev = cleaned_dataset_no_outliers[column].std()
        
        # Define lower and upper bounds for outliers
        lower_bound = mean - 3 * std_dev
        upper_bound = mean + 3 * std_dev
        
        # Filter out the outliers
        cleaned_dataset_no_outliers = cleaned_dataset_no_outliers[(cleaned_dataset_no_outliers[column] >= lower_bound) & 
                                                                  (cleaned_dataset_no_outliers[column] <= upper_bound)]

# Step 3: Print the cleaned dataset without outliers
print("\nCleaned Dataset (after removing outliers):")
print(cleaned_dataset_no_outliers)
````
