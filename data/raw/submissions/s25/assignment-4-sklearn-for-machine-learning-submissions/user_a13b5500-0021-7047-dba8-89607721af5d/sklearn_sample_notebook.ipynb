```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

df = pd.read_csv('science_data_large.csv')
print(df.head(15))
print(df.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X = df[['Temperature °C', 'Mols KCL']]  # Features
y = df['Size nm^3']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
print("Training feature set shape (X_train):", X_train.shape)
print("Testing feature set shape (X_test):", X_test.shape)
print("Training label set shape (y_train):", y_train.shape)
print("Testing label set shape (y_test):", y_test.shape)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

model = LinearRegression()
model.fit(X_train, y_train)

sample_data = pd.DataFrame([[500, 600]], columns=['Temperature °C', 'Mols KCL'])  
predicted_size = model.predict(sample_data)

print("Predicted size for the sample data point [Temperature = 500°C, Mols KCl = 600]:", predicted_size[0])

train_score = model.score(X_train, y_train)
print("Training score (R²):", train_score)

coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

The equation for our linear regression model is:

$$ h(x) = \theta_0 + \theta_1 \cdot \text{Temperature °C} + \theta_2 \cdot \text{Mols KCL} $$

Where:
- \(\theta_0\) is the intercept
- \(\theta_1\) is the coefficient for Temperature °C

Report on score and explain meaning:

The score reported by the Linear Regression model is the coefficient of determination, commonly known as R². This value ranges from 0 to 1 (though it can be negative with very poor fits), where:

- 1 indicates perfect prediction where the model explains all the variability in the data
- 0 means the model fails to explain any of the variability (as good as just predicting the mean)
- A negative score (possible with test data) means the model performs worse than just predicting the mean

In simpler terms, it measures how well our linear equation fits the actual data points. A higher score means our predictions are closer to the real values

In this case, the R² score of the model is approximately **0.8610**. This means that the model explains approximately **86.10%** of the variance in slime size based on the features provided (Temperature and Mols of KCl).

The R² score of **0.8610** indicates a good fit for the model, suggesting that it effectively captures the relationship between the input features and the slime growth size. Further validation with unseen data is recommended to confirm the model's generalizability.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Report on their finding and their significance

model = LinearRegression()
scores = cross_val_score(model, X, y, cv=5)

print("Cross-validation scores for each fold:", scores)
print("Mean cross-validation score:", scores.mean())
print("Standard deviation of cross-validation scores:", scores.std())
```

Write findings here:

We conducted a 5-fold cross-validation on our linear regression model using the `cross_val_score` function from scikit-learn. This process divided our dataset into 5 subsets, training the model on 4 folds and testing on the remaining 1 fold, iterating this process 5 times with different test folds.

#### Findings:
- **Individual Fold Scores**: The R² scores for each of the 5 folds were: [0.8392, 0.8705, 0.8587, 0.8720, 0.8436]
- **Mean R² Score**: The average R² score across all folds was 0.8568
- **Standard Deviation**: The standard deviation of the R² scores was 0.0135

#### Significance:


1. **Model Performance**: The mean R² score of 0.8568 indicates that the model, on average, explains approximately 85.68% of the variance in the size of the slime growth based on the input features (temperature and mols of KCl). A score close to 1 indicates that our model performs well in generalizing to unseen data, suggesting a robust relationship between the input features and the growth size of the slime.

2. **Standard Deviation**: The calculated standard deviation of the cross-validation scores was 0.0135. This low standard deviation implies that the model's performance is both consistent and stable across different data shuffles, highlighting that our model’s performance does not vary significantly between different training and testing splits.
3. **Comparison to Single Split**: In Part 3, we obtained a training R² score of 0.8610 from a single 90/10 train-test split. The cross-validation mean R² of 0.8568 is very close to this value (a difference of only 0.0042, or 0.42%), which strongly supports the reliability of our original model. This small difference suggests that the model is not overfitting to the single split from Part 3 and performs similarly well across multiple data shuffles. The cross-validation score, being an average over multiple tests, is likely a more realistic estimate of how the model would perform on unseen data.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

df = pd.read_csv('ckd_feature_subset.csv')
df.head()
```

```python
X = df.drop(columns='Target_ckd')  # Feature variables
y = df['Target_ckd']                # Target variable

# Check the shapes to confirm the split
print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Support Vector Machine': SVC(random_state=42),
    'k-Nearest Neighbors': KNeighborsClassifier(),
    'Neural Network': MLPClassifier(max_iter=1000, random_state=42)
}

# Dictionary to hold results
results = {}

# Evaluate each model using 5-fold cross-validation
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5)
    results[name] = {
        'Mean Accuracy': np.mean(scores),
        'Standard Deviation': np.std(scores)
    }

# Convert results into a DataFrame for better readability
results_df = pd.DataFrame(results).T
results_df
```

```python
# Determine which model performed best
best_model = results_df['Mean Accuracy'].idxmax()
best_score = results_df['Mean Accuracy'].max()

print(f"The best model is: {best_model} with an average accuracy of {best_score:.4f}.")
```

## Key Observations

1. **Neural Network Performance**:
   - The Neural Network achieved the highest mean accuracy of **0.9413**, indicating that it correctly classified approximately 94.13% of CKD cases during cross-validation.
   - It also had the lowest standard deviation of **0.0240**, suggesting robust and consistent performance across different folds.

2. **Support Vector Machine and k-Nearest Neighbors**:
   - Both SVM and KNN achieved a mean accuracy of **0.9282**. This strong performance shows that both models effectively classify CKD cases.
   - However, KNN displayed a lower standard deviation of **0.0319** compared to SVM's **0.0599**, indicating slightly more consistency in its results.

3. **Logistic Regression**:
   - The Logistic Regression model had the lowest mean accuracy of **0.8561**.
   - Its standard deviation of **0.0542** indicates more variability in its performance compared to the Neural Network and KNN.

## Summary of Results

The results indicate that the **Neural Network** was the top performer among the tested models. It not only achieved the highest accuracy but also demonstrated the most consistent results across different data splits.

I think the neural network performed the best as neural networks are capable of modeling complex, non-linear relationships between the input features and the target variable. In the context of the CKD dataset, the interactions between various clinical features may not be linearly separable, and a Neural Network can effectively learn these intricate relationships.

I also think another major factor for its performance is the Hierarchical Feature Learning architecture. Neural Networks operate by layering neurons in multiple layers (hidden layers). This architecture allows them to learn increasingly abstract representations of the input data. For instance, initial layers may detect simple features (like ranges of age or blood pressure), while deeper layers can identify complex patterns indicative of CKD.

Logistic regression and k nearest neighbors performed the worst and this could be since they are not as good as capturing non-linear relationships and patterns with many features involved.

## Results and Conclusion for Classification Experiments

```python
# Experiments with Neural Network.

# Different configurations for Neural Networks
best_configs = {
    'Config 1': MLPClassifier(hidden_layer_sizes=(20, 20), 
                              max_iter=2000, 
                              alpha=0.01,  # L2 regularization
                              random_state=42),
    
    'Config 2': MLPClassifier(hidden_layer_sizes=(10,), 
                              max_iter=2000, 
                              learning_rate_init=0.01, 
                              activation='tanh', 
                              random_state=42),
    
    'Config 3': MLPClassifier(hidden_layer_sizes=(10,), 
                              max_iter=2000, 
                              alpha=0.001,  # Smaller L2 regularization
                              learning_rate_init=0.001, 
                              random_state=42)
}

# Dictionary to hold results for the best configurations
best_results = {}

# Evaluate each configuration using 5-fold cross-validation
for config_name, model in best_configs.items():
    scores = cross_val_score(model, X, y, cv=5)
    best_results[config_name] = {
        'Mean Accuracy': np.mean(scores),
        'Standard Deviation': np.std(scores)
    }

# Convert results into a DataFrame for better readability
best_results_df = pd.DataFrame(best_results).T
best_results_df
```

```python
# Determine which neural network configuration performed best
best_nn_model = best_results_df['Mean Accuracy'].idxmax()
best_nn_score = best_results_df['Mean Accuracy'].max()
print(f"The best neural network configuration is: {best_nn_model} with an average accuracy of {best_nn_score:.4f}.")
```
