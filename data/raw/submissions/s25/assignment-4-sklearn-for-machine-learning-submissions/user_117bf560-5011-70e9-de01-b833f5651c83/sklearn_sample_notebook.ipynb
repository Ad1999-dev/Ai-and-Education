```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
data = pd.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
print(data.head(15))
print('\n')
# Display a summary of the table information (number of datapoints, etc.)
data.info()
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data[["Temperature °C", "Mols KCL"]]
Y = data["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
reg = model.fit(X_train, Y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
print(reg.predict(pd.DataFrame({'Temperature °C': [100], 'Mols KCL' : [100]})))
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print(reg.score(X_test, Y_test))
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print(reg.coef_)
print(reg.intercept_)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, Y, cv=kf)

# Report on their finding and their significance
print(scores)
scores.mean()
```

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
newData = PolynomialFeatures(2)
newX = newData.fit_transform(X)


newScores = []
coefficients = []

for train_index, test_index in kf.split(newX):
    x_train, x_test = newX[train_index], newX[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    # Create and fit the linear regression model
    model = LinearRegression()
    newReg = model.fit(x_train, y_train)

    # Store the coefficients for each fold
    coefficients.append(newReg.coef_)  
    newScores.append(newReg.score(x_test, y_test))
# Convert coefficients list to a DataFrame for easier analysis
coeffs_df = pd.DataFrame(coefficients, columns = newData.get_feature_names_out())
newScores_df = pd.DataFrame(newScores)

# Report on the metrics and output the resultant equation as you did in Part 3.

print(coefficients)

print(newScores)
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
data = pd.read_csv("ckd_feature_subset.csv")

X = data[["age", "bp", 'wbcc', 'rbcc', 'appet_poor', 'appet_good']]
Y = data[['Target_ckd']].values
Y = Y.ravel()

kf = KFold(n_splits=5, shuffle=True, random_state=42)

logModel = LogisticRegression(max_iter=200)
log_reg_scores = cross_val_score(logModel, X, Y, cv=kf) 
log_reg_score = log_reg_scores.mean()
log_reg_SD = log_reg_scores.std()
print("Logistic Regression Score Mean : ", log_reg_score, " Standard Deviation : ", log_reg_SD)

svcModel = SVC()
svc_scores = cross_val_score(svcModel, X, Y, cv=kf) 
svc_score = svc_scores.mean()
svc_SD = svc_scores.std()
print("Support Vector Machines Score Mean : ", svc_score, " Standard Deviation : ", svc_SD)


neighborsModel = KNeighborsClassifier(n_neighbors = 3)
neighbors_scores = cross_val_score(neighborsModel, X, Y, cv=kf) 
neighbors_score = neighbors_scores.mean()
neighbors_SD = neighbors_scores.std()
print("k-Nearest Neighbors Score Mean : ", neighbors_score, " Standard Deviation : ", neighbors_SD)

NNModel = MLPClassifier(hidden_layer_sizes=(100, ), max_iter= 200, random_state=42)
NN_scores = cross_val_score(NNModel, X, Y, cv=kf) 
NN_score = NN_scores.mean()
NN_SD = NN_scores.std()
print("Neural Network Score Mean : ", NN_score, " Standard Deviation : ", NN_SD)
```

## Results and Conclusion for Classification Experiments
| Model | Average Score | Standard Deviation of Scores |
|----------|----------|----------|
| Logistical Regression | 0.85656 | 0.06627 |
| Support Vector Machines | 0.92817 | 0.04760 |
| k-Nearest Neighbors | 0.94753 | 0.03937 |
| Neural Network | 0.86989 | 0.07330 |

Every model had a high coefficient of determination, and a low standard deviation (under 0.1). Their coefficients of determination were all quite close, but the one that performed best was the k-Nearest neighbors.

The model that performed best was the k-nearest neighbors model. This makes sense since k-nn usually performs best on datasets with lower dimensions, and with smaller datasets. It can handle non-linear data well, unlike logistic regression. In addition, there were a lot of clusters, especially in bp, and the binary appet variables, which k-nn works well with. That being said, I think the neural network may eventually be able to perform better if I tune it more.
