```python
# Imports section
import pandas as pd
import sklearn as sk
import warnings

# Suppress specific warnings
warnings.filterwarnings("ignore", category=sk.exceptions.ConvergenceWarning)
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

dataset = pd.read_csv("science_data_large.csv")

print("First 15 Rows:")
print(dataset.head(15))

print("\n Table Information")
print(dataset.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = dataset[['Temperature °C', 'Mols KCL']]  # Features should be in a DataFrame
y = dataset['Size nm^3']                    # Labels should be in a Series

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

model = sk.linear_model.LinearRegression()

model.fit(X_train,y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sampleDatapoint = X_test.iloc[0].values.reshape(1,-1)
predicted = model.predict(sampleDatapoint)

score = model.score(X_test,y_test)

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
print("R^2 score")
print(score)
print("\n")

#Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

# Below is the latex
# y = 866.1464 \cdot x_1 + 1032.6951 \cdot x_2 - 409391.4796
print("Coefficeints followed by intercept")
print(model.coef_)
print(model.intercept_)


```

Write the linear equation of a slime: (example equation: $E = mc^2$):  $\cdot x_1$ is Temperature and $\cdot x_2$ is Mols KCL

$h(x) = 866.1464 \cdot x_1 + 1032.6951 \cdot x_2 - 409391.4796$

Report on score and explain meaning:
The score in this model represents the R^2 function, with possible values from -inf to 1, with the closer the value to 1 meaning a more accurate regression model. It tells you how accurate the X values are in predicting the y value. Our value of R^2 = 0.85524 is pretty good as extremely high values to 1 could also be a result of overfitting.

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
num = sk.model_selection.KFold(n_splits=5,shuffle=True,random_state=42)

scores = sk.model_selection.cross_val_score(model, X, y, cv=num,scoring='r2')

print("R^2 Scores through 5 batches:")
print(scores)
print("Mean of the R^2:")
print(sum(scores)/len(scores))

# Report on their finding and their significance

```

Write findings here: The model achieved a mean R^2 score of 0.8597, indicating it explains about 86% of the variance in the target variable, demonstrating strong predictive performance. The scores across the five folds are consistent, suggesting the model generalizes well and is not overly sensitive to different data splits. Overall, this is a good result with no glaring errors.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
num = sk.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)

poly = sk.preprocessing.PolynomialFeatures(2)
X_poly = poly.fit_transform(X)

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X_poly, y, test_size=0.1, random_state=42)
polymodel = sk.linear_model.LinearRegression()
polymodel.fit(X_train, y_train)


# Perform cross-validation
scores = sk.model_selection.cross_val_score(polymodel, X_poly, y, cv=num, scoring='r2')
    
# Print model coefficients
print("Coefs:")
print(polymodel.coef_)
print("Intercepts")
print(polymodel.intercept_)

# Print Cross Val scores
print("Cross Val Scores:")
print(scores)

# Report on the metrics and output the resultant equation as you did in Part 3.
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

ckd_ds = pd.read_csv("ckd_feature_subset.csv")
#print(ckd_ds.info())
rand = 42
X = ckd_ds[['age', 'bp', 'wbcc', 'appet_poor', 'appet_good','rbcc']] 
y = ckd_ds['Target_ckd']                 


X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.1, random_state=rand)

# Logistic Regression: 

LR_model = sk.linear_model.LogisticRegression()

LR_model.fit(X_train,y_train)


# SVC

SVC_model = sk.svm.SVC(decision_function_shape='ovo')

SVC_model.fit(X_train,y_train)

# k-Nearest Neighbors

kNear_model = sk.neighbors.KNeighborsClassifier()

kNear_model.fit(X_train,y_train)

# Neural Networks
base_iterations = 1000  # Original 1000
base_activation = 'relu'  # Original ReLU
base_hidden_layers = (100,)  # Original 100 hidden layers

# Configurations to change one variable at a time
configurations = [
    {'iterations': base_iterations, 'activation': base_activation, 'hidden_layers': base_hidden_layers},
    {'iterations': 200, 'activation': base_activation, 'hidden_layers': base_hidden_layers},
    {'iterations': 500, 'activation': base_activation, 'hidden_layers': base_hidden_layers},
    {'iterations': base_iterations, 'activation': 'logistic', 'hidden_layers': base_hidden_layers},
    {'iterations': base_iterations, 'activation': 'identity', 'hidden_layers': base_hidden_layers},
    {'iterations': base_iterations, 'activation': base_activation, 'hidden_layers': (1,)},
    {'iterations': base_iterations, 'activation': base_activation, 'hidden_layers': (500,)},
]

# Iterate through configurations and train models
neurall = []
for config in configurations:
    iterations = config['iterations']
    activationfun = config['activation']
    layer_size = config['hidden_layers']
    
    # Create the neural network model
    neural = sk.neural_network.MLPClassifier(
        max_iter=iterations,
        activation=activationfun,
        hidden_layer_sizes=layer_size
    )
    
    # Train the model
    neurall.append(neural.fit(X_train, y_train))
```

## Results and Conclusion for Classification Experiments

```python
# Testing for Linear Regression Model: 

cfv = sk.model_selection.cross_val_score(LR_model, X, y, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (cfv.mean(), cfv.std()))

# Testing for SVC Model

cfv1 = sk.model_selection.cross_val_score(SVC_model, X, y, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (cfv1.mean(), cfv1.std()))

# Testing for k-Neighbors Model

cfv2 = sk.model_selection.cross_val_score(kNear_model, X, y, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (cfv2.mean(), cfv2.std()))

# Testing the neural network: 
print("Neural Networks below, first is the base network:")
tt = ["Iterations Changed:","Activation Function Changed:","Hidden Layers Changed:"]
j = 0
for i,v in enumerate(neurall):
    if i != 0 and ((i-1)%2 == 0):
        print(tt[j])
        j+=1

    cfv3 = sk.model_selection.cross_val_score(v, X, y, cv=5)
    print("%0.2f accuracy with a standard deviation of %0.2f" % (cfv3.mean(), cfv3.std()))


```
