```python
# Imports section
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score

from sklearn.svm import SVC

from sklearn.neighbors import KNeighborsClassifier

from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
slime_data = pd.read_csv("science_data_large.csv")
display(slime_data.head(15))
display(slime_data.info)
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

features = slime_data.iloc[:, :2]  # temp and moles of KCL
label = slime_data.iloc[:, 2:]     # size of slime

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

feature_train, feature_test, label_train, label_test = train_test_split(features, label, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(feature_train, label_train)
# Create a sample datapoint and predict the output of that sample with the trained model

sample =  pd.DataFrame({
    'Temperature °C': [469],
    'Mols KCL': [647]
})

prediction = model.predict(sample)
print(f"Predicted Output for Sample Datapoint: {prediction}")

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

score = model.score(feature_train, label_train)
print("The score is: ", score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

coefficients = model.coef_
intercept = model.intercept_
print("Coefficients: ",coefficients)
print("Intercept: ",intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42

# Create a KFold object with n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(LinearRegression(), features, label, cv=kf, scoring='r2')
    
# Report on their finding and their significance
print("Scores: ", scores)
print("Mean of Scores: ", np.mean(scores))
print("Standard Deviation: ", scores.std())
```

Write findings here: The score for each of the 5 splits is very close to the score of the training set. This implies that the training set is a good representation of our data. This also implies that it is unlikely for there to be many outliers in our test set.

## 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Perform k-fold cross validation (as above)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(features)

# splitting the augmented dataset of degree 2 into train and test sets
X_poly_train, X_poly_test, label_train, label_test = train_test_split(X_poly, label, test_size=0.1, random_state=42)

# fit using the training set
ok = LinearRegression()
ok.fit(X_poly_train, label_train)

intercept = ok.intercept_
coefficients = ok.coef_

kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores_poly = cross_val_score(LinearRegression(), X_poly, label, cv=kf, scoring='r2')
# Report on the metrics and output the resultant equation as you did in Part 3.

print("Scores: ", scores_poly)
print("Mean of Scores: ", np.mean(scores_poly))

print(intercept)
print(coefficients)
print(poly.get_feature_names_out())
```
