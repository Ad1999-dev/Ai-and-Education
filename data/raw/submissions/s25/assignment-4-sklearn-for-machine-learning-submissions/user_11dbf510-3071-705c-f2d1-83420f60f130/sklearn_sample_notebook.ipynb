```python
# Imports section
import pandas
import numpy
import sklearn
from sklearn import linear_model
from sklearn import neural_network
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
science_data = pandas.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
print(science_data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(science_data.info())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = science_data.drop(columns=["Size nm^3"])  # Assuming 'growth' is the label
y = science_data["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1, random_state=42, shuffle = True)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
lin_reg = sklearn.linear_model.LinearRegression()
lin_reg.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = numpy.array([[404, 309]])
prediction = lin_reg.predict(sample_data)
print("Sample Datapoint: [[404],[309]]. Prediction:", prediction[0])
# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
print("Linear Regression Score:", lin_reg.score(X_test, y_test))
# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
print("Coefficients:", lin_reg.coef_)
print("Intercept:", lin_reg.intercept_)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)  
cv_scores = sklearn.model_selection.cross_val_score(lin_reg, X, y, cv=k_fold, scoring="r2")
# Report on their finding and their significance
print("Cross-validation scores:", cv_scores)
print("Mean CV Score:", numpy.mean(cv_scores))
```

Write findings here:

The $R^2$ values across all five folds are relatively high, ranging from 0.8392 to 0.8720, which suggests to me  that the model performs well across different subsets of the data. The mean cross-validation score of 0.8568 confirms that the model generalizes well beyond the training data.

The small variation in scores suggests the model is stable, meaning its performance is not significantly affected by different data splits.

Since cross-validation helps mitigate overfitting, this result reassures us that the linear regression model is reliable and generalizable.

The mean CV score of 0.8568 is close to the original test set score of 0.8552, which means our model is not overly dependent on a particular train-test split.

The CV results confirm that the linear regression model is pretty effective. However, the consistent unexplained variance suggests that additional factors could improve the model, such as polynomial terms or additional features.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.
ckd_data = pandas.read_csv("ckd_feature_subset.csv")
X_ckd = ckd_data.drop(columns=["Target_ckd"])  
y_ckd = ckd_data["Target_ckd"]
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X_ckd, y_ckd, test_size=0.1, random_state=42)

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(kernel="rbf", random_state=42),
    "k-NN": KNeighborsClassifier(n_neighbors=5),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000, random_state=42)
}

# Perform 5-fold cross-validation and store results
for name, model in models.items():
    k_fold = KFold(n_splits=5, shuffle=True, random_state=42)  
    cv_scores = sklearn.model_selection.cross_val_score(model, X_ckd, y_ckd, cv=k_fold, scoring="accuracy")
    mean_score = numpy.mean(cv_scores)
    std_dev = numpy.std(cv_scores)
    print(f"{name}: Mean Accuracy = {mean_score}, Std Dev = {std_dev}")
```

## Results and Conclusion for Classification Experiments
| Model                | Mean Accuracy | Std Dev  |
|----------------------|--------------|----------|
| Logistic Regression | 0.8566        | 0.0663   |
| SVM                 | 0.9282        | 0.0476   |
| k-NN                | 0.9280        | 0.0524   |
| Neural Network      | 0.9351        | 0.0405   |

---

The neural network achieved the highest mean accuracy (0.9351) and had the lowest standard deviation (0.0405), indicating strong performance and stability. The SVM model came in second, with an accuracy of 0.9282, performing almost as well but with slightly higher variance.

Logistic regression had the lowest accuracy (0.8566), likely due to its linear decision boundary, which may not capture the complexity of the dataset as well as nonlinear models like SVM or neural networks.

The k-NN model performed similarly to SVM, with an accuracy of 0.9280, but had a higher standard deviation (0.0524), suggesting it was less stable across different splits of the data.

The neural network likely performed best because it can model complex, nonlinear decision boundaries better than logistic regression. The use of hidden layers allowed it to learn intricate patterns in the data that other models could not capture. Its lower standard deviation indicates that its performance was more consistent across different folds of the data.

Neural networks performed the best overall, balancing both high accuracy and low variance. SVM and k-NN were strong contenders, but k-NN had more variability. Logistic regression had the weakest performance, likely due to its linear nature.
