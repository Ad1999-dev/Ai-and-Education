```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
import numpy as np
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

ds = pd.read_csv('science_data_large.csv')
print(ds.head(15))
ds.info()
print("Number of datapoints:", ds.shape[0])
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42 

features = ds[['Temperature °C', 'Mols KCL']]
label = ds[['Size nm^3']]

X_train, X_test, y_train, y_test, = train_test_split(features, label, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX

reg = LinearRegression().fit(X_train.values, y_train.values)
print(reg.predict([[1000, 100]]))
print(reg.score(X_train.values, y_train.values))
print(reg.coef_)
print(reg.intercept_)
```

Report on score and explain meaning:

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(LinearRegression(), features.values, label.values, cv=kf, scoring='r2')
# Report on their finding and their significance
print(scores)
print(scores.mean())
```

Write findings here:
The average score of a LinearRegression on the data was 0.8597294202684646, indicating that the linear regression fits the model data very well, with the variables exhibiting a highly linear relationship.

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python
# Load the dataset. Then train and evaluate the classification models.

ckd = pd.read_csv('ckd_feature_subset.csv')
X = ckd[['age', 'bp', 'wbcc', 'appet_poor', 'appet_good', 'rbcc']]
y = ckd[['Target_ckd']]
kf = KFold(n_splits=5, shuffle=True, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

#Logistic Regression
model = LogisticRegression()
model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr = np.array([scores.mean(), scores.std()])
print(arr)

#SVC
model = SVC(kernel='linear')
model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr = np.vstack((arr, [scores.mean(), scores.std()]))
print(arr)

#K-nearest
model = KNeighborsClassifier()
model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr = np.vstack((arr, [scores.mean(), scores.std()]))
print(arr)
```

```python
#Neural Network
model = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=2000, random_state=42)

model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr = np.vstack((arr, [scores.mean(), scores.std()]))
print(arr)
```

## Results and Conclusion for Classification Experiments

The final table, with the mean and std of each classification in each row was:

From these results, we can conclude that the neural network, implemented using sklearn's MLPClassifier, performed the best, with the highest average $r^2$ score and the lowest standard deviation, indicating a consistently high performance. A clear second was the k-nearest neighbors classification, with only a 0.05 lower mean when compared to the neural network and the second lowest standard deviation. Logistic Regression and SVC both performed much worse, with half as high averages and much higher standard deviations, indicating large variability in performance. The neural network probably performed the best because it is the most complex of the 4 classification models, with the ability to recognize patterns in the data and change according to patterns. Neural networks also continually improve, with deeper layers adding onto the observations of shallower layers.

```python
# Experiments with Neural Network.
model = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=2000, random_state=42)

model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr2 = np.array([scores.mean(), scores.std()])

#number 2: 
model = MLPClassifier(hidden_layer_sizes=(20, 20), max_iter=2000, random_state=42)

model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr2 = np.vstack((arr2, [scores.mean(), scores.std()]))

#number 3:
model = MLPClassifier(activation='tanh', hidden_layer_sizes=(10, 10), max_iter=2000, random_state=42)

model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr2 = np.vstack((arr2, [scores.mean(), scores.std()]))

#number 4
model = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=200, random_state=42)

model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr2 = np.vstack((arr2, [scores.mean(), scores.std()]))

#number 5:
model = MLPClassifier(hidden_layer_sizes=(3, 3), max_iter=3000, random_state=42)

model.fit(X_train.values, y_train.values.ravel())

scores = cross_val_score(model, X.values, y.values.ravel(), cv=kf, scoring='r2')
arr2 = np.vstack((arr2, [scores.mean(), scores.std()]))
print(arr2)
```

## Results and Conclusion for Neural Network Experiments

The biggest difference for the classification was the activation parameter changing to 'tanh', the function used in the hidden layers of the network. This difference both increased the mean score by .14 and decreased the standard deviation by 0.01, meaning that the use of a hyperbolic tangent function correlated to an increase in performance. Both an increase and decrease in the hidden layer size corresponded to an increase of mean score, with the increase in hidden layer size about 3x as great as the decrease in hidden layer size. Finally, a massive decrease in the number of iterations drastically decreased the mean score and increased std, as the model was not able to fully converge.
