```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

## 1. Loading the dataset

```python
# Using pandas load the dataset
data = pd.read_csv("science_data_large.csv")

# Output the first 15 rows of the data
print("First 15 rows of the dataset:")
print(data.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print("\nDataset Information:")
data.info()
print("\nStatistical Summary:")
print(data.describe())
```

## 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data[['Temperature °C', 'Mols KCL']]
y = data['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# For grading consistency use random_state=42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model

sample_data = pd.DataFrame({
    "Temperature °C": [500],
    "Mols KCL": [300]         
})

predicted_size = model.predict(sample_data)
print("Predicted Size (nm^3):", predicted_size)

# Report the score for that model using the default score function property of the SKLearn model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print("R² Score:", score)

# Extract the coefficients and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

## 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# For grading consistency use n_splits=5 and random_state=42
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X, y, cv=kf)

# Report on their finding and their significance
print("Cross Validation Scores:", cv_scores)
print("Average CV Score:", cv_scores.mean())
```

# Part 2. Chronic Kidney Disease Prediction via Classification

Create code and markdown cells as needed to perform classification and report on your results

```python

# Load the cleaned CKD dataset
data = pd.read_csv("ckd_feature_subset.csv")

# Display the first few rows to verify the dataset structure
print("First 5 rows of the dataset:")
print(data.head())

# Define features and target variable
X = data.drop("Target_ckd", axis=1)
y = data["Target_ckd"]

# Create a 5-fold cross-validation object for consistency
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Define the classifiers
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Support Vector Machine": SVC(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(max_iter=1000, random_state=42)
}

# Dictionary to store results
results = {}

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=kf)
    results[name] = {"Mean Accuracy": scores.mean(), "Std": scores.std()}

# Convert the results to a DataFrame for a clean table display
results_df = pd.DataFrame(results).T
print("Classification Model Performance:")
print(results_df)
```

## Results and Conclusion for Classification Experiments

#### Findings and Conclusions 
##### Dataset Overview
__Dataset Structure:__ The CKD dataset contains normalized features such as age, blood pressure (bp), white blood cell count (wbcc), appetite indicators (appet_poor, appet_good), red blood cell count (rbcc), and a binary target variable (Target_ckd). The first five rows indicate the expected structure and normalization of the features.
Classification Models Performance

__Logistic Regression:__

Mean Accuracy: 85.66%

Std Dev: 6.63%

Observation: As a linear classifier, it provided a baseline performance but struggled with capturing more complex, non-linear relationships in the data.

__Support Vector Machine (SVC):__

Mean Accuracy: 92.82%

Std Dev: 4.76%

Observation: SVC improved performance by finding a robust decision boundary in the feature space, indicating non-linear separability in the data.

__K-Nearest Neighbors (KNN):__

Mean Accuracy: 92.80%

Std Dev: 5.24%

Observation: KNN performed comparably to SVC, confirming that local neighborhood information is quite informative for this classification task.

__Neural Network (MLP):__

Mean Accuracy: 93.51%

Std Dev: 4.05%

Observation: The neural network achieved the highest accuracy among the standard classifiers, suggesting that its capacity to model non-linear interactions is beneficial for this dataset.

```python

# Define different neural network configurations
nn_configurations = {
    "NN Default": MLPClassifier(max_iter=1000, random_state=42),
    "NN Two Hidden Layers": MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=3000, random_state=42),
    "NN Tanh Activation": MLPClassifier(activation='tanh', max_iter=1000, random_state=42),
    "NN Higher Learning Rate": MLPClassifier(learning_rate_init=0.01, max_iter=3000, random_state=42)
}

# Dictionary to store NN results
nn_results = {}

for name, model in nn_configurations.items():
    scores = cross_val_score(model, X, y, cv=kf)
    nn_results[name] = {"Mean Accuracy": scores.mean(), "Std": scores.std()}

# Convert the NN results to a DataFrame for display
nn_results_df = pd.DataFrame(nn_results).T
print("Neural Network Configuration Performance:")
print(nn_results_df)
```
